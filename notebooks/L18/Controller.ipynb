```python
from minesweeper import Game
import canvas_helpers as H
from view import View
from ipywidgets import HBox, Button, ToggleButton, Output
from IPython.display import display


# nur waehrend der Entwicklungsphase
err_out = Output(layout={'border': '1px solid black'})


class Controller:
    def __init__(self, game, view):
        self.game = game
        self.view = view
        self.flag_mode = False

        self.out = view.out
        self.canvas = view.mcanvas
        self.canvas.on_mouse_down(self.on_mouse_down)

        tooltip_msg = 'Aktiv: Klick = Flagge setzen/entfernen\nInaktiv: Klick = Feld aufdecken'
        self.flag_mode_button = ToggleButton(description='Flaggen-Modus',
                                             value=False,
                                             tooltip=tooltip_msg)


        self.new_game_button = Button(description='Neues Spiel')
        self.new_game_button.on_click(self.new_game_clicked)

    @err_out.capture(clear_output=True)
    def on_mouse_down(self, x, y):
        col, row = H.xy2cr(x, y, self.view.boardspec)
        if self.flag_mode_button.value:
            self.game.toggle_flag(col, row)
        else:
            self.game.reveal_cell(col, row)

    def new_game_clicked(self, _):
        self.game.new_game()
 
    def _ipython_display_(self):
        buttons_row = HBox([self.new_game_button, self.flag_mode_button])
        display(self.canvas, buttons_row, self.out, err_out)
```

```python
game = Game()
view = View(game)
controller = Controller(game, view)
controller
```
