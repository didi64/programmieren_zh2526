### Minesweeper

Ruft ein registriertes Callback `f` mit folgenden Argumenten auf:
```python
f('new_game')  # beim Start eines neuen Spiels
f('win')  # falls alle sicheren Felder aufgedeckt
f('game_over')  # falls Mine aufgedeckt
f('flag', pos=flag_pos, status=flag_status)
f('reveal', reveal=Menge_der_Felder_zu_Aufdecken)  # falls freies Feld aufgedeckt
```
