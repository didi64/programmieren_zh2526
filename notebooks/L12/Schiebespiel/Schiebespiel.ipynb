### Einfaches Schiebe-Spiel
Die Anordnung der rechts gezeigten 5 Zahlen kann auf 2 Weisen geändert werden.  
- Doppeltausch: Vertauscht die Zahlen in den durch rote Pfeile verbundenen Kreisen.
- Rotation: Verschiebt die Zahlen im blauen Kreis in Pfeilrichtung.

Beim Spielbeginn werden die Zahlen durcheinander gebracht. 
Ziel ist es, die rechts gezeigte Anordnung wieder herzustellen.  

<br>

Die Endversion unseres Spiels soll den Spielzustand auf der Leinwand darstellen.
Das Spiel soll mit der Tastatur und der Maus gesteuert werden können.
- `n`: neues Spiel starten
- *Pfeil nach oben* oder Klick auf obersten Kreis: Doppeltausch 
- *Pfeil nach rechts* oder Klick auf untersten Kreis: Rotation 

### Spiellogik separat implementieren
Unabhängig von konkreten Spiel implementieren wir in einem ersten Schritt **nur** die Spiellogik. Wir schreiben Code, der 
- den **Spielzustand** in geeigneter Form speichert, 
- **Funktionen zum Modifizieren des Spielzustandes** bereit stellt.
  Diese Funktionen sollten **nur regelkonforme** Modifikationen erlauben.
  
In einer ersten Version wird das Einhalten der Regeln oft noch dem Benutzer 
überlassen.

Für unser Schiebespiel sieht das so aus.

```python
#%%file game.py
SCRAMBLE = [2, 1, 3, 4, 5]  # wird spaeter ev. zufaellig gewaehlt
state = []  # Spielzustand wird hier gespeichert


def new_game():
    state[:] = SCRAMBLE


def swap():
    state[1], state[0] = state[0], state[1]
    state[4], state[2] = state[2], state[4]


def rotate():
    state[1:] = state[-1:] + state[1:-1]
```

Zum Testen rufen wir die Funktionen zum Ändern des Spielzustandes direkt auf und
geben dann die Variable `state`, die den  Spielzustand speichert, aus. 
Funktioniert der Code wie gewünscht, speichern wir ihn in einem File `game.py`.

```python
# neues Spiel starten
new_game()
```

```python
# Doppeltausch
swap()
```

```python
# Rotation
rotate()
```

```python
# Spielzustand anzeigen
state
```
