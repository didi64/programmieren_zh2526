### Schiebespiel mit verbesserter Darstellung
Wir nutzen ein  MultiCanvas-Widget mit 2 Layern, einem Hinter- und Vordergrund.


- Der Hintergrund wird mit  `draw_bg(hintergrund, center_x, center_y, radius)` gezeichnet.  
- Auf dem Vordergrund wird jeweils der aktuelle Spielzustand
dargestellt, durch Aufruf von `update(vordergrund, state)`.


Daszu nutzen wir die Funktionen
aus dem Notebook `Schiebespiel_helpers.ipynb`, die im File `darstellung_2.py`gespeichert wurden.

```python
import game
import darstellung_2 as D
from ipywidgets import Output
from ipycanvas import MultiCanvas
from IPython.display import display

LAYOUT = {'border': '1px solid black'}
CENTER_X, CENTER_Y, RADIUS = 100, 140, 40

# Mittelpunkte der 5 Kreise fuer die Zahlen
nodes = D.get_centers(CENTER_X, CENTER_Y, RADIUS)

out = Output(layout=LAYOUT)
mcanvas = MultiCanvas(2, width=200, height=200, layout=LAYOUT)
bg, fg = mcanvas


@out.capture(clear_output=True)
def on_key_down(key, *flags):
    print(f'key pressed: {key}')
    state_old = game.state.copy()
    if key == 'ArrowUp':
        game.swap()
    if key == 'ArrowRight':
        game.rotate()
    if key == 'n':
        game.new_game()

    if game.state != state_old:
        D.update(fg, game.state, nodes)


mcanvas.on_key_down(on_key_down)
D.draw_bg(bg, CENTER_X, CENTER_Y, RADIUS)

display(mcanvas, out)
```
