### Projektarbeit

Ziel ist es, in 3 Lektionen (23.2, 2.3, 9.3) plus Hausarbeit ein einfaches Spiel mit Python im Jupyterlab zu programmieren, welches im Canvas-Widget graphisch darstellt und mit der 
Maus und/oder Tastatur gesteuert werden kann.  

Der Programmcode ist selbständig zu erstellen. Basiert Code auf Codefragmente z.B. von Stack Overflow oder AI sind die Quelle oder der Code im Quellenverzeichnis anzugeben.
**Den Code eures Projekts müsst ihr im Detail verstehen**.

Der Programmcode ist zu dokumentieren: 
- Es sind beschreibende Variablen- und Funktionsnamen zu wählen,
- Funktionen sind mit Doc-Strings zu versehenen,
- wo sinnvoll sind inline Kommentare anzubringen.  

Weiter ist eine **Dokumentation** zu verfassen (2-3 Seiten):
- Einleitung
- Beschreibung des Spiels und der (geplanten) Implementation.
- Spezifikation eures Spiels  
  - Präzise Spielregeln (legale Züge, wofür gibt es Punkte, wann hat jemand gewonnen, ...)
- Implementation
  - Wie wird der Spielzustand im Code repräsentiert. Wie  überprüft das Programm das Einhalten der Regeln.
  - Wie wird das Spiel graphisch dargestellt?
  - Wie wird das Spiel gesteuert?  
- überblick über den Code
  - Welche Komponenten Module/Klassen/Funktionen erfüllen welche Aufgaben,
  wie spielen die verschiedenen Komponenten zusammen. Beschreibung der wichtigsten Funktionen.
- Diskussion
  - Erfüllt das Programm die Spezifikation? Wo/Wieso gibt es Abweichungen? 
  - Mögliche Verbesserungen und Erweiterungen. 
  - Schlussfolgerungen und Lehren
  - Quellenverzeichnis

**Bewertungskriterien**:
- Funktionalität: läuft Spiel fehlerfrei, sind die Regeln korrekt implementiert
- Darstellung (korrekt, übersichtlich)
- Steuerung (interaktiv mit Maus oder Tastatur)
- Umfang (eher minimalistisch oder gut ausgebaut)
- Codequalität (Struktur, Lesbarkeit, Docstrings, Effizienz)
- Dokumentation (siehe oben)
