Die Funktionen `enumerate` und `zip`
***

```python
# start legt den Startwert des ersten Elements des Tuples fest
tuple(enumerate('abc', start=1))
```

```python
# Default-Wert ist start=0
tuple(enumerate('abc'))
```

```python
numbers = range(5, 10)
for i, n in enumerate(numbers, start=5):
    print(f'i: {i}, i, n: {n}')
```

```python
items_1 = (1, 2, 3)
items_2 = 'abcd'
items_3 = 'ABCD'
for x, y, z in zip(items_1, items_2, items_3):
    print(f'x={x}, y="{y}", z="{z}"')
```

```python
def find(s, char):
    for i, c in enumerate(s):
        if c == char:
            return i
    return -1
```

```python
s = 'foobar'
find(s, 'b')
```

```python
# A (Entferne aus text alle Ziffern
digits = '0123456789'
text = 'foo23, ba2r'

letters = []
for c in text:
    if c not in digits:
        letters.append(c)
''.join(letters)
```

```python
# B (Welche Zahl liegt am naechsten bei 20)
numbers = [1, 13, 26, 30]
distances_from_20 = []
for n in numbers:
    d = abs(n - 20)
    distances_from_20.append(d)


least_dist = min(distances_from_20)
i = distances_from_20.index(least_dist)
numbers[i]
```

```python
def draw_cards_(deck, n):
    '''entferne die obersten n Karten aus deck und
       gib diese als Liste zurueck
    '''
    cards = []
    for _ in range(n):
        card = deck.pop()
        cards.append(card)
    return cards


def draw_cards(deck, n):
    '''entferne die obersten n Karten aus deck und
       gib diese als Liste zurueck
    '''
    return [deck.pop() for _ in range(n)]
```

```python
def new_deck_(ranks='23456789TJQKA', suits='♥♠♦♣'):
    '''erstellt eine Liste mit Karten mit Rang aus ranks und
       Farbe aus suits
    '''
    deck = []
    for suit in suits:
        for rank in ranks:
            card = rank + suit
            deck.append(card)
    return deck


def new_deck(ranks='23456789TJQKA', suits='♥♠♦♣'):
    '''erstellt eine Liste mit Karten mit Rang aus ranks und
       Farbe aus suits
    '''
    return [rank + suit for suit in suits for rank in ranks]
```

```python
# Liste mit Ortschaftsnamen erstellen (ignoriere die Details!)
filename = 'PLZO_CSV_LV95.txt'
SEP = ';'

with open(filename, mode='r') as f:
    lines = [line.rstrip() for line in f]

orte = sorted(set(line.split(SEP)[0] for line in lines[1:]))
orte[:3] + orte[-3:]
```

```python
# Orte der Laenge 2
[ort for ort in orte if  len(ort) == 2]
```

```python
orte_mit_berg = [ort for ort in orte if 'berg' in ort.lower()]
len(orte_mit_berg)
```

```python
orte_mit_tal = [ort for ort in orte if ' ' not in ort and ort[-3:] == 'tal']
```

```python
[ort for ort in orte if ' ' not in ort and ort.endswith('tal')]
```

```python
symbols = ''
for ort in orte:
    for c in ort:
        if c not in symbols:
            symbols = symbols + c
symbols          
```

```python
''.join(sorted(symbols))
```
