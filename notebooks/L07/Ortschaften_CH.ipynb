### Ortschaften der Schweiz
Das File `PLZO_CSV_LV95.txt` enthält eine Tabelle mit Angaben zu Ortschaften in der Schweiz. Öffne dieses File in Jupyterlab um die Spaltennamen und das Trennzeichen zu sehen.

Im Moment interessieren uns nur die Ortssnamen.

```python
# Zeilen des Files in Liste lines lesen, newlines entfernen
filename = 'PLZO_CSV_LV95.txt'
SEP = ';'

with open(filename, mode='r') as f:
    lines = [line.rstrip() for line in f]


orte_mehrfach = []

for line in lines[1:]:
    row = line.split(SEP)
    orte_mehrfach.append(row[0])

orte_mehrfach.sort()
```

```python
orte_mehrfach[:3] + orte_mehrfach[-3:]
```

```python
orte = []
current = orte_mehrfach[0]
for ort in orte_mehrfach:
    if ort != current:
        orte.append(ort)
        current = ort
```

```python
orte[:3] + orte[-3:]
```

```python
[ort for ort in orte if '/' in ort]
```

```python
[ort for ort in orte if 'sonne' in ort.lower()]
```

```python
# Liste der Ortsname, die Ziffern enthalten
digits = '0123456789'
orte_mit_ziffern = []
for ort in orte:
    for d in digits:
        if d in ort:
            orte_mit_ziffern.append(ort)
            break
orte_mit_ziffern
```

```python
def contains_XX(s):
    for x, y in zip(s, s[1:]):
        if x.isupper() and y.isupper():
            return True
```

```python
s = 'fOO'
contains_XX(s)
```

```python
orte_XX = [ort for ort in orte if contains_XX(ort)]
orte_XX[:3] + orte_XX[-3:]
```

```python
[ort for ort in orte_XX if not ort[-1].isupper()]
```

```python
kts = []
for ort in orte_XX:
    XX = ort[-2:]
    if XX.isupper() and XX not in kts:
        kts.append(XX)
kts.sort()
len(kts)
```
