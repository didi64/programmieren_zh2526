### Aufgabe 1 [3+3+3 Punkte]
1. Erstellen eine Liste `numbers` mit den Zahlen von 1 bis und mit 100, und eine Liste `cubes` mit den ersten 100 Kubikzahlen (`[1, 8, 27, ..., 970299, 1000000]`).
    Gib zur Kontrolle jeweils die ersten und letzten 3 Elemente dieser Listen aus.
2. Berechne die Summe der Zahlen in `numbers` (5050), ohne die Funktion `sum` zu benutzen. 
3. Schreibe eine Funktion `get_sum(zahlen)`, die die Summe der Zahlen in der Liste `zahlen` wie oben berechnet und zur√ºck gibt.  
Teste, ob das Quadrat von `get_sum(numbers)` gleich `get_sum(cubes)` ist (das ist der Fall).

```python
# 1
numbers = list(range(1, 101))
cubes = [n**3 for n in numbers]
print(f'numbers: {numbers[:3]}, {numbers[-3:]}')
print(f'cubes: {cubes[:3]}, {cubes[-3:]}')
```

```python
# 2a
total = 0
for n in numbers:
    total = total + n
total
```

```python
# 2b
def get_sum(zahlen):
    total = 0
    for n in zahlen:
        total = total + n
    return total
```

```python
# Funktion testen
get_sum(cubes)
```

```python
# 2c
get_sum(numbers)**2 == get_sum(cubes)
```
