### Aufgabe 3 [4 + 4 + 4 Punkte]
Verschiedene Athleten absolvierten einen Hindernisparcour.
Die Liste `finishers` enthält Tuples mit dem Namen und der Zeit der Athleten.
Die benötigte Zeit ist in der Form Minuten:Sekunden angegeben, z.B.
`'10:51'` meint 10 Minuten und 51 Sekunden. 


1. Spalte den String `zeit='10:51'` beim Doppelpunkt mit der Stringmethode `split` und
  speichere die Minuten und Sekunden als **Integer** in den Variablen `minutes` und `seconds`. Schreibe dann eine Funktion `zeit2seconds(zeit)`,
die eine Zeitangabe in Minuten und Sekunden in Sekunden umrechnet.
Teste die Funktion mit den Zeiten in der Liste `zeiten` und erstelle eine Liste
mit den zugehörigen Sekunden.
```
zeiten = ['10:51', '10:07', '9:51', '8:13', '8:45']
zeiten_in_sekunden -> [651, 607, 591, 493, 525]
```

2. Sortiere die Liste `finishers`. Erstelle neue, sortierte Listen, sortiert nach folgenden Kriterien:
      1. Alphabetisch nach Namen.
      1. Nur nach der Zeit (benutze die Funktion `zeit2seconds`).
      1. Nach der Zeit (mit Namen als Tiebreaker).  
         &nbsp; 
3. Erstelle aus den Schlüssel-Wert Paaren in `finishers` einen Dictionary `name_zeit`

   (`{'Emil': '10:51', ..., 'Hans', '2:13'}`).

   Erstelle dann aus dem Dictionary `name_zeit` sortierte Dictionaries, einmal nach
   den Namen, und einmal nach den Werten (mit den Namen als Tiebreaker). Iteriere dann mit einem For-Loop über Letzteren und produziere folgenden Output (Rangliste).

```
1. Carl, 8:13
2. Greg, 8:13
3. Hans, 8:13
4. Dani, 8:45
5. Alex, 9:51
6. Fred, 9:51
7. Bob, 10:07
8. Emil, 10:51
```

````python
zeit = '10:51'
zeiten = ['10:51', '10:07', '9:51', '8:13', '8:45']

finishers = [('Emil', '10:51'),
             ('Bob', '10:07'),
             ('Greg', '8:13'),
             ('Hans', '8:13'),
             ('Fred', '9:51'),
             ('Alex', '9:51'),
             ('Carl', '8:13'),
             ('Dani', '8:45'),
             ]
````

````python
# 1a
zeit = '10:51'
parts = zeit.split(':')
numbers = [int(part) for part in parts]
minutes, seconds = numbers
60 * minutes + seconds
````

````python
# 1b
def zeit2seconds(s):
    parts = s.split(':')
    return 60 * int(parts[0]) + int(parts[1])
````

````python
zeit2seconds(zeit)
````

````python
# 1c Funktion testen mit Zeiten in Liste zeiten
[zeit2seconds(zeit) for zeit in zeiten]
````

````python
# 2A
sorted(finishers)
````

````python
# 2B
sorted(finishers, key=lambda x: x[1])
````

````python
# 2C
sorted(finishers, key=lambda x: (zeit2seconds(x[1]), x[0]))
````

````python
# 3a Dict erstellen
name_zeit = dict(finishers)
name_zeit
````

````python
# 4b sortieren nach Namen
kv_pairs = sorted(name_zeit.items())
dict(kv_pairs)
````

````python
# 4c sortieren nach (zeit, namen)
kv_pairs = sorted(name_zeit.items(), key=lambda x: (zeit2seconds(x[1]), x[0]))
standings = dict(kv_pairs)
standings
````

````python
# 4d Rangliste
for i, (name, zeit) in enumerate(standings.items()):
    print(f'{i+1}. {name}, {zeit}')
````
