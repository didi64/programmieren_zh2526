1. Gegeben sei der String `names = 'Alice, Bob, Carl'`.
   Benutze eine For-Schleife, um daraus den String
   `result = 'Alice Bob Carl'` zu erstellen (Kommas entfernen).
   Benutze **keine** Stringmethoden. Starte mit einem leere String `result`.
   Iteriere dann Ã¼ber die Zeichen in `names` und addiere das Zeichen zu `result`, falls es kein Komma ist.
1. Gegeben sei der String `test = 'Test'`.
   Benutze eine For-Schleife, einmal mit und einmal ohne `enumerate` zu benutzen, um folgenden Output zu produzieren.
   ```
   
   Der 1. Buchstabe ist T
   ```  

````python
s = '123456789'
test = 'Test'
names = 'Alice, Bob, Carl'
````

````python
# 1a
s[:3], s[3:6], s[-3:]
````

````python
# 1b
s_new = s[:3] + s[3:6][::-1] + s[-3:]
s_new
````

````python
# 2
result = ''
for c in names:
    if c != ',':
        result = result + c
result
````

````python
# 3a
i = 0
for c in test:
    i = i + 1
    print(f'Der {i}. Buchstabe ist {c}')
````

````python
# 3b
for i, c in enumerate(test):
    print(f'Der {i+1}. Buchstabe ist {c}')
````
