### Aufgabe 5 [2+3+3+4 Punkte]
Es wird mit 5 Würfeln gewürfelt.
Ein Wurf ist ein Tuple mit 5 Zahlen von 1 bis 6, z.B. `(4,1,3,1,3)`.
Der zu einem Wurf gehörige Count-Dict ist der
Dictionary, der zu jeder Zahl angibt, wie oft diese vorkommt.  
Zum Wurf `(4,1,3,1,3)` gehört der Dict `{4: 1, 1: 2, 3: 2}` und  
zum Wurf `(1,2,3,4,5)` der Dict  `{1: 1, 2: 1, 3: 1, 4: 1, 5: 1}`.

Einen Wurf nennen wir 
- *Quartett*, falls 4 ein Wert im Count-Dict ist z.B. (2,3,2,2,2),
- *Fullhouse*, falls 2 und 3 Werte im Count-Dict sind z.B. (4,2,4,2,2),
- *Straight*, falls der Count-Dict Länge 5 hat (5 Schlüssel-Wert Paare) und nicht sowohl 1 und 6 Schlüssel sind z.B. (1,2,3,4,5) und (2,3,4,5,6).

1. Schreibe eine Funktion `wuerfle()`, die ein Tuple mit 5 Zufallszahlen von 1 bis 6 zurück gibt (benutze die Funktion `randint` im Modul `random`).
1. Schreibe eine Funktion `make_count_dict(wurf)`, die den zum Wurf
gehörigen Count-Dict zurück gibt.
1. Schreibe Funktionen `is_quartett(count_dict)`,
`is_straight(count_dict)` und `is_fullhouse(count_dict)`, die testen, ob
`count_dict` der Count-Dict eines *Quartetts*, bez. *Fullhouse* oder 
*Straight* ist.
Teste die Funktionen mit den Count-Dicts der folgenden Würfen.
    ```python
    
    wuerfe = [(1, 2, 3, 4, 5), (2, 3, 4, 5, 6), (1, 2, 4, 5, 6),
              (2, 2, 2, 4, 4), (4, 2, 2, 2, 4), (1, 1, 2, 2, 3),
              (1, 3, 1, 1, 1), (5, 6, 6, 6, 6)]
              
    ```
    Produziere folgenden Output (S steht für Straight, F für Fullhouse und Q für Quartett).
    ```
    
    (1, 2, 3, 4, 5), S: True, F: False, Q: False
    (2, 3, 4, 5, 6), S: True, F: False, Q: False
    (1, 2, 4, 5, 6), S: False, F: False, Q: False
    (2, 2, 2, 4, 4), S: False, F: True, Q: False
    (4, 2, 2, 2, 4), S: False, F: True, Q: False
    (1, 1, 2, 2, 3), S: False, F: False, Q: False
    (1, 3, 1, 1, 1), S: False, F: False, Q: True
    (5, 6, 6, 6, 6), S: False, F: False, Q: True
    
    ```
    &nbsp;
1. Würfle 10'000 Mal und zähle, wieviele Straights, Fullhouses und Quartette gewürfelt wurden (ungefähr 310 Straights, 390 Fullhouses und 190 Quartette).

````python
from random import randint


wuerfe = [(1, 2, 3, 4, 5), (2, 3, 4, 5, 6), (1, 2, 4, 5, 6),
          (2, 2, 2, 4, 4), (4, 2, 2, 2, 4), (1, 1, 2, 2, 3),
          (1, 3, 1, 1, 1), (5, 6, 6, 6, 6)]
````

````python
# 1
def wuerfle():
    return tuple(randint(1, 6) for _ in range(5))
````

````python
# Funktion testen
wurf = wuerfle()
wurf
````

````python
# 2
def make_count_dict(items):
    d = {}
    for item in items:
        d[item] = d.get(item, 0) + 1
    return d
````

````python
# Funktion testen
d = make_count_dict(wurf)
d
````

````python
# 3a
def is_quartett(count_dict):
    return 4 in count_dict.values()


def is_fullhouse(count_dict):
    return len(count_dict) == 2 and 3 in count_dict.values()


def is_straight(count_dict):
    return len(count_dict) == 5 and not (1 in count_dict and 6 in count_dict)
````

````python
is_quartett({2: 4, 1: 1})
````

````python
is_fullhouse({2: 4, 1: 3})
````

````python
is_straight({1: 1, 2: 1, 3: 1, 4: 1, 5: 1})
````

````python
# 3b
for wurf in wuerfe:
    d = make_count_dict(wurf)
    print(f'{wurf}, S: {is_straight(d)}, F: {is_fullhouse(d)}, Q: {is_quartett(d)}')
````

````python
# 4
N = 10_000

n_straight = 0
n_fullhouse = 0
n_quartett = 0

for _ in range(N):
    wurf = wuerfle()
    count_dict = make_count_dict(wurf)
    if is_quartett(count_dict):
        n_quartett = n_quartett + 1
    if is_straight(count_dict):
        n_straight = n_straight + 1
    if is_fullhouse(count_dict):
        n_fullhouse = n_fullhouse + 1

print(f'Straigts: {n_straight}, Fullhouses: {n_fullhouse}, Quartette: {n_quartett}')
````
