### Aufgabe  [3 + 4 + 3 + 2 Punkte]
Das File `words.txt` enthält Zeilen mit durch Semikolons separierten 
 Wörtern. Es soll ein File `stripped_words.txt` erstellt werden, wo
 aller Space zwischen den Wörter entfernt ist und 
 anstelle des Semikolons ein Komma als Trennzeichen verwendet wird.


1. Öffne das File `words.txt` und erstelle eine Liste `lines` mit den
   Zeilen des Files. Whitespace am Zeilenende ist zu entfernen.
   Erstelle eine Liste `samples` mit den
   ersten und letzten 3 Zeilen von `lines`.
2. Speichere die erste Zeile in einer Variable `first_line`.
   Spalte diese Zeile beim `';'` in eine Liste `words` und erstelle eine neue Liste `stripped_words`, mit den Wörtern aus `words` ohne Space.  
Verbinde dann die Wörter in `stripped_words` mit einem Komma zu einem String `stripped_line` und füge den Newline-Charakter `'\n'` ans Zeilenende an.
3. Schreibe eine Funktion `strip_line(line)`, die mit dem String `line` macht, was in 2. mit `first_line` gemacht wurde und den resultierenden String zurück gibt. Wende die Funktion `strip_line` auf jede Zeile in `lines` an
und speichere die resultierenden Zeilen in einer Liste `new_lines`.
Gib zur Kontrolle die ersten und letzten beiden Zeilen aus.


6. Schreibe die Zeilen in `new_lines` in ein File `stripped_words.txt`.

```python
fn_old = 'words.txt'
fn_new = 'stripped_words.txt'
sep_old = ';'
sep_new = ','
```

```python
# 1
with open(fn_old, mode='r') as f:
    lines = [line.rstrip() for line in f]

samples = lines[:3] + lines[-3:]
samples
```

```python
# 2a
first_line = lines[0]
first_line
```

```python
# 2b
words = first_line.split(sep_old)
words
```

```python
# 2c
stripped_words = [word.strip() for word in words]
stripped_words
```

```python
# 3a
def strip_line(line):
    words = line.split(';')
    stripped_words = [word.strip() for word in words]
    stripped_line = ','.join(stripped_words) + '\n'
    return stripped_line
```

```python
# Funktion testen
strip_line(first_line)
```

```python
# 3b
new_lines = [strip_line(line) for line in lines]
new_lines[:2] + new_lines[-2:]
```

```python
# 4
with open(fn_new, mode='w') as f:
    f.writelines(new_lines)
```
