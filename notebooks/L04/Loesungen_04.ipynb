````python
def swap_items(items, i, j):
    '''items: list
       vertauscht items[i] und items[j]
    '''
    tmp = items[i]
    items[i] = items[j]
    items[j] = tmp
````

````python
xs = [1, 2, 3, 4, 5]
````

````python
swap_items(xs, 1, 4)
xs
````

````python
def count_digits(digits):
    '''digits: str
       gibt eine Liste counts der Laenge 10 zurueck, wobei counts[i] angibt,
       wie oft die Ziffer i im String digits vorkommt
    '''
    counts = [0] * 10
    i = 0
    while i < len(digits):
        idx = int(digits[i])
        counts[idx] = counts[idx] + 1
        i = i + 1
    return counts
````

````python
def get_random_numbers(n):
    '''n: int
       gibt eine Liste mit n Zufallszahlen zw. 1 und 10*n zurueck
    '''
    numbers = [0] * n
    i = 0
    while i < n:
        numbers[i] = randint(1, 10*n)
        i = i + 1
    return numbers
````

````python
def is_sorted(items):
    '''items: list (mit vergleichbaren Elementen)
       gibt True zurueck falls die items sortiert ist, sonst False
    '''
    n = len(items)
    i = 0
    while i < n-1:
        if items[i] > items[i+1]:
            return False
        i = i + 1
    return True
````

````python
xs = [1, 2, 3, 4, 3]
is_sorted(xs)
````

````python
def bubble_at(items, i):
    '''items: list (mit vergleichbaren Elementen)
       i: int (0 < i < len(items))
       vertausche items[i] und items[j] falls items[i] > items[j]
    '''
    if items[i] > items[i+1]:
        tmp = items[i]
        items[i] = items[i+1]
        items[i+1] = tmp
````

````python
xs = [4, 5, 3, 1, 2]
bubble_at(xs, 2)
xs
````

````python
def bubble_up(items):
    '''items: list (mit vergleichbaren Elementen)
       vertausche von links nach rechts benachbarte Elemente,
       falls das linke groesser ist als das rechte.
    '''
    n = len(items)
    j = 0
    while j < n-1:
        bubble_at(items, j)
        j = j + 1
````

````python
xs = [4, 5, 3, 1, 2]
````

````python
bubble_up(xs)
xs
````

````python
def bubble_sort(items):
    '''items: list (mit vergleichbaren Elementen)
       sortiert items aufsteigend
    '''
    n = len(items)
    i = 0
    while i < n-1:
        bubble_up(items)  # eigentlich genuegt ein bubble_up der Liste [items[0,...,items[n-1-i]]
        i = i + 1
````

````python
xs = [5, 4, 3, 2, 1]
````

````python
bubble_sort(xs)
xs
````

### Indexing und Slicing

````python
word = 'Achtung'
word[-3:]  # die letzten 3 Buchstaben von word
````

````python
word[-3:] == 'ung'  # endet word mit 'ung'
````

````python
def select(items, suffix):
    result = []
    n = len(suffix)
    i = 0
    while i < len(items):
        item = items[i]
        if item[-n:] == suffix:
            result = result + [item]
        i = i + 1
    return result
````

````python
words = ['Übung', 'Haus', 'Atmung', 'Begrüssung',
         'Hund', 'Ehrung', 'Einigung', 'Endung',
         ]

select(words, 'ung')
````
