### Recap Lektion 3
- **Tuples**: Ein Datentyp zum Speichern verschiedener Werte.
  Ist `items` ein Tuple, so kann man wie bei Strings mit `items[0]`
  ein Element herausgreifen und mit `len(items)` die L&auml;nge erfragen.
  Tupel kann man mit `+` zusammenh&auml;ngen, mit `*` vervielfachen und
  mit `==` und `<`, `<=`, ... vergleichen.
  ```python
  ()  # leeres Tuple
  (1,) # einelenemtige Tuple
  ('Max', 'Muster', 2004)  # 3-elementiges Tuple
  ```
- **f-Strings**: ein String mit Platzhaltern der Form `{<Ausdruck>}`.
  Der Ausdruck `<Ausdruck>` wird ausgewertet, in einen String umgewandelt und an die Stelle von `{<Ausdruck>}` gesetzt.
    ```python
    x = 2
    y = 3
    f'Die Summe von {x} und {y} ist {x+y}.'
    ```
- **Triple-quoted-Strings**.
  ```python
  '''\
  1. Zeile
  2. Zeile\
  '''
  ```
  Ein `\` **als letztes Zeichen einer Zeile** zeigt an, dass der String ohne Zeilenumbruch auf der n√§chsten Zeile weitergeht.
- **Funktionen**:
    ```python
        def add(x, y):
            '''x: int
               y: int
               gibt die Summe von x und y zurueck
            '''
            return x + y
    ```
    Ist die Funktion `add` in einem `.py`-File (z.B. `my_functions.py`) im Ordner `modules` definiert,
    so kann sie nach dem Import (`import my_functions`) dieses Files mit `my_functions.add(1, 2)` aufgerufen werden.  

````python
numbers = (1, 2, 3)  # Tupel erstellen
numbers = numbers + (99,)  # Tupel zusammenhaengen

len(numbers), numbers[3]  # Werte als Tupel anzeigen (da Klammern optional)
````

````python
characters = tuple('hallo')  # Strinig in Tuple umwandeln
characters = characters + ('!',)  # Tuple hinzufuegen
characters
````

````python
# Tupel umdrehen (fast wie String umdrehen)
items = ('Alice', 'Bob', 'Carl')
i = 0
items_reversed = ()

while i < len(items):
    items_reversed = (items[i],) + items_reversed  # beachte das Komma!
    i = i + 1

items_reversed
````

````python
def add(x, y):
    '''x: int
       y: int
       gibt die Summe von x und y zurueck
    '''
    return x + y
````
