```python
def update_tuple(items, i, val):
    '''items: tuple
       i: int 
       gibt ein tuple zurueck, wo items[i] = val
    '''
    res = ()  
    j = 0
    while j < len(items):
        if i == j:
            res = res + (val,)
        else:
            res = res + (items[j],)
        j = j + 1
    return res
```

```python
counts = (0,) * 10
update_tuple(counts, 7, 15)
```

```python
digits = (1, 1, 1, 2, 2, 2, 3, 2, 2, 3, 1, 9)
counts = (0,) * 10

n = len(digits)
i = 0
while i < n:
    j = digits[i]
    counts = update_tuple(counts, j, counts[j] + 1)
    i = i + 1

counts
```

```python
def show_counts(counts):
    i = 0
    while i < len(counts):
        n = counts[i]
        if n > 0:
            print(f'Ziffer {i} kommt {n} Mal vor')
        i = i + 1
```

```python
show_counts(counts)
```

```python
digits = (1, 1, 1, 2, 2, 2, 3, 2, 2, 3, 1)
counts = [0] * 10

n = len(digits)
i = 0
while i < n:
    j = digits[i]
    counts[j] = counts[j] + 1
    i = i + 1

counts
```
