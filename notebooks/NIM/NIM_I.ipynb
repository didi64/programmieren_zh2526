### Das Spiel **NIM**:
Gegeben sind Haufen mit Steinen.
- 2 Spieler entfernen abwechselnd ein oder mehrere Steine von einem der Haufen.
- Wer den letzten Haufen abr&auml;umt hat gewonnen.
***

### NIM programmieren
Der **Spielzustand** wird in einer Liste
```python
heaps = [3, 4, 5, 6]
```
gespeichert. $3,4,5$ und $6$ sind die Anzahl Steine auf den Haufen.  
Ein **Zug** wird als Tuple `(Haufen, Anzahl)` gespeichert. (2, 3) bedeutet, dass
vom 2. Haufen 3 Steine entfernt werden sollen. Unser NIM-Programm soll Folgendes tun:

- Solange noch Steine da sind, fragt das Programm abwechselnd die Spieler nach ihrem Zug.
Der Spielzustand wird aktuallisiert und dargestellt, das Zugrecht wird weitergegeben.
- Sind alle Steine weg, wird der Gewinner ausgegeben.

***
Eine erste Version unseres Programms k&ouml;nnte etwa so aussehen.
```python
heaps = [3, 4, 5, 6]  # Spielzustand/Anzahl Steine auf Haufen
players = ['Alice', 'Bob']
ptm = 0  # Player to move, 0 oder 1

Zeige Spielzustand an
Solange es noch Haufen mit Steinen hat:
    gib Name des Spielers am Zug aus
    Frage nach Zug -> (Haufen, Anzahl) 
    Update Spielzustand
    Zeige Spielzustand an
    Gib Zugrecht an nächsten Spieler weiter
   
gib Gewinner aus
```

Um daraus Python Code zu machen brauchen wir Hilfsfunktionen:
- `show(heaps)`: stelle die Haufen textlich dar, z.B. so:
 ```
 1) **
 2) ****
 3) *****
  ```
- `ask_for_move()`: Frage den Spieler, von welchem Haufen er wieviele Steine wegnehmen möchte und
  gib diese Information als Tuple `move` (Haufen, Anzahl) zurück.

- `update_heaps(heaps, move)`: passt die Liste `heaps` entsprechend dem gemachten Zug an.

Um totale Anzahl Steine auf den Haufen liefert uns die built-in Funktion `sum`.

````python
from nim import show, ask_for_move, update_heaps
````

````python
heaps = [1, 2, 3]  # Spielzustand
players = ['Alice', 'Bob']
ptm = 0  # Player to move, 0 oder 1

show(heaps)
while sum(heaps) != 0:
    print(f'{players[ptm]} ist am Zug')
    move = ask_for_move()
    update_heaps(heaps, move)
    show(heaps)
    ptm = 1 - ptm

print(f'{players[1-ptm]} hat gewonnen!')
````

````python
def play_nim(heaps, players):
    ptm = 0  # Player to move, 0 oder 1
    show(heaps)
    while sum(heaps) != 0:
        print(f'{players[ptm]} ist am Zug')
        move = ask_for_move()
        update_heaps(heaps, move)
        show(heaps)
        ptm = 1 - ptm

    print(f'{players[1-ptm]} hat gewonnen!')
````

````python
heaps = [1, 2, 3]
players = ['Alice', 'Bob']
play_nim(heaps, players)
````
