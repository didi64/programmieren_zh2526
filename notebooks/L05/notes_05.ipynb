```python
def get_table(lines, indices):
    '''lines: str (Zeilen kommaseparierter Werte)
       indices: tuple[int]
       gibt eine Tabelle mit den Spaltennummern indices zurueck
    '''
    result = []
    for line in lines:
        items = line.split(',')
        person = tuple(items[i] for i in indices)
        result.append(person)
    return result
```

```python
def get_head_and_tail(line, sep):
    if line[-1] != SEP:
        line = line + SEP
        print('Trennzeichen hinzugefuegt', line)
    n = len(line)
    i = 0
    while i < n and line[i] != SEP:
        i = i + 1
    head = line[:i]
    tail = line[i+1:]
    return (head, tail)
```
