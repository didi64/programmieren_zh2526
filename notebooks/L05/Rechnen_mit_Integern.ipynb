```python
line = '2, 3'
i = 0
n = len(line)

while i < n:
    print(line[i])
    i = i + 1
```

```python
line = '20, 30'
i = 0
n = len(line)

while i < n:
    char = line[i]
    if char != ' ' and char != ',':
        print(char)
    i = i + 1
```

```python
line = '20, 30,'
i = 0
n = len(line)

zahl = ''
while i < n:
    char = line[i]
    if char != ' ' and char != ',':
        zahl = zahl + char
    elif char == ',':
        print(zahl)
        zahl = ''
    i = i + 1
```
