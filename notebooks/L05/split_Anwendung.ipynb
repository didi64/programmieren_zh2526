*Index*, *User Id*, *First Name*, *Last Name*, *Sex*, *Email*, *Phone*, *Date of birth*, *Job Title*.

Die Tabelle besteht aus Tupel (Zeilen).
Jedes Tupel (Zeile) ist selber ein Tupel mit Vorname, Name und Email.  

**Vorgehen**:
- Erstelle einen String mit dem Inhalt des Files `personen.txt`.
  Achte darauf, keine zusätzlichen Zeilenumbrüche einzufügen.
- Benutze unsere Funktion `split`, um diesen String in die einzelnen Zeilen zu splitten (splitte beim Zeilenumbruchzeichen `\n`).
- Splitte nun jede Zeile beim Komma und greife Vornamen, Namen und Email heraus.
