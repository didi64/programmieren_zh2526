```python
def select(items, suffix):
    n = len(suffix)
    result = []
    for item in items:
        if item[-n:] == suffix:
            result = result + [item]
    return result
```

```python
words = ['Übung', 'Haus', 'Atmung', 'Begrüssung',
         'Hund', 'Ehrung', 'Einigung', 'Endung',
         ]

select(words, 'ung')
```

```python
def count(heaps):
    '''heaps: list[int]
       gibt die Summe der Zahlen in heaps zurueck
    '''
    total = 0
    for heap in heaps:
        total = total + heap
    return total
```

```python
heaps = [3, 4, 5, 6]
anzahl = sum(heaps)
print('Anzahl:', anzahl)
```
