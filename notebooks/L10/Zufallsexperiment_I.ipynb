### Wie oft kommen die einzelnen Wurfbilder vor
Wie oft kommen 
> *Straight*, *Yahtzee* (5 gleiche), *Quads*, *Fullhouse*, *Triple*, *two pairs*, one pair und *nothing*

beim Würfeln mit 5 würfeln vor?  

Wir würfeln und testen dann die Wurfbilder in obiger Reihenfolge, bis das passende gefunden ist.
Wenn wir auf *Fullhouse* testen, wissen wir bereits, dass 
der Test auf *Quads* negativ ausgefallen ist.
Es genügt prüfen, dass die Würfel genau 2 Zahlen zeigen.

Wenn wir  auf *Triple* testen, wissen wir bereits, dass 
der Test auf *Quads* und *Fullhouse* negativ ausgefallen ist. 
Es genügt dann zu testen, ob 3 Würfel die gleiche Zahl zeigen.

Beim Test auf *nothing* können wir immer mit `True` antworten, den mit *nothing* meinen wir ja, dass es sich beim Wurfbild um keines des vorangehenden handelt.

```python
from random import randint


def wuerfle(n=5):
    return tuple(randint(1, 6) for _ in range(n))


def make_count_dict(items):
    d = {}
    for item in items:
        d[item] = d.get(item, 0) + 1
    return d
```

```python
wurf = wuerfle()
make_count_dict(wurf)
```

```python
def is_straight(count_dict):
    return len(count_dict) == 5 and not (1 in count_dict and 6 in count_dict)


def is_yahtzee(count_dict):
    return 5 in count_dict.values()


def is_quads(count_dict):
    return 4 in count_dict.values()


def is_fullhouse(count_dict):
    return len(count_dict) == 2


def is_triple(count_dict):
    return 3 in count_dict.values()


def is_two_pairs(count_dict):
    return len(count_dict) == 3


def is_one_pair(count_dict):
    return len(count_dict) == 4


def is_nothing(count_dict):
    return True
```

```python
wurfbild_test = {'Yahtzee': is_yahtzee,
                 'quads': is_quads,
                 'fullhouse': is_fullhouse,
                 'triple': is_triple,
                 'two pairs': is_two_pairs,
                 'one pair': is_one_pair,
                 'straight': is_straight,
                 'nothing': is_nothing,
                 }


def count_wurfbilder(n=100_000):
    bild_count = {bild: 0 for bild in wurfbild_test}
    for _ in range(n):
        wurf = wuerfle(5)
        count_dict = make_count_dict(wurf)
        for bild, test in wurfbild_test.items():
            success = test(count_dict)
            if success:
                bild_count[bild] = bild_count[bild] + 1
                break

    return bild_count
```

```python
counts = count_wurfbilder()
counts = dict(sorted(counts.items(), key=lambda x: x[1]))
counts
```

```python
total = sum(v for v in counts.values())
total
```
