### Einkäufe
Nachstehender Dictionary speichert zu jeder Product-ID den Namen des zugehörigen Produktes und dessen Preis
```python
pid_prodPrice = {1: ('Brot', 2.30),
                 2: ('Milch', 1.50),
                 3: ('Schokolade', 2.85),
                 4: ('Buendnerfleich', 8.90),
                 5: ('Eier', 4.80),
                 }

cid_pids = {1561: (1, 2, 5, 4),
            1085: (3, 2, 2),
            2034: (5, 3, 3, 1, 1),
            2021: (5, 3, 4, 4, 1)}
```
1. Gib die Produkte mit Preis aus, die der Kunde mit Customer-ID `1561` gekauft hat.
   ```
    Produktname: Brot, Preis: 2.3
    Produktname: Milch, Preis: 1.5
    Produktname: Eier, Preis: 4.8
    Produktname: Buendnerfleich, Preis: 8.9
   ```
2. Erstelle eine Liste mit den Preisen der Produkte, die dieser Kunde gekauft hat und berechne die Summe dieser Preise.
3. Erstelle einen Dictionary `cid_total`, der jeder Customer-ID den totalen Warenwert der gekauften Produkte zuordnet. Sortiere diesen Dict absteigend nach dem Warenwert.
   ```python
   {2021: 27.75, 1561: 17.5, 2034: 15.1, 1085: 5.85}
   ```
   
4. Erstelle einen Dict `cid_npids`, der jeder Customer-ID eine Liste mit Tuple `(pid, n)` zuordnet.
   Jedes Tuple enthält eine Produkt_ID und wie oft dieses Produkt vom Kunden gekauft wurde.
   ```
    {⋮
     2034: [(1, 2), (3, 2), (5, 1)],
     2021: [(1, 1), (3, 1), (4, 2), (5, 1)]}
   ```
5. Erstelle einen Dict `cid_nproducts`, der jeder Customer-ID eine Liste mit Tuple `(Produktname, n)` zuordnet.
   Jedes Tuple enthält einen Produktnamen und wie oft dieses Produkt vom Kunden gekauft wurde.
   ```
   {⋮
    2034: [('Brot', 2), ('Schokolade', 2), ('Eier', 1)],
    2021: [('Brot', 1), ('Schokolade', 1), ('Buendnerfleich', 2), ('Eier', 1)]}
   ```

````python
pid_prodPrice = {1: ('Brot', 2.30),
                 2: ('Milch', 1.50),
                 3: ('Schokolade', 2.85),
                 4: ('Buendnerfleich', 8.90),
                 5: ('Eier', 4.80),
                 }

cid_pids = {1561: (1, 2, 5, 4),
            1085: (3, 2, 2),
            2034: (5, 3, 3, 1, 1),
            2021: (5, 3, 4, 4, 1)}
````

````python
cid = 1561
pids = cid_pids[cid]
for pid in pids:
    name, price = pid_prodPrice[pid]
    print(f'Produktname: {name}, Preis: {price}')
````

````python
prices = [pid_prodPrice[pid][1] for pid in pids]
sum(prices)
````

````python
cid_total = {}
for cid, pids in cid_pids.items():
    prices = [pid_prodPrice[pid][1] for pid in pids]
    total = sum(prices)
    cid_total[cid] = total

cid_total
````

````python
dict(sorted(cid_total.items(), key=lambda x: x[1], reverse=True))
````

````python
def make_count_dict(items):
    d = {}
    for item in items:
        d[item] = d.get(item, 0) + 1
    return d
````

````python
 make_count_dict('abcababca')
````

````python
cid_npids = {}
for cid, pids in cid_pids.items():
    count_dict = make_count_dict(pids)
    np_pairs = sorted(count_dict.items())
    cid_npids[cid] = np_pairs

cid_npids
````

````python
cid_nproducts = {}
for cid, np_pairs in cid_npids.items():
    np_pairs = [(pid_prodPrice[pid][0], n) for pid, n in np_pairs]
    cid_nproducts[cid] = np_pairs
    
cid_nproducts
````
