````python
def choices(items, k=5):
    results = []
    n = len(items) - 1
    for _ in range(k):
        i = randint(0, n)
        item = items[i]
        results.append(item)
    return results
````

````python
def unique_choices(items, k=5):
    results = set()
    n = len(items) - 1
    while len(results) < k:
        i = randint(0, n)
        item = items[i]
        results.add(item)
    return tuple(results)
````

````python
unique_choices(RANKS)
````

````python
def get_suits(n=5):
    suits = []
    for _ in range(5):
        n = len(SUITS)
        i = randint(1, n-1)
        suit = SUITS[i]
        suits.append(suit)

    return suits
````

````python
def get_ranks(n=5):
    ranks = []
    for _ in range(5):
        n = len(RANKS)
        i = randint(1, n-1)
        rank = RANKS[i]
        ranks.append(rank)

    return ranks
````

````python
def get_hand(n=5):
    hand = []
    ranks = get_ranks()
    suits = get_suits()
    for i in range(n):
        card = ranks[i] + suits[i]
        hand.append(card)

    return hand
````
