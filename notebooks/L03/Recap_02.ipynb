### Recap Lektion 2
- Wir kennen die Typen `int` (Integer), `float` (Fliesskommazahlen), `str` (Strings), und `bool` (Booleans/Wahrheitswerte).
- Mit `int(x)` , `float(x)` , `str(x)` , `bool(x)` k&ouml;nnen wir versuchen, `x` in den entsprechenden Typ umzuwandeln. Jeder Typ, den wir kennen l&auml;sst sich in einen String und einen Wahrheitswert umwandeln.
- Mit `input('Bitte gib etwas Text ein:')` k&ouml;nnen wir den Benutzer nach Text fragen.
  Fragen wir nach einer Zahl, so ist die Eingabe (ein String) noch in einen Integer oder Float umzuwandeln.
- Einr&uuml;cken ist Teil der Syntax von Python (siehe Aufgabe).

```python
zahl = input('Durch 3 teilbare ganze Zahl:')
n = int(zahl)  # Eingabe in int umwandeln

while n % 3 != 0:
    msg = str(n) + ' ist nicht durch 3 teilbar. Versuche es nochmals.'
    n = int(input(msg))

print('Ja,', n, 'ist', n // 3, 'mal 3')
```

```python
prompt1 = 'Wort, das mit einem Buchstaben beginnt:'
prompt2 = 'Deine Eingabe beginnt nicht mit einem Buchstaben! Nochmals:'

word = input(prompt1)
ok = word and 'A' <= word[0] and word[0] <= 'z'  # word ist nicht leer, word[0] zw. 'A' und 'z'

while not ok:
    word = input(prompt2)
    ok = word and 'A' <= word[0] and word[0] <= 'z'

print('Danke fÃ¼r deine Eingabe:', word)
```
