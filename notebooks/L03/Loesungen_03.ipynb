### Funktionen schreiben

```python
def contains_character(text, character):
    '''text: str
       character: str of length 1
       return: bool
       returns True if the text contains the symbol character
    '''
    n = len(text)
    i = 0
    contains_character = False

    while i < n and not contains_character:
        if text[i] == character:
            contains_character = True
        i = i + 1

    return contains_character
```

```python
contains_character('foobar', 'x')
```

```python
def count_character(text, character):
    '''text: str
       character: str of length 1
       return: int
       returns the number of times char occurs in s
    '''
    n = len(text)
    i = 0
    count = 0

    while i < n:
        if text[i] == character:
            count = count + 1
        i = i + 1

    return count
```

```python
count_character('foobar', 'o')
```

```python
def split(word, m):
    '''word: str
       m: int with 0<= m <len(word)
       return: tuple[str, str]
       returns a tuple (head, tail), where
       head consists of the first m characters of word
       and tail of the remaining characters
    '''
    head = ''
    tail = ''

    n = len(word)
    i = 0

    while i < m:
        head = head + word[i]
        i = i + 1

    while i < n:
        tail = tail + word[i]
        i = i + 1

    return head, tail
```

```python
split('foobar', 3)
```

```python
def numberlength(n):
    '''n: int
       return: int
       returns the number of digits of a number, or 0 if the number is 0
    '''
    n_digits = 0
    while n > 0:
        n = n // 10
        n_digits = n_digits + 1

    return n_digits
```

```python
numberlength(12345)
```

```python
def contains_digit(n, digit):
    '''n: int
       digit: int (a single digit)
       return: bool
       returns True if n contains digit
    '''
    if n == 0 and digit == 0:
        return True

    found = False
    while n > 0 and not found:
        if digit == n % 10:
            found = True
        n = n // 10

    return found
```

```python
contains_digit(123, 5)
```

```python
def is_prime(n):
    '''n: int
       return: bool
       returns True if n is prime
    '''
    is_prime = True
    i = 2
    while i*i <= n and is_prime:
        if n % i == 0:
            is_prime = False
        i = i + 1

    return is_prime
```

```python
is_prime(112)
```

### if-elif-else

```python
def sign(x):
    if x < 0:
        return -1
    elif x ==0:
        return 0
    else:
        return 1
```

```python
sign(-2), sign(0), sign(1)
```

```python
def fizzbuzz(n):
    if n % 3 == 0 and n % 5 == 0:
        return 'fizzbuzz'
    elif n % 3 == 0: 
        return 'fizz'
    elif n % 5 == 0:
        return 'buzz'
    else:
        return str(n)
```

```python
values = ('1', '2', 'fizz', '4', 'buzz', 'fizz', '7', '8', 'fizz' , 'buzz', 
          '11', 'fizz', '13', '14', 'fizzbuzz', '16', '17', 'fizz', '19', 'buzz', 
          'fizz', '22', '23', 'fizz', 'buzz', '26', 'fizz', '28', '29', 'fizzbuzz')

i = 0
while i < len(values) and values[i] == fizzbuzz(i+1):
    i = i + 1

if i < len(values):
    print(f'first error at {i+1}: got {fizzbuzz(i+1)}, expected {values[i]}.')
else:
    print('test passed')
```
