### if-Anweisungen (allgemeine Form)

`if`-Anweisung erm&ouml;glichen es, den Ablauf eines Programms zu steuern, indem
gewisse Codebl&ouml;cke in Abh&auml;ngigkeit einer Bedingung ausgef&uuml;hrt werden. Einer `if`-Anweisung k&ouml;nnen optional eine oder mehrere `elif`-Anweisungen folgen, abschliessend kann eine `else`-Anweisung stehen.


```python
if  <Ausdruck>: 
    <Anweisungen>
elif <Ausdruck>:  # optional eine oder mehrere elif-Bloecke
    <Anweisungen>
â‹®
elif <Ausdruck>:
    <Anweisungen>
else:  # optional ein abschliessender else-Block
    <Anweisungen>
```


Ist der Ausdruck  nach `if` *wahr*, wird der nachfolgende Codeblock ausgef&uuml;hrt und dann ans Ende des ganzen Blocks gesprungen. Ist der Ausdruck *false*, wird zum n&auml;chsten `elif`-Block gesprungen.
Gibt es keinen, wird der Code im `else`-Block ausgef&uuml;hrt.

Ist der Ausdruck nach `elif`*true*, wird der nachfolgende Codeblock ausgef&uuml;hrt und dann ans Ende des ganzen Blocks gesprungen.
Ist der Ausdruck *false*, wird zum n&auml;chsten `elif`-Block gesprungen. 
Gibt es keinen, wird der Code im `else`-Block ausgef&uuml;hrt.

````python
#  aendere die Werte von x und y und teste, ob der richtige Text ausgegeben wird
x = 2
y = 3

if x > y:
    print(f'{x} ist groesser als {y}')
elif x == y:
    print(f'{x} ist gleich {y}')
else:
    print(f'{y} ist groesser als {x}')
````

````python
# aendere den Wert der Tuples numbers, um alle Faelle zu testen!
numbers = ()


if not numbers:  
    print('keine Zahlen')
elif len(numbers) == 1:
    print(f'Nur die Zahl {numbers[0]}')
elif len(numbers) == 2:
    print(f'Nur die beiden Zahlen {numbers[0]} und {numbers[1]}')
else:
    n = len(numbers)
    print(f'{n} Zahlen:  {numbers[0]}, ..., {numbers[n-1]}')
````
