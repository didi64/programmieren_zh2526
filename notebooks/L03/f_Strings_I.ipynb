### f-Strings
Ein f-String ist ein String mit Ausdr&uuml;cken in geschweiften Klammern `{<Ausdruck>}`, sog. Platzhalterklammern.

```python
f'Die Summe von {x} und {y} ist {x+y}.' 
```

Der **Wert des Ausdrucks** in der Klammer wird in einen **String umgewandelt** und tritt an die Stelle des Platzhalters. Damit der Interpreter weiss, dass die Ausdr&uuml;cke in den Klammern auszuwerten sind, ist dem String ein `f` voranzustellen. 
Ein Klammerpaar `{}` ohne Ausdruck verursacht einen Fehler.

**Bemerkung**: Innerhalb eines f-Strings wird `{{` als `{` und `}}` als `}` dargestellt. 

````python
f'Die Summe von {1} und {2} ist {1+2}.'  # f-String
````

````python
x, y = 2, 3
s = f'Die Summe von {x} und {y} ist {x+y}.'  # f-String
print(s)
````

````python
f'leeres Klammerpaar {}'
````

````python
name = 'Bob'
alter = 24
print(f'{name} ist {alter} Jahre alt.')
print(f'Nächstes Jahr wird {name} {alter+1} Jahre alt sein.')
````

````python
text = 'test'
f'f-String mit geschweiften Klammern:  {{text}}'
````

````python
text = 'test'
f'f-String mit doppelten geschweiften Klammern:  {{{{text}}}}'
````
