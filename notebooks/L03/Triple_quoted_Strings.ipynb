### Nachtrag zu Stringliteralen, Triple-quoted Strings
Wie haben Stringliterale eingef&uuml;hrt als
>in Anführungszeichen gesetzter Text wie 'hallo' oder "hallo", der das Anführungszeichen nicht enthält.

Sieht der Python-Interpreter ein Stringliterale erzeugt er einen String.  
`'hallo'` und `"hallo"` erzeugen den gleichen String. Ich bevorzuge *single quotes* (`'`) gegen&uuml;ber
*double quotes* (`"`).  

**Sonderzeichen und non-printable Characters**:  
Wie erzeugt man einen String, der Anf&uuml;hrungszeichen enth&auml;lt?
```python
'Er sagt: "Hallo".' # ok
"Er sagt: 'Hallo'." # ok
'Er fragt: "Wie geht's"?' # nicht ok
```

Begegnet der Interpreter `'Er fragt: "Wie geht's"?'`, so sieht er 
1. ein Stringliteral `'Er fragt: "Wie geht'` gefolgt von
1. einem Variabelnamen `s`,
1. gefolgt von einem angefangenen Stringliteral `"?'` (schliessendes `"` fehlt).  

Hier kommt der **Backslash `\`** als **Escape-Charakter** ins Spiel.  
Ein Backslash gefolgt von **bestimmten** weitere Zeichen wird als ein **einziges Zeichen** interpretiert. 
Jupyterlab weist auf ung&uuml;ltige Kombinationen hin.
Auf diese Weise kann man Strings erzeugen, die Anf&uuml;hrungszeichen und Backslashes und
non-printable Characters wie *newline* und *tab* enthalten:  

- `\'` single quote,
- `\"` double quote,
- `\\` Backslash,
- `\n` newline (Zeilenumbruch)
- `\t` Tabulator

````python
'Er fragt: "Wie geht's"?'
````

````python
'Er fragt: "Wie geht\'s"?'  # \' ist ein ' innerhalb Singlequotes
````

````python
s = 'a\tb\tc\td'  # tab-separierte Buchstaben
print(s)
````

````python
s = '\' (single quote), \" (double quote), \\ (a single backslash)'
print(s)
````

````python
print('\\\\ (a double backslash)')
````

### Triple-quoted Stringliterale
Einen String kann man auch erzeugen, indem man Text in
sog. triple-Quotes (`'''` oder `"""`) einschliesst.  Der Text kann sich über mehrere Zeilen erstrecken.

Ist das letzte Zeichen einer Zeile ein Backslash, so wird die folgende Zeile **ohne Zeilenumbruch** an den bestehenden Text angefügt.


```python
langer_string_mit_3_zeilenumbrüchen = '''
erster Teil
zweiter Teil
'''

langer_string_mit_ohne_zeilenumbrüchen = '''\
erster Teil \
zweiter Teil\
'''
```
***

````python
langer_string_mit_3_zeilenumbrüchen = '''
erster Teil
zweiter Teil
'''

langer_string_mit_ohne_zeilenumbrüchen = '''\
erster Teil \
zweiter Teil\
'''
````

````python
langer_string_mit_3_zeilenumbrüchen
````

````python
langer_string_mit_ohne_zeilenumbrüchen
````
