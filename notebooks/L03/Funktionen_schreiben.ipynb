### Vorgehen zum Schreiben einer Funktion
Code ist schnell geschrieben, die Fehlersuche dagegen ist oft sehr zeitintensiv.
Deshalb versucht man oft, den **Funktionsbody** aus bereits **getesteten** Codefragmenten zusammenzusetzen.  

Exemplarisch erstellen wir eine Funktion, welche ein Wort umdreht. 
Um die Funktion bei Bedarf verwenden zu können, speichern wir diese dann in einem `.py`-File `my_functions.py` im Ordner `modules`.  
Dabei st&uuml;tzen wir uns auf folgenden Code.

````python
# Folgender Code dreht ein Wort um
word = 'abcd'

reversed_word = ''
n = len(word)
i = 0

while i < n:
    reversed_word = word[i] + reversed_word
    i = i + 1

reversed_word
````

Mit obigem Code k&ouml;nnen wir irgend einen String umdrehen.  
Wir m&uuml;ssen den String einfach
anstelle von `'abcd'` der Variable `word` zuweisen und dann die Zelle ausf&uuml;hren. 
Der umgedrehte String ist dann in der Variable `reversed_word` gespeichert.    

Um eine Funktion zu erhalten, welche diese Aufgabe löst, 
definieren wir eine Funktion, die `word` als Argument nimmt,
obigen Code, der `reversed_word` berechnet als Funktionsbody enth&auml;lt und dann
`reversed_word` zur&uuml;ck gibt.  
Die Funktion k&ouml;nnen  wir ebenfalls `reversed_word` nennen.
Das verursacht kein Problem, da der Funktionsname eine globale Variable und 
der Rückgabewert eine lokale Variable ist.

```python
def reverse_word(word):
    reversed_word = ''
    n = len(word)
    i = 0
    
    while i < n:
        reversed_word = word[i] + reversed_word
        i = i + 1
    
    reversed_word
    return reversed_word
```

Dann fügen wir noch den DocString ein und testen die Funktion mit einigen W&ouml;rtern.

````python
def reverse_word(word):
    '''word: str
       return: str
       gibt das umgedrehte Wort zurueck
    '''
    reversed_word = ''
    n = len(word)
    for i in range(n):
        reversed_word = word[i] + reversed_word

    return reversed_word
````

````python
# Funktion testen
word = 'sugus'
reverse_word(word)
````
