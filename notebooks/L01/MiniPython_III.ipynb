### Teste, ob ein bestimmtes Zeichen in einem String vorkommt
Vergleiche Zeichen um Zeichen des Strings mit dem bestimmten Zeichen.

````python
text = 'agagdfsgdsf'
character = 'x'

n = len(text)
i = 0
contains_character = False

while i < n and not contains_character:
    if text[i] == character:
        contains_character = True
    i = i + 1

contains_character
````

### Z&auml;hle, wie oft ein bestimmtes Zeichen in einem String vorkommt

````python
text = 'abakadabra'
character = 'a'

n = len(text)
i = 0
count = 0

while i < n:
    if text[i] == character:
        count = count + 1
    i = i + 1

count
````

### Drehe ein Wort um 
Starte mit dem leeren Wort als reversed_word. Iteriere &uuml;ber die Buchstaben des umzudrehenden Wortes und addiere diese
von links zum reversed_word
reserved word ist

````python
word = 'abcd'
reversed_word = ''

n = len(word)
i = 0
while i < n:
    reversed_word = word[i] + reversed_word
    i = i + 1

reversed_word
````

### Teste, ob ein Wort ein Palindrom ist (gleich dem umgedrehten Wort)
Drehe das Wort wie oben um und teste dann ob es gleich ist.

````python
word = 'sugus'
reversed_word = ''

n = len(word)
i = 0
while i < n:
    reversed_word = word[i] + reversed_word
    i = i + 1

# Resultat ausgeben
if reversed_word == word:
    print(word, 'ist ein Palindrom')

if reversed_word != word:
    print(word, 'ist kein Palindrom')
````

### Zerlege ein Wort in zwei Teile
Gegeben sei ein Wort `word` und eine Zahl `m`.
Zerlege das Wort so in zwei W&ouml;rter `head` und `tail`,  
dass `head+tail == word` und `len(head) == m`.  

Addiere die ersten m Buchstaben des Wortes sukzessive zum head, dann die restlichen Buchstaben zum tail.

````python
word = 'foobar'
m = 3

head = ''
tail = ''

n = len(word)
i = 0

while i < m:
    head = head + word[i]
    i = i + 1

while i < n:
    tail = tail + word[i]
    i = i + 1

print(head, '+', tail)
````

### Sortierte W&ouml;rter zusammenf&uuml;gen
Wir nennen ein Wort sortiert, falls seine Buchstaben aufsteigend angeordnet sind.  
Kombiniere die Buchstaben zweier sortierter W&ouml;rter zu einem einzigen sortierten Wort.  
Bsp: `'adef'` und `'bc'` ergibt `'abcdef'`.  

Entscheide, aus welchem Wort der n&auml;chste Buchstaben kommt und h&auml;nge diesen an das neue Wort an.

````python
word1 = 'adef'
word2 = 'bc'
merged_word = ''

m = len(word1)
n = len(word2)

i = 0
j = 0

while i+j < m+n:
    from_word1 = (j >= n) or (i < m and word1[i] <= word2[j])
    if from_word1:
        next_char = word1[i]
        i = i + 1
    if not from_word1:
        next_char = word2[j]
        j = j + 1

    merged_word = merged_word + next_char

merged_word
````

### Anzahl Stellen einer positiven ganzen Zahl bestimmen
Benutze, dass `435 // 10 == 43` (letzte Ziffer wird abgeschnitten).  
Schneide solange die letzte Ziffer ab, bis nur noch 0 &uuml;brig bleibt.

````python
n = 4998
n_digits = 0
while n > 0:
    n = n // 10
    n_digits = n_digits + 1

n_digits
````

### Anzahl Stellen einer nicht-negativen ganzen Zahl bestimmen
Spezialfall n=0 separat behandeln, Rest wie oben.

````python
n = 0
n_digits = 0

if n == 0:
    n_digits = 1

while n > 0:
    n = n // 10
    n_digits = n_digits + 1

n_digits
````

### Teste, ob ein bestimmte Ziffer in einer Zahl vorkommt
Benutze das `453 % 10 == 3` die letzte Ziffer einer Zahl ist.  
Schneide jeweils die letzte Ziffer ab, falls diese ungleich der gesuchten Zahl ist.

````python
n = 12345689  # Zahl
_n = n  # _n Zahl, von der wir Ziffern abschneiden
digit = 5  # Ziffer die in der Zahl gesucht wird
found = False

while _n > 0 and not found:
    if digit == _n % 10:
        found = True
    _n = _n // 10

# Ergebins ausgeben
if found:
    print(digit, 'occurs in', n)
if not found:
    print(digit, 'does not occur in', n)
````

### Teste, ob eine Zahl eine Primzahl ist
Teste, ob 2, 3, ..., k (wo k*k < n) die Zahl n teilen. 

````python
n = 113  # n: ganze Zahl >= 2, teste, ob Primzahl
i = 2
is_prime = True

while i*i <= n and is_prime:
    if n % i == 0:
        is_prime = False
        print(n, ' =', i, ' *', n // i)
    i = i + 1

if is_prime:
    print('n',  'is prime')
````

**Wurzel berechnen**  
Gegeben sei ein (blaues) Quadrat mit Seitenl&auml;nge $a$ und Fl&auml;cheninhalt $A=a^2$ und ein etwas gr&ouml;sseres Quadrat mit Seitenl&auml;nge $x$.  
Die Idee ist, dass $x$ eine etwas zu grosse N&auml;herung f&uuml;r $a=\sqrt{A}$ ist.  
In der Figur sehen wir, dass $a=x-dx$.  
Wie kann man die Differenz $dx$ der Seitenl&auml;ngen absch&auml;tzen?  

<img align="left;" src="/files/images/square.svg">

Es gilt $x^2 =A + 2x\cdot dx - (dx)^2$ (das rote Quadrat mit Seitenl&auml;nge $dx$ wurde doppelt gez&auml;hlt).  

Im Vergleich zur Gesamtfl&auml;che ist die Fl&auml;che $dx^2$ des roten Quadrates klein. Es gilt n&auml;herungsweise  
$x^2 \simeq A + 2x\cdot dx$, also $dx\simeq\frac{x^2-A}{2x}=\frac{x-A/x}{2}$.

````python
# Wurzel auf 15 Nachkommastellen genau berechnen
epsilon = 1E-15
A = 2
x = 1  # erste Naeherung

dx = (x - A/x) /2
i = 0
while (dx < -epsilon) or (epsilon < dx):  # dx nicht im Intervall (-epsion, epsilon)
    dx = (x - A/x) /2
    x = x - dx
    i = i + 1

print('Wurzel aus', A, 'ist', x)
print('Berechnet in', i, 'Schritten')
print('Genauigkeit:', A**0.5-x)
````
