### Einkäufe
Nachstehender Dictionary speichert zu jeder Product-ID den Namen des zugehörigen Produktes und dessen Preis
```python
pid_prodPrice = {1: ('Brot', 2.30),
                 2: ('Milch', 1.50),
                 3: ('Schokolade', 2.85),
                 4: ('Buendnerfleich', 8.90),
                 5: ('Eier', 4.80),
                 }
```
Im File `Einkaeufe.txt` enthält Zeilen mit einer Customer-ID und den Produkt-IDs der 
gekauften Produkten.
1. Iteriere mit einer For-Schleife über die Key-Value Paare obigen Dicts und gib
   die Produkt-ID zusammen mit dem Produktnamen und dem Preis aus.
   ```python
   Produkt-ID: 1, Produktname: Brot, Preis: 2.3
   ⋮
   Produkt-ID: 5, Produktname: Eier, Preis: 4.8
   ```
2. Erstelle eine Liste `lines` mit den Zeilen dieses Files.
3. Greife die erste Datenzeile aus `lines` heraus. Speichere die Customer-ID als **Integer** in einer
   Variable `cid` und die Produkt-IDs als **Tuple von Integern** in einer Variable `pids`.  
   `cid -> 1561, pids -> (1, 2, 5, 4)`
7. Erstelle nun einen Dictionary der jeder Customer-ID ein Tuple mit den Produkt-IDs zuordnet.
   ```python
   {1561: (1, 2, 5, 4),
    1085: (3, 2, 2),
    2034: (5, 3, 3, 1, 1),
    2021: (5, 3, 4, 4, 1)}
   ```
8. Erstellen einen Dictionary `cid_products`, der jeder Customer-ID eine alphabetisch sortierte Liste mit den
   von diesem Kunden gekauften Produkten zuordnet. Jedes gekaufte Produkt soll **höchstens einmal** vorkommen.
   ```python
   {1561: ['Brot', 'Buendnerfleich', 'Eier', 'Milch'],
    ⋮
    2021: ['Brot', 'Buendnerfleich', 'Eier', 'Schokolade']}
   ```

````python
pid_prodPrice = {1: ('Brot', 2.30),
                 2: ('Milch', 1.50),
                 3: ('Schokolade', 2.85),
                 4: ('Buendnerfleich', 8.90),
                 5: ('Eier', 4.80),
                 }
````

````python
# 1a
for pid, (product, price) in pid_prodPrice.items():
    print(f'Produkt-ID: {pid}, Produktname: {product}, Preis: {price}')
````

````python
# 1b)
for pid, prodPrice in pid_prodPrice.items():
    product, price = prodPrice
    print(f'Produkt-ID: {pid}, Produktname: {product}, Preis: {price}')
````

````python
# 2
filename = 'Einkaeufe.txt'
with open(filename, mode='r') as f:
    lines = [line.rstrip() for line in f]

lines
````

````python
# 3
line = lines[1]
fields = line.split(',')
fields
````

````python
# 3 cont.
cid = int(fields[0])
pids = tuple(int(n) for n in fields[1:])
cid, pids
````

````python
# 4
cid_pids = {}
for line in lines[1:]:
    fields = line.split(',')
    cid = int(fields[0])
    pids = tuple(int(n) for n in fields[1:])
    cid_pids[cid] = pids

cid_pids
````

````python
# 5
cid_products = {}
for cid, orders in cid_pids.items():
    products = set(pid_prodPrice[n][0] for n in orders)
    cid_products[cid] = sorted(products)

cid_products
````
