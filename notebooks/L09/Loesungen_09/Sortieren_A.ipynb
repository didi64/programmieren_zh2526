1. Ein Punkt ist ein Tuple aus x- und y-Koordinate.
   Sortiere `points` nach dem Abstand vom Ursprung.  
   (`[(2, 3), (-3, -2), (4, 2), (0, 5), (1, 7), (-2, 7)]`).  
   Schreibe dazu eine Funktion `distance(v, w)`, die den Abstand zweier Punkte berechnet.
   Teste die Funktion, indem du den Abstand aller Punkte vom Ursprung berechnest. 
1. Erstelle einen Dict, der jedem Wort in `words` das entsprechende englische Wort in `words_en` zuweist.
1. Sortiere die Werte dieses Dicts alphabetisch.
1. Sortiere die Schlüssel dieses Dicts wie in 2.

````python
words = ['Blume', 'aber', 'weg', 'rot', 'Abfall', 'blau', 'Weg', 'Rost']
words_en = ['flower', 'but', 'missing', 'red', 'garbage', 'blue', 'path', 'rust']
points = [(2, 3), (4, 2), (-2, 7), (-3, -2), (0, 5), (1, 7)]
````

````python
# 1
sorted(words)
````

````python
# 2
sorted(words, key=lambda x: x.upper())
````

````python
# 3
sorted(words, key=lambda x: (x.upper(), x))
````

````python
# 4
sorted(words, key=lambda x: (len(x), x.upper(), x))
````

````python
# 5 Hilfsfunktion
def distance(v, w):
    dist = ((w[0]-v[0])**2 + (w[1] - v[1])**2)**(1/2)
    return dist
````

````python
# 5 Hilfsfunktion testen
ursprung = (0, 0)
for pt in points:
    print(f'Abstand vom Ursprung: {distance(ursprung, pt)}')
````

````python
# 5
[round(distance(ursprung, pt), 2) for pt in points]
````

````python
# 6
ursprung = (0, 0)
sorted(points, key=lambda pt: distance(ursprung, pt))
````
