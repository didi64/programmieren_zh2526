### For-Schleifen
Implementiere nachstehende Funktionen und teste diese mit den folgenden Argumenten. 
Der Kommentar ist jeweils der Rückgabewert.
1.  ```python
    xs = [2, 6, 1, 4, 5]
    minimum(xs)  # 1
1.  ```python
    xs = [2, 5, 10]
    product(xs)  # 100
    ```       

````python
def minimum(xs):
    '''gib die kleinste Zahl der Liste xs zurueck
       ohne die Funktion min zu benutzen.
    '''
    m = xs[0]
    for x in xs[1:]:
        if x < m:
            m = x
    return m
````

````python
xs = [2, 6, 1, 4, 5]
minimum(xs)
````

````python
def product(xs):
    '''berechne das Produkt aller Zahlen in der Liste xs'''
    product = 1
    for x in xs:
        product = product * x
    return product
````

````python
xs = [2, 5, 10]
product(xs)
````

### While-Schleifen
Wir wollen jeweils eine Menge, eine Liste und ein Tuple mit
5 verschiedenen zufälligen einstelligen Zahlen erstellen.

1. Starte mit einer leeren Menge `numbers`.
   Füge solange zufällige einstellige Zahlen zur Menge hinzu, bis diese
   5 Elemente enthält.
1. Starte mit einer leeren Liste `numbers`.
   Füge solange zufällige einstellige Zahlen zur Liste hinzu, bis diese
   5 **verschiedene** Elemente enthält. Nimm nur Zahlen in die Liste auf, die noch
   nicht in der Liste sind.
1. Starte mit einem leeren Tuple `numbers`. Falls eine zufällige einstellige Zahlen nicht im Tuple ist,
   verlängere das Tuple um diese Zahl, bis das Tuple
   5 **verschiedene** Elemente enthält. 

````python
from random import randint
````

````python
# Menge
numbers = set()

while len(numbers) < 5:
    number = randint(0, 9)
    numbers.add(number)

numbers
````

````python
# Liste
numbers = []

while len(numbers) < 5:
    number = randint(1, 9)
    if number not in numbers:
        numbers.append(number)

numbers
````

````python
# Tuple
numbers = ()

while len(numbers) < 5:
    number = randint(1, 9)
    if number not in numbers:
        numbers = numbers + (number,)

numbers
````
