### Count-Dict, While
1. Frage den Benutzer solange noch einer Ziffer, bis eine Ziffer eingegeben wurde.
1. Frage den Benutzer immer wieder nach einer Ziffer, bis eine Ziffer 3 Mal eingegen wurde.
Erstelle dabei einen Dict `d`, der zu jeder Ziffer angibt, wie oft sie eingegeben wurde.
Ignoriere die Eingabe, falls es sich nicht um eine Ziffer handelt.
2. Iteriere über die Schlüssel-Wert Paare von `d` und finde die Ziffer, welche 3 Mal eingegeben wurde.
3. Gib aus, wieviele Ziffern insgesamt eingegeben wurden (Summe der Werte im Dict `d`).
4. Berechne mit Hilfe von `d` den Durchschnitt aller eingegebenen Ziffern.

```python
# 1
ziffern = '0123456789'
while True:
    answer = input('Ziffer:')
    if len(answer) == 1 and answer in ziffern:
        print(f'Ja, {answer} ist eine Ziffer.')
        break
```

```python
# 2
d = {}
while 3 not in d.values():
    answer = input('Ziffer:')
    if len(answer) == 1 and answer in ziffern:
        d[answer] = d.get(answer, 0) + 1

d
```

```python
# 3
for k, v in d.items():
    if v == 3:
        print(f'Die Ziffer {k} wurde 3 Mal eingegeben.')
        break
```

```python
# 4a
sum(d.values())
```

```python
# 4b
total = 0
for v in d.values():
    total = total + v

total
```

```python
# 5
total = 0
for k, v in d.items():
    total = total + int(k)*v

print('Durchschnitt:', total/len(d))
```
