````python
def form_pairs_1(xs):
    '''xs ist eine Liste
       gibt eine Liste mit Paaren (Tupeln der Laenge 2)
       aufeinanderfolgender Elemente von xs zurueck.
       Ist die Anzahl Elemente ungerade, igonirere das letzte Element
    '''
    pairs = []
    n = len(xs)
    i = 0

    while i+1 < n:
        pair = (xs[i], xs[i+1])
        pairs.append(pair)
        i = i + 2
    return pairs
````

````python
numbers = [1, 2, 3, 4, 5]
form_pairs_1(numbers)
````

````python
def form_pairs_2(xs):
    '''xs ist eine Liste
       gibt eine Liste mit Paaren (Tupeln der Laenge 2)
       aufeinanderfolgender Elemente von xs zurueck.
       Das letzte Tuple hat ev. nur Laenge 1.
    '''
    pairs = []
    n = len(xs)
    i = 0

    while i+1 < n:
        pair = (xs[i], xs[i+1])
        pairs.append(pair)
        i = i + 2

    if i < n:
        pair = (xs[i],)
        pairs.append(pair)

    return pairs
````

````python
numbers = [1, 2, 3, 4, 5]
form_pairs_2(numbers)
````

````python
def form_pairs(xs):
    '''xs ist eine Liste
       gibt eine Liste mit Paaren (Tupeln der Laenge 2)
       aufeinanderfolgender Elemente von xs zurueck.
       Das letzte Tuple hat ev. nur Laenge 1.
    '''
    pairs = []
    for i in range(0, len(xs), 2):
        pair = tuple(xs[i:i+2])
        pairs.append(pair)
    return pairs
````

````python
numbers = [1, 2, 3, 4, 5]
form_pairs(numbers)
````

Wir suchen einen Integer $x$, so dass $x^2\leq n < (x+1)^2$.
Wir starten mit `lower = 0` und `upper = n+1`.
Es soll immer gelten, dass `lower`${}^2\leq n <$ `upper`${}^2$.
Ziel ist es, einen Wert für `lower` zu finden, so dass `lower+1` gleich `upper` ist.
Dazu raten wir einen Wert zwischen `lower` und `upper`, nämlich `x = (lower + upper) // 2`.
Ist $x^2 \leq n$, dann können wir `lower = x` setzen, andernfalls
können wir `upper = x` setzen.

````python
def square_root(n):
    lower = 0
    upper = n+1

    while lower + 1 < upper:
        x = (lower + upper) // 2
        print(f'lower: {lower}, x: {x}, upper: {upper}')
        if x**2 <= n:
            lower = x
        else:
            upper = x
    return lower, upper
````

````python
def collatz(n):
    steps = []
    while n > 1:
        steps.append(n)
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3*n + 1

    steps.append(1)
    return steps
````

````python
for i in range(2, 10):
    print(collatz(i))
````

````python
def log2(n):
    result = 0
    while n > 0:
        n = n // 2
        result = result + 1
    return result - 1
````
