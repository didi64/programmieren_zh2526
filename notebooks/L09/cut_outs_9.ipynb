```python
def form_pairs_1(xs):
    '''xs ist eine Liste
       gibt eine Liste mit Paaren (Tupeln der Laenge 2)
       aufeinanderfolgender Elemente von xs zurueck.
       Ist die Anzahl Elemente ungerade, igonirere das letzte Element
    '''
    pairs = []
    n = len(xs)
    i = 0

    while i+1 < n:
        pair = (xs[i], xs[i+1])
        pairs.append(pair)
        i = i + 2
    return pairs
```

```python
numbers = [1, 2, 3, 4, 5]
form_pairs_1(numbers)
```

```python
def form_pairs_2(xs):
    '''xs ist eine Liste
       gibt eine Liste mit Paaren (Tupeln der Laenge 2)
       aufeinanderfolgender Elemente von xs zurueck.
       Das letzte Tuple hat ev. nur Laenge 1.
    '''
    pairs = []
    n = len(xs)
    i = 0

    while i+1 < n:
        pair = (xs[i], xs[i+1])
        pairs.append(pair)
        i = i + 2

    if i < n:
        pair = (xs[i],)
        pairs.append(pair)

    return pairs
```

```python
numbers = [1, 2, 3, 4, 5]
form_pairs_2(numbers)
```

```python
def form_pairs(xs):
    '''xs ist eine Liste
       gibt eine Liste mit Paaren (Tupeln der Laenge 2)
       aufeinanderfolgender Elemente von xs zurueck.
       Das letzte Tuple hat ev. nur Laenge 1.
    '''
    pairs = []
    for i in range(0, len(xs), 2):
        pair = tuple(xs[i:i+2])
        pairs.append(pair)
    return pairs
```

```python
numbers = [1, 2, 3, 4, 5]
form_pairs(numbers)
```
