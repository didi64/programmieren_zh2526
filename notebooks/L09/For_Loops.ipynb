````python
def minimum(xs):
    '''gib die kleinste Zahl der Liste xs zurueck
       ohne die Funktion min zu benutzen.
       xs: list[int]
    '''
    m = xs[0]
    for x in xs[1:]:
        if x < m:
            m = x
    return m
````

````python
xs = [2, 6, 1, 4, 5]
minimum(xs)
````

````python
def find_palindroms(words):
    '''gib eine Liste mit den Woertern in words
       zurueck, die Palindrome sind.
       Unterscheide nicht zw. Gross- und Kleinbuchstaben
    '''
    return [w for w in words if w.lower() == w.lower()[::-1]]
````

````python
words = ['Sugus', 'rotor', 'foo']
find_palindroms(words)
````

````python
def one_space_breaks(s):
    '''entferne Whitespace am Anfang und Ende und
       reduziere Whitespace zw. Woertern zu ' '
    '''
    return ' '.join(s.split())
````

````python
s = ' foo  bar   baz  '
one_space_breaks(s)
````

````python
def form_pairs(xs):
    '''xs ist eine Liste
       gibt eine Liste mit Paaren (Tupeln der Laenge 2)
       aufeinanderfolgender Elemente zurueck.
       Das letzte Tuple hat ev. nur Laenge 1.
    '''
    pairs = []
    for i in range(0, len(xs), 2):
        pair = tuple(xs[i:i+2])
        pairs.append(pair)
    return pairs
````

````python
numbers = [1, 2, 3, 4, 5]
form_pairs(numbers)
````

````python
def product(xs):
    '''berechne das Produkt aller Zahlen in der Liste xs'''
    product = 1
    for x in xs:
        product = product * x
    return product
````

````python
xs = [2, 5, 10]
product(xs)
````
