### Elemente auswählen 
1. Gegeben sei eine Liste mit Zahlen `numbers`.
Schreibe Code, der ausgibt, wieviele verschiedene Zahlen die Liste enthält.
2. Erzeuge eine Zufallszahl $i$, so dass `0 <= i < len(numbers)` und 
greife dann die i-te Zahl heraus.
3. Benutze eine Schleife, um ein Tuple mit 5 zufälligen Zahlen aus `numbers` zu erstellen.
4. Schreibe eine Funktion  `choices(items, k=5)`, die aus einem Iterable k zufällige Elemente herauspickt und diese als Tuple zurück gibt.
5. Schreibe eine Funktion  `unique_choices(items, k=5)`, die ein Tuple mit k zufälligen, **paarweise verschiedenen** Elementen aus `items` zurück gibt.
Enthält `items` k oder weniger verschienene Elemente, so gib alle Elemente von `items` als Tuple zurück.
(Benutze eine While-Schleife, um zu einer Menge solange zufällige Elemente aus `items` hinzuzufügen, bis diese k Elemente enthält).

```python
from random import randint

numbers = [1, 2, 1, 3, 4, 3, 5, 6, 7, 8, 9]
```

```python
len(set(numbers))
```

```python
n = len(numbers) - 1
i = randint(0, n)
number = numbers[i]
number
```

```python
results = []

n = len(numbers) - 1
for _ in range(5):
    i = randint(0, n)
    number = numbers[i]
    results.append(number)
results
```

```python
def choices(items, k=5):
    results = []
    n = len(items) - 1
    for _ in range(k):
        i = randint(0, n)
        item = items[i]
        results.append(item)
    return tuple(results)
```

```python
choices('abcdefgh')
```

```python
numbers = [1, 2, 1, 3, 4, 3, 5, 6, 7, 8, 9]
k = 5

result = set()
n = len(numbers) - 1

while len(result) < k:
    i = randint(0, n)
    number = numbers[i]
    result.add(number)

tuple(result)
```

```python
def unique_choices(items, k=5):
    n = len(items)
    if len(set(items)) <= k:
        return tuple(items)

    result = set()
    n = len(items) - 1
    while len(result) < k:
        i = randint(0, n)
        item = items[i]
        result.add(item)
    return tuple(result)
```

```python
unique_choices(numbers, k=5)
```
