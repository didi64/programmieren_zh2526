### Prüfung vom 12. Januar 2026

**Allgemeines**
- Zeit: 90 Minuten.
- Hilfsmittel: Unterrichtunterlagen, Notizen, ...
- **Vorboten** ist die Kommunikation mit anderen Personen in jeglicher Form, AI, Chat-GPT.
- Deine Lösung darf nur Sprachelemente enthalten, die wir im Unterricht behandelt haben.

**Prüfungsstoff** Der in den Lektionen 1-9 behandelte Stoff. Insbesondere:
- input und print
- Rechnen mit Integer, Floats und Strings, inkl. Stringmethoden und f-Strings
- Tuple, Listen, Mengen und Dictionaries
- Typumwandlungen: Z.B. `int` zu `str`, oder `list` zu `set`
- Slice- und Indexnotation
- Sortieren von Listen und Dicts (nach Schlüssel oder Werten)
- Listen und Dict-Comprehension
- Text-Files lesen und schreiben
- For- und While-Schleifen, inkl. range, enumerate und zip
- Bedingte Anweisungen: if, elif, else
- Funktionen
