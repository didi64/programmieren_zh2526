### Cheatsheet

**int**: Integer/ ganze Zahlen.   
Operationen: `+, -, *, /, **, //, %` (plus, minus, mal, durch, hoch, Ganzzahldivision,
  modulo).    
**float**: Floats/Fliesskommazahlen.   
Operationen: `+, -, *, /, **, //, %`.  
Funktionen: `abs(x), min(x, y), max(x, y), round(x, ndigits=None)`.  
**str**: String.  
Operationen: `+`, `*` (Multiplikation mit Integer).  
Funktionen: `len`.  
Methoden: `split(sep)`, `join(items)`, `index(item)`, `find(item)`, `replace(old, new)`, `lstrip()`, `rstrip()`, `strip()`, `ljust`, `rjust`.  
**bool**: `True` und `False`.  
`x` heisst *wahr*, falls `bool(x)` gleich `True`, sonst *falsch*.   
Nur `0`, `0.0`, `None` und leere Iterables (`''`, `()`, `[]`, `{}`, ...) sind *falsch*.  
Operationen: `and`, `or`, `not`, `+`, `*`, `-`.  
**tuple**: Tuple `x = ('Max', 'Muster')`.  
Operationen: `+`, `*` (Multiplikation mit Integer).  
Funktionen: `len`.  
**list**: Liste `x = ['Max', 'Muster']`.  
Operationen: `+`, `*` (Multiplikation mit Integer).   
Funktionen: `len`.  
Methoden: `append(item)`, `pop(index=-1)`, `extend(items)`, `insert(index, item)`, `remove(item)`, `clear()`.      
**set**: Menge `x = set()`, `x = {1, 2, 3}`, `x = set(iterable)`.  
Operationen: `|` (Verein), `&` (Schnitt), `-` (Differenz).  
Funktionen: `len`.  
Methoden: `add(item)`, `remove(item)`, `pop()` (liefert und entfernt ein Element).  
**dict**: Dictionary: `d = {}`, `d = dict(zip(keys, values))`.  
Operationen: `|` (Verein).  
Funktionen: `len`.  
Methoden: `keys()`, `values()`, `items()`, `pop(key)`.  

**Typ eines Objekts**: `type(x)` liefert den Typ von `x`. Jeden Typ gibt es nur einmal, deshalb werden Typen mit `is` verglichen (z.B. `type(x) is int`). 


**Index und Slicenotation bei Strings, Tuple und Listen**:  
Verwendung: `s[stop], s[start:stop], s[start:stop:step]` (ist start oder stop Anfang oder Schluss kann es weggelassen werden).
- `s[0], s[1], s[-1], s[-2]`: erstes, zweites, letztes und zweitletztes Element.
- `s[:2], s[-2:]`: Die ersten und letzten beiden Elemente.
- `s[:], s[::-1]`: Alle Elemente, alle Elemente von hintennach vorn.
- `s[::2], s[::-2]`: jedes 2. Element, jedes 2. Element von hintennach vorn.


**Die  `range` Funktion**: 
Verwendung: `range(stop), range(start, stop), range(start, stop, step)`

**Auf Iterables anwendbare Funktionen**:  
- `max`, `min`, `len`, `any`, `all`.   
  `any(items)` gibt `True` zurück, falls mind. ein Element in `items` *wahr*
  ist.  
  `all(items)` gibt `True` zurück, falls alle Element in `items` *wahr*
  sind.
- `enumerate(xs)`: range-Objekt mit Elementen `((0, xs[0]), (1, xs[1]), ...)`,  
  `zip(xs, ys)`:   range-Objekt mit Elementen `((xs[0], ys[0]), (xs[1], ys[1]), ...)`
  `reversed(xs)`:  range-Objekt mit Elementen `(xs[-1], ..., xs[0])`
- `sorted(iterable, key=None, reverse=False)`: sortiert das Iterable.
`key` ist eine Funktion, sortiert wird aufsteigend nach dem Sortierwert `key(item)`, oder absteiend falls `reverse=True`.  

**f-Strings**:
- `f'Zahl {x}, Quadrat: {x*x}'` (String mit `{<Ausdruck>}`. Der `<Ausdruck>` wird ausgewertet und in einen String umgewandelt.
  

**Iterables und Dictionaries auspacken**:  
Ist `args=[1, 2]`, so hat `f(x, *args)` den gleichen Effekt wie `f(x, 1, 2)`.
Die Elemente des Iterable `args` werden ausgepackt.  
Ist `kwargs={'width': 200, 'height': 100}`, so hat 
`f(x, **kwargs)` den gleichen Effekt wie `f(x, width=200, height=100)`. Der Dictionary `kwargs` wird ausgepackt.
