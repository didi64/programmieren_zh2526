### While-Loops
Oft hat ein While-Loop folgende Form:
```python
result = ...  # initialer Wert des Resultates
n = ...  # initialer Wert 
i = 0

while i < n:
    # modifiziere result
    i = i + 1
    

```

````python
def collatz(n):
    steps = []
    while n > 1:
        steps.append(n)
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3*n + 1

    steps.append(1)
    return steps
````

````python
for i in range(2, 10):
    print(collatz(i))
````

````python
def mul(a, b):
    stack = [a, b]
    while stack[1] > 1:
        stack[0] = stack[0] + a
        stack[1] = stack[1] - 1
    return stack[0]
````

````python
mul(3, 4)
````

````python
from random import randint
````

````python
def distinct_random_digits_1(n=5):
    numbers = set()
    while len(numbers) < n:
        number = randint(1, 9)
        numbers.add(number)
    return numbers
````

````python
def distinct_random_digits_2(n=5):
    numbers = []
    while len(numbers) < n:
        number = randint(1, 9)
        if number not in numbers:
            numbers.append(number)
    return numbers
````

````python
distinct_random_digits_2()
````

````python
def distinct_random_digits_3(n=5):
    numbers = ()
    while len(numbers) < n:
        number = randint(1, 9)
        if number not in numbers:
            numbers = numbers + (number,)
    return numbers
````

````python
distinct_random_digits_3()
````

Wir suchen einen Integer $x$, so dass $x^2\leq n < (x+1)^2$.
Wir starten mit `lower = 0` und `upper = n+1`.
Es soll immer gelten, dass `lower`${}^2\leq n <$ `upper`${}^2$.
Ziel ist es, einen Wert für `lower` zu finden, so dass `lower+1` gleich `upper` ist.
Dazu raten wir einen Wert zwischen `lower` und `upper`, nämlich `x = (lower + upper) // 2`.
Ist $x^2 \leq n$, dann können wir `lower = x` setzen, andernfalls
können wir `upper = x` setzen.

````python
def square_root(n):
    lower = 0
    upper = n+1

    while lower + 1 < upper:
        x = (lower + upper) // 2
        print(f'lower: {lower}, x: {x}, upper: {upper}')
        if x**2 <= n:
            lower = x
        else:
            upper = x
    return lower, upper
````

````python
square_root(2)
````

````python
def log2(n):
    result = 0
    n = n // 2
    while n > 0:
        n = n // 2
        result = result + 1
    return result
````

````python
def form_pairs(xs):
    '''xs ist eine Liste
       gibt eine Liste mit Paaren (Tupeln der Laenge 2)
       aufeinanderfolgender Elemente von xs zurueck.
       Ist die Anzahl Elemente ungerade, igonirere das letzte Element
    '''
    pairs = []
    n = len(xs)
    i = 0

    while i+1 < n:
        pair = (xs[i], xs[i+1])
        pairs.append(pair)
        i = i + 2
    return pairs
````
