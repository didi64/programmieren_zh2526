### Weitere Aufgaben
Implementiere nachstehende Funktionen und teste diese mit den folgenden Argumenten. 
Der Kommentar ist jeweils der Rückgabewert.
1.  ```python
    words = ['Sugus', 'rotor', 'foo']
    find_palindroms(words)  # ['Sugus', 'rotor']
    ```
1.  ```python
    s = ' foo  bar   baz  '
    one_space_breaks(s)  # 'foo bar baz'
    ```       

````python
def find_palindroms(words):
    '''gib eine Liste mit den Woertern in words
       zurueck, die Palindrome sind.
       Unterscheide nicht zw. Gross- und Kleinbuchstaben
    '''
    return [w for w in words if w.lower() == w.lower()[::-1]]
````

````python
words = ['Sugus', 'rotor', 'foo']
find_palindroms(words)
````

````python
def one_space_breaks(s):
    '''entferne Whitespace am Anfang und Ende und
       reduziere Whitespace zw. Woertern zu ' '
    '''
    parts = s.split()
    return ' '.join(parts)
````

````python
s = ' foo  bar   baz  '
one_space_breaks(s)
````
