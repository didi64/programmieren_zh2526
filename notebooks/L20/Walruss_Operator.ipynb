### Der Walruss Operator `:=`

Der Walruss Operator kombiniert Zuweisung und Evalutation eines Ausdrucks.  

Z.B. ist `(x := 24)` gleichzeitig eine Zuweisung und ein Ausdruck.  
Der Wert des Ausdrucks ist der Wert der rechten Seite von `:=`.
Gleichzeitig wird der Variable `x` dieser Wert zugewiesen.


Einige typische Verwendungsweisen:  
```python
if (n := get_number()) < 0:
    # do stuff with n


while (value := get_nest_value()) is not None:
    # do stuff with value


results = [result for i in range(10) if (result := get_result(i)) > threshold]
```

Die Verwendungsweise im letzen Fall macht Sinn, da
der List-Comperhensionsausdruck eine Kurzform folgenden Programms ist:
```python
results = []
for i in range(10):
    if (result := get_result(i)) > threshold:
        results.append(result)
```

````python
numbers = [1, 2, 3, -1, -2, -3]

n = numbers[-1]
if n < 0:
    print(f'do stuff with {n}')
````

````python
numbers = [1, 2, 3, -1, -2, -3]

if (n := numbers[-1]) < 0:
    print(f'do stuff with {n}')
````

````python
stack = [1, 2, 3, -1, -2, -3]
n = stack.pop()
while n < 0:
    print(n)
    n = stack.pop()

n
````

````python
stack = [1, 2, 3, -1, -2, -3]
while (n := stack.pop()) < 0:
    print(n)

n
````

````python
def get_square(i):
    print(f'computing {i}*{i}')
    return (i*i)
````

````python
big_squares = [get_square(i) for i in range(5) if get_square(i) > 5]
big_squares
````

````python
big_squares = [square for i in range(5) if (square := get_square(i)) > 5]
big_squares
````
