### Aufgabe
Im Notebook `L19/Simple_Grid_Controller/TikTakToe_in_3_Komp.ipynb`
haben wir mit Hilfe des Modules `grid_controller` erstellt.



Folge diesem Muster, um eine ganz simple App zu programmieren, die
es erlaubt, in ein 10x10 Gitter zu klicken.
- Das geklickte Feld wird mit der ausgewählten Farbe ausgefüllt.
- Drücken von 1, 2 oder 3 wählt die Farben rot, grün oder blau.
- Drücken von c löscht die Gitterfelde und setzt den Colorcode auf 1.

So sieht es aus.

- Spiellogik: Im Modul `tiktak_L19`
- Darstellung: Im Modul `view`.  
  Das Modul `view` muss mit der Funktion `init(grid_spec, width=100, height=100)` initialisiert werden.
  Dieser Funktion muss das Modul `game` übergeben werden, damit die Funktion `game.update` überschreiben werden kann.
  Weiter wird ein Canvas-Widget und eine Grid-Spec erstellt und
  in `view.canvas` und `view.grid_spec` gespeichert.
- Controller: Im Modul `grid_controller`.
  Der Funktionsaufruf `init(view.canvas, view.grid_spec, callbacks)` sagt dem Controller,
  dass das Canvas-Widget der Darstellungskomponente benutzt werden soll, um auf Maus- und Tastatureingaben zu hören. Ebenso wird die `gird_spec` der View genutzt, um
  herauszufinden, in welches Feld geklickt wird.
  Weiter werden die Callbacks im Dict `callbacks` registriert.
  
Die grobe Struktur der Module `app.py` und `view.py` steht bereits.
- Das Modul `app` hat eine globale Variable `colorcode`, welche die ausgewählten Farbe speichert
  und die Variable `matrix` speichert die Farbcodes der Gitterfelder.
- View: siehe `view.py`
  

Die App soll dann wie folgt gestartet werden können:
```python
import app
import view
import grid_controller as C


callbacks = {'mouse_down': app.update_field,
             'c': app.clear,
             '1': lambda: app.set_colorcode(1),
             '2': lambda: app.set_colorcode(2),
             '3': lambda: app.set_colorcode(3),
             }

view.init(app)
C.init(view.mcanvas, callbacks, view.grid_spec)
C.show(debug=True)
```
