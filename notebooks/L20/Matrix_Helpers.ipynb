### Matrix Helpers
`matrix_helpers` ist ein Modul mit einigen Hilfsfunktionen zu erstellen und manipulieren 
von Matrizen/Tabellen, bez. Listen von Listen.

**Tücken** beim Erstellen von Matrizen:  
Auf den ersten Blick sehen `A` und `B` beide gleich aus (2 x 3 Matizen)

```python
A = [[0] * 3] * 2  # [[0, 0, 0], [0, 0, 0]]
B = [[0] * 3 for _ in range(2)]  # [[0, 0, 0], [0, 0, 0]]
```
Visualisiere die Ausführung des Codes auf [pythontutor.com](https://pythontutor.com/visualize.html#).  
A und B sind beides Listen von Listen.  
A enthält zweimal **dieselbe** Liste, B enthält zwei verschiedene Listen mit den gleichen Elementen.


````python
A = [[0] * 3] * 2  #
B = [[0] * 3 for _ in range(2)]
print(A)
print(B)
````

````python
A[0] is A[1] is A[2]
````

````python
A[0][0] = 1
A
````

````python
B[0] is B[1]
````

````python
B[0] == B[1]
````

````python
B[0][0] = 1
B
````

````python
import matrix_helpers as M


# help(M)
````

````python
matrix = M.make_matrix(ncol=5, nrow=3, default=0)
matrix
````

````python
for i in range(15):
    pos = M.idx2pos(i, ncol=5)
    M.set_cell(matrix, pos, i+1)
matrix
````

````python
M.show_matrix(matrix, cell_width=2)
````

````python
side_neighbors = M.get_neighbors(matrix, (2, 1), kinds='s')
print(side_neighbors)

[M.get_cell(matrix, pos) for pos in side_neighbors]
````

````python
corner_neighbors = M.get_neighbors(matrix, (2, 1), kinds='c')
print(corner_neighbors)

[M.get_cell(matrix, pos) for pos in corner_neighbors]
````
