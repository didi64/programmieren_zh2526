### Matrix Helpers
`matrix_helpers` ist ein Modul mit einigen Hilfsfunktionen zum Erstellen und Manipulieren 
von Matrizen/Tabellen, bez. Listen von Listen.

**Tücken** beim Erstellen von Matrizen:  
Auf den ersten Blick sehen `A` und `B` beide gleich aus (2 x 3 Matizen)

```python
A = [[0] * 3] * 2  # [[0, 0, 0], [0, 0, 0]]
B = [[0] * 3 for _ in range(2)]  # [[0, 0, 0], [0, 0, 0]]
```
Visualisiere die Ausführung des Codes auf [pythontutor.com](https://pythontutor.com/visualize.html#).  
A und B sind beides Listen von Listen.  
A enthält zweimal **dieselbe** Liste, B enthält zwei verschiedene Listen mit den gleichen Elementen.


````python
A = [[0] * 3] * 2  #
B = [[0] * 3 for _ in range(2)]
print(A)
print(B)
````

````python
A[0] is A[1] is A[2]
````

````python
A[0][0] = 1
A
````

````python
B[0] is B[1]
````

````python
B[0] == B[1]
````

````python
B[0][0] = 1
B
````
