Wir schrieben eine Funktion`get_idx(s, sep)`, welche den Index/Position eines Zeichens `sep` liefert,
und eine Funktion `split(s, sep)`, welche
den String `s` beim Separationszeichen `sep` in Werte spaltet und diese als Tupel zur端ck gibt.  
Dabei nutzten wir die Slicenotation `s[:end]`, `s[start:]` und `s[start:end]`.

Nachstehend die beiden Funktionen.
`get_idx(s, sep)` heisst neu `find(s, char)` und gibt `-1` zur端ck, falls
das Zeichen nicht gefunden wird.  `split(s, sep)` trennt nur und entfernt nicht
f端hrenden SPACE. 

Wir werden sehen, dass ein String-Methoden hat, um 
die Position eines Zeichen anzugeben oder sich zu zerlegen:

s.split(sep)

````python
def find(s, char):
    '''s: String
       gibt den Index des Zeichen char in s zurueck, 
       oder -1 falls es nicht vorkommt
    '''
    i = 0
    n = len(s)
    while i < n and s[i] != char:  # stoppe falls s[i] == char (Zeichen gefunden) oder i zu gross.
        i = i + 1

    if i == n:  # Zeichen nicht gefunden
        i = -1
    return i


def split(s, sep):
    '''s: str (kommaseparierte Werte)
       gibt ein Tupel mit den Werten zurueck
    '''
    values = ()

    i = find(s, sep)
    while i != -1:  # solange Trennzeichen im String
        value = s[:i]  # Teilstring vor Separator
        s = s[i+1:]    # Teilstring nach Separator

        values = values + (value,)
        i = find(s, sep)  # Postition des naechsten Separators

    values = values + (s,)  # letzter Wert hinzufuegen
    return values
````

````python
s = '0123456789'
print(f'Unser find: {find(s, '5')}, String-Methode str.find: {s.find('5')}')
print(f'Unser find: {find(s, 'x')}, String-Methode str.find: {s.find('x')}')
````

````python
s = 'foo, bar, baz'
print(f'Unser split: {split(s, ',')}, String-Methode str.split: {s.split(',')}')
print(f'Unser slit: {split(s, ',')}, String-Methode str.split: {s.split(',')}')
````

### Text in Zeilen spalten und Tabelle erstellen
Wir hatten dann `split(s, sep)` genutzt, um einen Text in Zeilen zu spalten,
und die Zeilen in Tuple mit Werten.  
Aus diesen Tupeln haben wir dann Werte ausgesondert,
mit der Hilfsfunktion, die wir nun `get_items(items, indices)` nennen.  
Die Funktion `get_table(lines, indices, sep)` erstellt dann eine Tabelle mit den
gew端nschten Werten.  

Python hat auch eine String-Methoden `str.splitlines(s)`, um durch `\n` getrennete Zeilen zu zerlegen.
```python
s.splitlines()
```

````python
def get_items(items, indices):
    '''items: tuple
       indices: tuple[int]
       Gib Elemente an den Positionen in indices zurueck
    '''
    result = ()
    n = len(indices)
    i = 0

    while i < n:
        idx = indices[i]  # Position des gewuenschten Elements
        item = items[idx]  # gewuenschtes Elements
        result = result + (item,)  # item an result anhaengen
        i = i + 1
    return result


def get_table(lines, indices, sep):
    '''lines: str (durch sep separierte Werte)
       indices: tuple[int]
       gibt eine Tabelle mit den Spaltennummern indices zurueck
    '''
    result = ()
    n = len(lines)
    i = 0

    while i < n:
        line = lines[i]
        items = split(line, sep)  # Zeile spalten
        items = get_items(items, indices)  # gewuenschte Werte aussondern
        result = result + (items,)
        i = i + 1

    return result
````
