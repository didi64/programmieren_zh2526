```python
# Tuple mit Positionen von 'a' in 'abakadabra' erstellen
s = 'abakadabra'
indices = ()
i = s.find('a')
while i != -1:
    indices = indices + (i,)
    i = s.find('a', i+1)
indices
```

```python
def find_all(s, char):
    indices = ()
    i = s.find(char)
    while i != -1:
        indices = indices + (i,)
        i = s.find(char, i+1)
    return indices
```

```python
s = 'abakadabra'
find_all(s, s[0])
```

```python
# aufeinenderfolgende Elemente vertauschen
numbers = [6, 3, 8, 6, 7]
n = len(numbers)

for i in range(n-1):
    if numbers[i] > numbers[i+1]:
        numbers[i], numbers[i+1] = numbers[i+1], numbers[i]

numbers
```

```python
def bubble(items):
    n = len(items)
    for i in range(n-1):
        if items[i] > items[i+1]:
            items[i], items[i+1] = items[i+1], items[i]
```

```python
numbers = [5, 4, 3, 2, 1]
```

```python
# fuehre diese Zelle mehrmals aus
bubble(numbers)
numbers
```

```python
def bubble_sort(items):
    n = len(items)
    for _ in range(n-1):
        bubble(items)
```

```python
numbers = [5, 4, 3, 2, 1]
bubble_sort(numbers)
numbers
```
