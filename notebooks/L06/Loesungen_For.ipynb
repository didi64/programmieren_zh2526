```python
def get_items_mit_while(items, indices):
    '''items: tuple
       indices: tuple[int]
       Gib Elemente an den Positionen in indices zurueck
    '''
    result = ()
    n = len(indices)
    i = 0

    while i < n:
        idx = indices[i]  # Position des gewuenschten Elements
        item = items[idx]  # gewuenschtes Elements
        result = result + (item,)  # item an result anhaengen
        i = i + 1
    return result


def get_items(items, indices):
    '''items: tuple
       indices: tuple[int]
       Gib Elemente an den Positionen in indices zurueck
    '''
    result = ()
    for idx in indices:
        item = items[idx]  # gewuenschtes Elements
        result = result + (item,)  # item an result anhaengen
    return result
```

```python
s = '0123456789'
indices = (2, 0, 7, 2)
get_items(s, indices)
```

```python
def find_mit_while(s, char):
    '''s: String
       gibt den Index des Zeichen char in s zurueck, 
       oder -1 falls es nicht vorkommt
    '''
    i = 0
    n = len(s)
    while i < n and s[i] != char:  # stoppe falls s[i] == char (Zeichen gefunden) oder i zu gross.
        i = i + 1

    if i == n:  # Zeichen nicht gefunden
        i = -1
    return i


def find_v1(s, char):
    '''s: String
       gibt den Index des Zeichen char in s zurueck, 
       oder -1 falls es nicht vorkommt
    '''
    n = len(s)
    for i in range(n):
        if s[i] == char:
            break

    if s[i] != char:  # Zeichen nicht gefunden
        i = -1
    return i


def find_v2(s, char):
    '''s: String
       gibt den Index des Zeichen char in s zurueck, 
       oder -1 falls es nicht vorkommt
    '''
    n = len(s)
    for i in range(n):
        if s[i] == char:
            break

    if s[i] != char:  # Zeichen nicht gefunden
        i = -1
    return i


def find_v3(s, char):
    '''s: String
       gibt den Index des Zeichen char in s zurueck, 
       oder -1 falls es nicht vorkommt
    '''
    n = len(s)
    for i in range(n):
        if s[i] == char:
            return i

    return -1


def find_v4(s, char):
    '''s: String
       gibt den Index des Zeichen char in s zurueck, 
       oder -1 falls es nicht vorkommt
    '''
    i = 0
    for zeichen in s:
        if zeichen == char:
            return i
        i = i + 1

    return -1
```

```python
funs = (find_mit_while,
        find_v1,
        find_v2,
        find_v3,
        find_v4,
        )

s = '0123456789'
zeichen = '4'

for fun in funs:
    print(f'{fun.__name__}: {fun(s, zeichen)}')  # fun._name__ ist der Funktionsname von fun
```
