### Fehlerbehandlung
Dem Slogan  "It is Easier to Ask for Forgiveness than Permission" folgend,
erlaubt es Python, etwas zu versuchen und im Falle eines Misserfolgs
den Schaden zu beheben.

```python
try:
    may_be_works()  # Code der ev. ValueError auslöst
except ValueError:
    behandle_den_Fehler()
else:
    # wird nur im Erfolgsfall ausgeführt
    ...
finally:
    # aufräumen, wird immer ausgeführt
    ...
```

**Best Practice**:
- Sei konkret: Brauche **nie** nur `except:`.  
  Fange nur konkrete Errors ab, z.B. `except ValueError:`.
- Halte den try-Block klein: Nur die Codezeile, die den Error auslösten könnte.

**Gängige Errors**:
- `NameError`: undefinierte Variable,
- `TypeError`: unpassender Typ,
- `ValueError`: passender Typ, aber unpassender Wert,
- `IndexError`: Index unpassend,
- `KeyError`: Schlüssel nicht im Dict,
- `ImportError`: Importversuch nicht erfolgreich.

````python
def get_number():
    try:
        val = int(input('Zahl:'))
    except ValueError:
        print('Das war keine Zahl! Ich interpretiere das als "None"')
        return None  # wird erst nach dem finally-Blockausgeführt!
    else:
        print(f'Deine Zahl: {val} years')
    finally:
        print('Danke, dass Sie get_number nutzen.')
    return val
````

````python
n = get_number()
print(f'got "{n}"')
````

***
Oft wird versucht, eine Funktion zu importieren.
Schlägt das fehl, wird eine Ersatzlösung verwendet.
***

````python
try:
    from random import randint
except ImportError:
    randint = lambda m, n: (m+n)//2

for i in range(3):
    print(f'Zufallszahl: {randint(1, 99)}')
````
