### Fehlerbehandlung
(siehe auch [https://docs.python.org/3/tutorial/errors.html](https://docs.python.org/3/tutorial/errors.html).    
Dem Slogan  "It is Easier to Ask for Forgiveness than Permission" folgend,
erlaubt es Python, etwas zu versuchen und im Falle eines Misserfolgs
den Schaden zu beheben/begrenzen.

```python
try:
    versuche_etwas()  # Code der ev. <Error> auslöst
                      # <Error>: Error oder Tuple von Errors
except <Error> as e:
    behandle_den_Fehler(e)
else:
    # wird nur im Erfolgsfall ausgeführt
    ...
finally:
    # aufräumen, wird immer ausgeführt
    ...
```

**Best Practice**:
- Sei spezifisch beim Abfangen von Fehlern. Nicht eingeplante Fehler sollten immer noch
  einen Fehler auslösen.  
- Brauche **nie** nur `except:`. Dies fängt zuviele Fehler ab.
  Besser: `except Exception as e:`.
- Halte den try-Block klein: Nur die Codezeile, die den Error auslösen könnte.

**Gängige Errors**:
- `NameError`: undefinierte Variable,
- `TypeError`: unpassender Typ,
- `ValueError`: passender Typ, aber unpassender Wert,
- `IndexError`: Index unpassend,
- `KeyError`: Schlüssel nicht im Dict,
- `ZeroDivisionError`: Divison durch 0,
- `ImportError`: Importversuch nicht erfolgreich.

````python
try:
    x  # ersetze diese Zeile mit einer anderen fehlerausloesenden Zeile
except Exception as e:
    print(f'Fehlermeldung: {e}, Error: {e.__class__}')
````

````python
divisor = 'foo'

try:
    result = 1 / divisor
except (TypeError, ZeroDivisionError):
    result = 'undefined'

result
````

````python
def get_number():
    try:
        val = int(input('Zahl:'))
    except ValueError:
        print('Das war keine Zahl! Ich interpretiere das als "None"')
        return None  # wird erst nach dem finally-Blockausgeführt!
    else:
        print(f'Deine Zahl: {val} years')
    finally:
        print('Danke, dass Sie diese Funktion nutzen.')
    return val
````

````python
n = get_number()
print(f'got "{n}"')
````

````python
try:
    from random import randint
except ImportError:
    randint = lambda m, n: (m+n)//2

for i in range(3):
    print(f'Zufallszahl: {randint(1, 99)}')
````
