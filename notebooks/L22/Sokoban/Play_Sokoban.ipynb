### Sokoban
Das Spielfeld ist ein 8x8 Schachbrett.
Graue Felder sind blockiert. 
Die Zielfelder für die Boxen sind mit orangen Kreuzen markiert.
Boxen sind braun, ausser sie stehen bereits auf einem Zielfeld, dann sind sie orange.
Der Spieler wird als roter Kreis dargestellt (orange auf einem Zielfeld).
Der Spieler kann sich auf ein benachbartes freies Feld bewegen, oder **eine** Box verschieben.
Ziel ist es, alle Boxen aud die Zielfelder zu schieben.  

Wir speichern die Position des Spieler in einer Liste.
Die blockierten Felder, die Positionen der Boxen und die Zielpositionen sind in Mengen gespeichert. 

```python
player_pos = [4, 4]
boxes =   {(2, 3), (2, 5), (3, 3), (3, 4), (3, 5)}

           (2, 7), (3, 0), (3, 1), (3, 7), (4, 0), (4, 7), (5, 0), (5, 2), (5, 4), (5, 5),
           (5, 6), (5, 7), (6, 0), (6, 4), (7, 0), (7, 1), (7, 2), (7, 3), (7, 4),
           }
```
