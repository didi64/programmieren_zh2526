### Generators
Ein Generator-Objekt `gen` ist ein Objekt, das endlich oder unendlich viele Werte liefert.
Es kann auf folgende Weisen verwendet werden:

- `next(gen)`: Liefert einen nächsten Wert, oder erzeugt einen `StopIteration` Error, falls
  der Generator ausgeschöpft ist.
- `next(gen, default_wert)`: Liefert einen nächsten Wert oder `default_wert`.
- `items = list(gen)`: Liste mit allen erzeugten Werten.
- `items = [x for x in gen]`: dito.
- `for x in gen: print(x)`: iteriert über alle Werte des Generators. Kurzform für
    ```python
     while True:
        try:
            x = next()
        except StopIteration:
            break
    
        print(x)
    ```

**Wichtig**: Sobald ein Generator-Objekt alle Werte geliefert hat, ist es leer.
  Von da an erzeugt jeder Aufruf `next(gen)` einen `StopIteration` Error, und
  z.B. `list(gen)` liefert dann immer eine leere Liste `[]`.

**Bemerkung**: `rg = range(3)` ist **kein** Generator. `rg` lässt sich wiederverwenden!


**Generatoren erstellen**
1. **Generator-Funktion**:
   Eine Generator-Funktion wird wie eine normale Funktion erstellt, muss aber das
   Schlüsselwort `yield` enthalten.
   `yield` bez. `yield <Expression>` gibt wie `return` einen Wert zurück, pausiert
   aber die Ausführung des Codes nur. Die Ausführung endet, falls auf das Ende der Funktion erreich ist, oder
    `return` angetroffen wird.

   Die Generator-Funktion gibt ein Generator-Objek zurück (die Generator-Funktion selber ist noch kein Generator). 
    
   
   ```python
   def numbers_generator(n):  # Generator-Funktion
       i = 1
       while i <= n:
           yield i
           i = i + 1


    numbers = numbers_generator(3)  # liefert ein Generator-Objekt numbers  
    [next(numbers), next(numbers), next(numbers)]  # [1, 2, 3]

    next(numbers)  # StopIteration  (Generator ist leer, bez. am Ende der Funktion oder einem return)
    next(numbers, 'I am done')  # 'I am done'
    ```

3. **Generator-Ausdruck**:  
   Ein Generator-Ausdruck ist ein List-Comprehenion Ausdruck mit runden statt eckigen Klammern.
   Generator-Ausdruck erstellt direkt ein Generator-Objekt.
   ```python
   squares = (x**2 for x in range(1, 10))  # kein Tuple, sondern ein Generator
   
   [next(squares) for _ in range(8)]  # [1, 4, 9, 16, 25, 36, 49, 64]
    next(squares)  # 81
    next(squares, 'I am done')  # 'I am done'
    next(squares)  # StopIteration
   ```

   
   
**Typisches Beispiel**
```python
def find_chars(s, chars):
    '''liefert die naechste Position an der eines der Zeichen in chars vorkommt'''
    for i, c in enumerate(s):
        if c in chars:
            yield i
```            

````python
def numbers_generator(n):  # Generator-Funktion
    i = 1
    while i <= n:
        yield i
        i = i + 1
````

````python
def fibonacci_gen():
    x, y = 1, 1
    while True:
        yield x
        x, y = y, x+y
````

````python
fibonacci = fibonacci_gen()
fibonacci
````

````python
[next(fibonacci) for _ in range(10)]  # die 10 naechsten Fibonacci Zahlen
````

````python
# list(fibonacci)  # runs forever!
````

````python
def get_letters():
    char = 'A'
    while char <= 'Z':
        yield char
        char = chr(ord(char)+1)
````

````python
letters = get_letters()

''.join(next(letters) for _ in range(13))
````

````python
list(letters)
````

````python
next(letters, '.')
````

````python
s = ''
for c in get_letters():
    s += c

s
````

````python
s = ''
letters = get_letters()
while True:
    try:
        c = next(letters)
    except StopIteration:
        break

    s += c

s
````

### Aufgaben
Schreibe Generatoren, die Folgendes liefern:
1. `get_lines(s)`: Zeilen eine Strings `s` (Zeilen sind durch `'\n'` getrennt).
2. `find_chars(s, chars))`: Alle Positionen, an denen eines der Zeichen in `chars` in `s` vorkommt.
3. `pos_and_values(matrix)`: Alle Tuple `(pos, val)` mit Position (col, row) und zugehörigem Matrixfeld.
4. `get_positions(matrix , chars)`: Alle Positionen, an denen eines der Zeichen in `chars` in `matrix` vorkommt (die Matrixeinträge seien Zeichen).

Output für die Strings und Matrix der folgenden Zelle:

```python
list(get_lines(lines))  # ['aaaa', 'bbbbbb', 'cccccc', 'dddddddd']
list(find_chars(level, '$*'))  # [26, 27, 35, 42, 43]
list(pos_and_values(mat))  # [((0, 0), ' '), ((1, 0), ' '), ((6, 7), ' '), ((7, 7), ' ')]
list(get_positions(mat, '$*'))  # [(2, 3), (3, 3), (3, 4), (2, 5), (3, 5)]
```
