### Generators
Ein Generator-Objekt `gen` ist ein Objekt, das endlich oder unendlich viele Werte liefert.
Es kann auf folgende Weisen verwendet werden:

- `next(gen)`: Liefert einen nächsten Wert, oder erzeugt einen `StopIteration` Error, falls
  der Generator ausgeschöpft ist
- `next(gen, default_wert)`: Liefert einen nächsten Wert, oder `default_wert`
- `items = list(gen)`: Liste mit allen erzeugten Werten
- `items = [x for x in gen]`: dito
- `for x in gen: print(x)`: Iteriert über alle Werte des Genewrators. Kurzform für
  ```python
    while True:
        try:
            x = next(gen)
        except StopIteration:
            break
    
        print(x)
    ```

**Wichtig**: Sobald ein Generator-Objekt alle Werte geliefert hat, ist es leer.
  Von da an erzeugt jeder Aufruf `next(gen)` einen `StopIteration` Error, und
  z.B. `list(gen)` liefert dann immer eine leere Liste `[]`.

**Bemerkung**: `rg = range(3)` ist kein Generator. `rg` lässt sich wiederverwenden!


**Generatoren erstellen**
1. **Generator-Funktion**:
   Eine Generator-Funktion wird wie eine normale Funktion erstellt, muss aber das
   Schlüsselwort `yield` enthalten.
   `yield` bez. `yield <Expression>` gibt wie `return` einen Wert zurück, pausiert
   aber die Ausführung des Codes nur.
   
   ```python
   def numbers_generator(n):  # Generator-Funktion
       i = 1
       while i <= n:
           yield i
           i = i + 1


    numbers = numbers_generator(3)  # liefert ein Generator-Objekt numbers  
    [next(numbers), next(numbers), next(numbers)]  # [1, 2, 3]

    next(numbers)  # StopIteration  (Generator ist leer, bez. am Ende der Funktion oder einem return)
    next(numbers, 'I am done')  # 'I am done'
    ```

2. **Generator-Ausdruck**:  
   Ein Generator-Ausdruck ist ein List-Comprehenion Ausdruck mit runden statt eckigen Klammern.
   Generator-Ausdruck erstellt direkt ein Generator-Objekt.
   ```python
   squares = (x**2 for x in range(1, 10))  # kein Tuple, sondern ein Generator
   
   [next(squares) for _ in range(8)]  # [1, 4, 9, 16, 25, 36, 49, 64]
    next(squares)  # 81
    next(squares, 'I am done')  # 'I am done'
    next(squares)  # StopIteration
   ```

````python
def fibonacci_gen():
    x, y = 1, 1
    while True:
        yield x
        x, y = y, x+y
````

````python
def get_letters():
    char = 'A'
    while char <= 'Z':
        yield char
        char = chr(ord(char)+1)
````

````python
import random


def get_random_neighbors(x, y, dims):
    directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
    random.shuffle(directions)

    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < dims[0] and 0 <= ny < dims[1]:
            yield (nx, ny)


def get_moves(start=(0, 0), dims=(4, 2)):
    stack = [(start)]
    visited = {start}

    while len(visited) < dims[0]*dims[1]:
        pos = stack.pop()
        for npos in get_random_neighbors(*pos, dims):
            if npos in visited:
                continue
            visited.add(npos)
            stack.append(npos)
            yield pos, npos
````

````python
import widget_helpers as W
import grid_helpers as G


def draw_maze(ncol, nrow, dx=20, dy=20):
    grid_spec = (0, 0, dx, dy, ncol, nrow)
    mcanvas = W.get_mcanvas(2, width=ncol*dx, height=nrow*dy)
    bg, fg = mcanvas

    G.draw_grid(bg, grid_spec)

    fg.stroke_style = 'blue'
    fg.line_width = 2
    for src, dst in get_moves(dims=(ncol, nrow)):
        p = G.cr2xy(*src, grid_spec, center=True)
        q = G.cr2xy(*dst, grid_spec, center=True)
        fg.stroke_line(*p, *q)
    display(mcanvas)
````
