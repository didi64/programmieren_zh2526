### Matrizen (Listen von Listen)
Wird einer Variable  `2`, `2.0` `'foo'`, `True`, `(1, 2)` oder `None` zugewiesen,
wird ein neues Objekt des entsprechenden Typs erstellt oder ein bestehendes wiederverwendet, und eine 
Referenz (Adresse) in der Variable gespeichert.
Objeke vom Typ `int`, `float`, `str`, `bool`, `tuple` oder `None` sind **immutable** (nicht modifizierbaren).
Ihr Wert kann **nicht** geändert werden.

Speichern dagegen die Variabeln `xs` und `ys` die Adresse eines **mutable** (modifizierbaren) Objekts, 
z.B. eine Liste, so verlängern `xs.append` und `ys.append` die gleiche Liste.

Der Operator `x is y` testet, ob `x` und `y` die gleiche Adresse (bez. das gleiche Objekt) speichern.  
Der Operator `x == y` testet, ob `x` und `y` den gleichern Wert haben.  
Bei Listen `xs` und `ys` testet `xs == ys`, ob entsprechende Elemente bez. `==` gleich sind, d.h., ob    
 `all(x == y for x, y in zip(xs, ys))`

````python
x = 1  # Adresse eines Int-Objekt mit Wert 1 wird in x gespeichert
print(f'Wert von x: {x}, Adresse von x: {hex(id(x))}')

y = x  # x und y referenzieren das gleiche Integer-Objekt
print(f'Wert von y: {y}, Adresse von y: {hex(id(y))}')

x = 2  # Adresse eines Int-Objekt mit Wert 2 wird in x gespeichert
# x und y referenzieren nicht mehr das gleiche Objekt
print(f'Wert von x: {x}, Adresse von x: {hex(id(x))}')
````

````python
xs = [1, 2, 123456]  # die Integer-Objekte 1 und 2 werden wiederverwendet
ys = [1, 2, 123456]
print([hex(id(x)) for x in xs])
print([hex(id(y)) for x in ys])
[x is y for x, y in zip(xs, ys)]
````

````python
xs = [1]  # neue Liste wird erstellt
print(f'Wert von xs: {xs}   , Adresse von xs: {hex(id(xs))}')

ys = xs
print(f'Wert von ys: {ys}   , Adresse von ys: {hex(id(ys))}')

xs.append(2)  # Liste wird modifiziert
print(f'Wert von xs: {xs}, Adresse von xs: {hex(id(xs))}')
print(f'Wert von ys: {ys}, Adresse von ys: {hex(id(ys))}')
````

### Kopieren von Listen
Sei `xs = [1, 2, 3]`. Dann ist  
```python
ys = xs                 # dieselbe Liste  

zs_1 = [x for x in xs]  # neue Liste mit den gleichen Elementen wie xs
zs_2 = xs[:]            # neue Liste mit den gleichen Elementen wie xs
zs_3 = xs.copy()        # neue Liste mit den gleichen Elementen wie xs

zs_4 = []
zs_4[:] = xs            # neue Liste mit den gleichen Elementen wie xs
```

Ist  `A = [[1, 2, 3], [4, 5, 6]]` eine Matrix, so ist
```python
B = A                          # A und B sind identische Matrizen,
C = A.copy()                   # A und C haben identischen Zeilen,
D = [row.copy() for row in A]  # A und D haben verschiedene Zeilen, 
                               # aber mit den gleichen Elementen
```

````python
A = [[1, 2, 3], [4, 5, 6]]
C = A.copy()
D = [row.copy() for row in A]

print(all(a is c for a, c in zip(A, C)))
print(any(a is d for a, d in zip(A, D)))
print(any(a == d for a, d in zip(A, D)))
````

````python
def copy_matrix_1(matrix):
    return matrix.copy()


def copy_matrix(matrix):
    return [row.copy() for row in matrix]
````

````python
A = [[1, 2], [3, 4]]
B = copy_matrix_1(A)
C = copy_matrix_1(A)
B[0] = [5, 6]  # B erhaelt neue Zeile
C[0][0] = 7    # C[0] is A[0], diese Zeile wird geandert

print(f'A[0]: {A[0]}')
print(f'B[0]: {B[0]}')
````

````python
A = [[1, 2], [3, 4]]
B = copy_matrix(A)
C = copy_matrix(A)
B[0] = [5, 6]  # B erhaelt neue Zeile
C[0][0] = 7    # C[0] ist eine andere Zeile als A[0], A[0] wird nicht geaendert

print(f'A[0]: {A[0]}')
print(f'B[0]: {B[0]}')
````

````python
def have_identical_rows(A, B):
    return all(x is y for x, y in zip(A, B))


def have_rows_with_same_elements(A, B):
    return all(x == y for x, y in zip(A, B))


def print_crosstable(matrices, names):
    # erstelle Tabelle
    n = len(matrices)
    crosstable = [[None]*n for _ in range(n)]
    for i, xs in enumerate(matrices):
        for j, ys in enumerate(matrices):
            crosstable[i][j] = int(have_identical_rows(xs, ys))


    # gib Tabelle aus
    header = '   ' + ' '.join(names)
    print(header)
    for name, row in zip(names, crosstable):
        line = ''.join(f' {x} ' for x in row)
        print(f'{name}{line}')
````

````python
A = [[1, 2, 3], [4, 5, 6]]

B = A

C1 = [row for row in A]
C2 = A[:]
C3 = A.copy()

C4 = []
C4[:] = A

D1 = [[x for x in row] for row in A]
D2 = [row[:] for row in A]
D3 = [row.copy() for row in A]

matrices = [A, B, C1, C2, C3, C4, D1, D2, D3]
names = ['A ', 'B ', 'C1', 'C2', 'C3', 'C4', 'D1', 'D2', 'D3']

print_crosstable(matrices, names)
````
