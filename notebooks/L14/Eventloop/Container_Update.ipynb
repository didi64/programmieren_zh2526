### Container-Update
- Führe im Windows Command Prompt Folgendes aus:  
> docker pull dieterprobst/jlab:3.1

- In Docker Destkop, klicke rechts auf Images. Bewege die Mause über die Zeile  
> dieterprobst/jlab:3.1
> 
und klicke "run".
- Expandiere *Optional Settings* und gib bei *Local Host*, *Host Path* und *Container Path*
  Folgendes an:
```
Ports
Local Host                      Container-Port
10000                           8888/tcp

Volumes
Host Path                       Container Path
selektiere deinen Ordner        /home/studi/work   
programmiern                       
```
