```python
import game
import darstellung as D
from eventloop import event_loop, stop_event, event_queue
from IPython.display import display


# um nicht immer D.out und D.canvas tippen zu muessen
out = D.out
canvas = D.canvas


@out.capture(clear_output=True)
def on_key_down(key, *flags):
    print(key)
    if key.startswith('Arrow'):
        event_queue.insert(0, key)
    if key == 'n':
        game.new_game()


def update(event, **kwargs):
    # out.append_stdout(f'{event}, {kwargs}')
    if event == 'new_game':
        D.new_game(**kwargs)
        stop_event.clear()
        event_loop()
    if event == 'game_over':
        D.game_over()
        stop_event.set()
    if event == 'move':
        D.move(**kwargs)


game.update = update
canvas.on_key_down(on_key_down)

display(canvas, out)
canvas.focus()
```
