### Snake mit Eventloop 
Wir folgen dem Vorgehen in Notebook `Eventloop.ipynb`

Wird eine Pfeiltaste gedrückt, wird der entsprechende Richtungsvektor als erstes Element in eine Liste `event_queue` eingefügt. 
Ist der Eventqueue  nicht leer, entfernt 
der Eventloop den letzten Richtungsvektor, andernfalls wird der letzte Richtungsvektor beibehalten.

***
Starte vor Ausführen der nächsten  Zelle den Kernel neu, um erneutes Importieren zu erzwingen, damit eine neue Canvas verwendet wird. 
***

```python
import snake as game
import darstellung_snake as D
from ipywidgets import Output
import eventloop as E
from IPython.display import display


E.move = game.move

key_direction = {'ArrowUp': (0, -1),
                 'ArrowDown': (0, 1),
                 'ArrowLeft': (-1, 0),
                 'ArrowRight': (1, 0),
                 }

layout = {'border': '1px solid black'}
out = Output(layout=layout)

canvas = D.canvas


@out.capture(clear_output=True)
def on_key_down(key, *flags):
    print(key)
    if key in key_direction:
        direction = key_direction[key]
        print(direction)
        E.event_queue.insert(0, direction)
    if key == 'n':
        game.new_game()


def update(event, **kwargs):
    # out.append_stdout(f'{event}, {kwargs}')
    if event == 'new_game':
        D.new_game(**kwargs)
        E.stop_event.clear()
        E.event_loop()
    if event == 'game_over':
        D.game_over()
        E.stop_event.set()
    if event == 'move':
        D.move(**kwargs)


game.update = update
canvas.on_key_down(on_key_down)

display(canvas, out)
canvas.focus()
```
