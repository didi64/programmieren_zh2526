### Threads
Threads erlauben es, mehrere Funktionen (scheinbar) parallel auszuführen.
Tatsächlich wird rasch von einem Thead zum nächsten gesprungen. 


***
`threading.Thread(target = <Funktion>, args = <Iterable mit Argumenten>)`
macht aus einer Funktion ein Thread-Objekt, welches dann mit `start()` gestartet werden kann.
***

````python
import threading


def f_1(n):
    for i in range(n):
        print('A', end='')


def f_2(n):
    for i in range(n):
        print('B', end='')
````

````python
thread_1 = threading.Thread(target=f_1, args=(700,))
thread_2 = threading.Thread(target=f_2, args=(700,))


thread_1.start()
thread_2.start()
````

***
`t = threading.Timer(delay, function, args)` kreiert einen Thread `t`.  
`t.start()` startet diesen Thread dann mit `delay` Sekunden Verz&ouml;gerung.

***

````python
out = Output(layout={'border': '1px solid black'})
display(out)

thread = threading.Timer(1, out.append_stdout, args=('Sorry fuer die Verspaetung',))
thread.start()
print('Thread gestartet ...')
````

***
**threading.Event**:  
Eine Klasse mit  Methoden

- `is_set()` : liefert `True` oder `False` (Default).
- `set()`    : `is_set()` liefert nun `True`
- `clear()`  : `is_set()` liefert nun `False`
***

````python
stop_event = threading.Event()
stop_event.is_set()
````

````python
stop_event.set()
stop_event.is_set()
````

````python
stop_event.clear()
stop_event.is_set()
````

***
Mit  `threading.Timer` kann z.B. ein Event-Loop implementiert werden:
Nachstehend definierte Funktion `event_loop` liest jede Sekunde das 1. Element einer Liste `source`
(z.B. Eingaben w&auml;hrend eines Spiels) und bearbeitet es (hier schieben wir die Elemente der Liste `source` nur weiter in eine Liste `targets`).
Ist die **Liste** `source` **leer**, **wird** das **letzte Element wiederholt**.

***

````python
out = Output(layout={'border': '1px solid black'})
canvas = Canvas(width=100, height=100, layout={'border': '1px solid black'})

state = {'pos': (50, 50), 'count': 0}
directions = {'Up': (0, -5), 'Down': (0, 5), 'Left': (-5, 0), 'Right': (5, 0)}

stop_event = threading.Event()
event_queue = []


@out.capture()
def on_key_down(key, *flags):
    if key.startswith('Arrow'):
        event_queue.append(key[5:])
    print(key)


def event_loop(event=None):
    if event_queue:
        event = event_queue.pop(0)

    dx, dy = directions.get(event, (0, 0))
    x0, y0 = state['pos']
    x, y = (x0+dx, y0+dy)
    canvas.stroke_lines([(x0, y0), (x, y)])
    state['pos'] = (x, y)
    state['count'] += 1

    if x < 0 or y < 0 or x > 100 or y > 100:
        canvas.fill_text('Game over!', 10, 50)
        stop_event.set()

    if not stop_event.is_set():
        thread = threading.Timer(1, event_loop, args=(event,))
        thread.name = f'MyThread-{state['count']}'
        thread.start()


canvas.on_key_down(on_key_down)


display(canvas, out)
event_loop('Right')
canvas.focus()
````

````python
# stoppt den event_loop,
stop_event.set()
````
