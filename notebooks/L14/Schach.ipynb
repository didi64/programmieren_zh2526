```python
# import schach2 as schach
from schach3 import Schach
import schach_D as D
import helpers as H
from ipycanvas import MultiCanvas
from ipywidgets import Output, Button
from IPython.display import display

schach = Schach()

state = {'mouse_down': None}

layout = {'border': '1px solid black'}
out = Output(layout=layout)
button = Button(description='New Game')
mcanvas = MultiCanvas(2, width=200, height=200, layout=layout)
bg, fg = mcanvas


@out.capture(clear_output=True)
def on_mouse_down(x, y):
    state['mouse_down'] = (x, y)
    print(f'A (state[\'mouse_down\']) set to ({int(x)}, {int(y)})')


@out.capture()
def on_mouse_up(x, y):
    target = H.xy2cr(x, y, D.BOARD_SPEC)
    src = H.xy2cr(*state['mouse_down'], D.BOARD_SPEC)
    schach.raw_move(src, target)
    print(f'move from {src} to {target}')


D.draw_chessboard(bg)

schach.update = lambda event, **kwargs: D.update(fg, event, **kwargs)
schach.new_game()

mcanvas.on_mouse_down(on_mouse_down)
mcanvas.on_mouse_up(on_mouse_up)

button = Button(description='New Game')
button.on_click(lambda _: schach.new_game())  # Argument ignorieren

display(mcanvas, button, out)
```

```python
from functools import wraps


def capture_output(f, clear_output=False):
    @wraps(f)
    @out.capture(clear_output=False)
    def g(*args, **kwargs):
        return f(*args, **kwargs)
    return g
```

```python
class View:
    def __init__(self):
        self.mouse_down = None

        self.layout = {'border': '1px solid black'}
        self.out = Output(layout=layout)

        self.mcanvas = MultiCanvas(2, width=200, height=200, layout=layout)
        self.bg, self.fg = mcanvas

        @self.out.capture(clear_output=True)
        def on_mouse_down(x, y):
            self.on_mouse_down(x, y)

        @self.out.capture()
        def on_mouse_up(x, y):
            self.on_mouse_up(x, y)

        self.mcanvas.on_mouse_down(on_mouse_down)
        self.mcanvas.on_mouse_up(on_mouse_up)

    def on_mouse_down(self, x, y):
        self.mouse_down = (x, y)
        print(f'A (mousedown) set to ({int(x)}, {int(y)})')

    def on_mouse_up(self, x, y):
        self.mouse_down = None
        print(f'move from {self.mouse_down} to {(x, y)}')

    def _ipython_display_(self):
        display(self.mcanvas, self.out)
```
