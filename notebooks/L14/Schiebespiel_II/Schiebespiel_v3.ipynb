### Schiebespiel mit Maussteuerung
- Ein Klick auf den obersten Kreis (Kreis mit Index 0) soll einen Doppeltausch auslösen.
- Ein Klick auf den untersten Kreis (Kreis mit Index 3) soll eine Rotation auslösen.

Der Dict
```python
idx_action = {0: game.swap, 3: game.rotate}
```
ordnet dem Index des Kreises die Funktion zu, die wir beim Klicken auf diesen Kreis
ausrufen wollen. 


Speichert `pts` die Kreismittelpunkte und wird an der Stelle `(x, y)` auf die
Leinwand geklickt, dann liefert
`get_nearest_idx((x, y), pts)` den Index `idx` des Kreises, und
`idx_action.get(idx)` gibt dann die aufzurufende Funktion oder `None` zurück.

````python
import game
import darstellung as D
from ipywidgets import Output
from ipycanvas import MultiCanvas
from IPython.display import display


LAYOUT = {'border': '1px solid black'}
CENTER_X, CENTER_Y, RADIUS = 100, 140, 40

# Mittelpunkte der 5 Kreise fuer die Zahlen
nodes = D.get_centers(CENTER_X, CENTER_Y, RADIUS)

out = Output(layout=LAYOUT)
mcanvas = MultiCanvas(2, width=200, height=200, layout=LAYOUT)
bg, fg = mcanvas
fg.text_align = 'center'
fg.text_baseline = 'middle'


def update(event, data):
    if event == 'new_game':
        D.draw_bg(bg, CENTER_X, CENTER_Y, RADIUS)
        D.update_scramble(fg, data['scramble'], nodes)
    if event in ('swap', 'rotate'):
        D.update_scramble(fg, data['scramble'], nodes)
    if event == 'solved':
        bg.fill_text('Congrats!', 100, 20)


@out.capture(clear_output=True)
def on_mouse_down(x, y):
    print(f'mouse down: {(x, y)}')
    idx = D.get_nearest_idx((x, y), nodes)

    if idx == 0:
        game.swap()
    elif idx == 3:
        game.rotate()


@out.capture(clear_output=True)
def on_key_down(key, *flags):
    if key == 'ArrowUp':
        game.swap()
    if key == 'ArrowRight':
        game.rotate()
    if key == 'n':
        game.new_game()


mcanvas.on_key_down(on_key_down)
mcanvas.on_mouse_down(on_mouse_down)


D.draw_bg(bg, CENTER_X, CENTER_Y, RADIUS)

game.update = update
game.new_game()

display(mcanvas, out)
````
