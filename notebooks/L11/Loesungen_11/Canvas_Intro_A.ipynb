### Aufgaben mit Lösungen
Erstellen eine neue Leinwand in der Grösse 200x200 mit schwarzem Rand.
1. Zeichne ein ein Schachbrett in Leinwandgrösse in den Farben grey und blue.
1. Schreibe eine Funktion `draw_chessboard_1(canvas, colors)`, die ein Schachbrett in den angegebenen Farben auf das Canvas-Objekt `canvas` zeichnet.
1. Der Dict der folgenden Form beschreibt ein Schachbrett.
   ```python
   chessboard_config = {
       'position': (20, 20),  # Position der linken oberen Ecke
       'width': 160,            
       'height': 160,
       'n': 8,                # Brett hat n x n Felder  
       'colors': ('grey', 'blue'),
   }
   ```
   Schreibe eine Funktion `draw_chessboard(canvas, position, width, height, n, colors)`,
   so dass  `draw_chessboard(canvas, **chessboard_config)` das gewünsche Brett zeichnet.
1. Schreibe eine Funktion
   `place_stone(canvas, col, row, color, position, width, height, n, colors, radius=1)`,  
   so dass `place_stone(canvas, col, row, color, **chessboard_config, radius=1)`
   einen Stein der Farbe `color` auf das Feld (`col`, `row`) setzt.
1. Schreibe eine entsprechende Funktion
   `remove_stone(canvas, col, row, color, position, width, height, n, colors, radius=1)`

````python
from ipycanvas import Canvas
````

````python
# Leinwand erstellen
canvas = Canvas(width=200, height=200, layout={'border': '1px solid black'})
canvas
````

````python
# Schachbrett zeichnen
colors = ('blue', 'grey')
n = 8
dx, dy = canvas.width / 8, canvas.height / 8

for col in range(n):
    for row in range(n):
        color_idx = (col + row) % 2
        color = colors[color_idx]
        canvas.fill_style = color
        canvas.fill_rect(col*dx, row*dy, dx, dy)
````

````python
def get_color(col, row, colors):
    idx = (row+col) % 2
    return colors[idx]


def draw_chessboard_1(canvas, colors):
    canvas.clear()
    dx, dy = canvas.width / 8, canvas.height / 8
    for col in range(8):
        for row in range(8):
            color = get_color(col, row, colors)
            canvas.fill_style = color
            canvas.fill_rect(col*dx, row*dy, dx, dy)
````

````python
# Canvas loeschen und anzeigen
canvas.clear()
canvas
````

````python
draw_chessboard_1(canvas, ('blue', 'teal'))
````

````python
def draw_chessboard(canvas, position, width, height, n, colors):
    x0, y0 = position
    dx, dy = width/n, height/n
    for row in range(n):
        for col in range(n):
            color = get_color(col, row, colors)
            canvas.fill_style = color
            canvas.fill_rect(x0+col*dx, y0+row*dy, dx, dy)
````

````python
chessboard_config = {
    'position': (20, 20),  # Position der linken oberen Ecke
    'width': 160,
    'height': 160,
    'n': 8,                # Brett hat n x n Felder
    'colors': ('grey', 'blue'),
}


canvas.clear()
draw_chessboard(canvas, **chessboard_config)
````

````python
def place_stone(canvas, col, row, color, position, width, height, n, colors, radius=1):
    x0, y0 = position
    dx, dy = width/n, height/n
    max_radius = min(dx, dy)/2 - 3
    radius = max_radius*radius

    canvas.fill_style = color
    canvas.fill_circle(x0+(col+0.5)*dx, y0+(row+0.5)*dy, radius)
````

````python
canvas
````

````python
place_stone(canvas, 2, 3, 'yellow', **chessboard_config)
````

````python
def remove_stone(canvas, col, row, position, width, height, n, colors):
    x0, y0 = position
    dx, dy = width/n, height/n
    x, y = x0+col*dx, y0+row*dy

    color = get_color(col, row, colors)
    canvas.fill_style = color
    canvas.fill_rect(x, y, dx, dy)
````

````python
remove_stone(canvas, 2, 3, **chessboard_config)
````
