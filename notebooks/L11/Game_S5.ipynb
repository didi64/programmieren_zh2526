```python
state = [1, 2, 3, 4, 5]


def update(event, data):
    print(f'event: {event}, data: {data}')


def swap():
    state[1], state[0] = state[0], state[1]
    state[4], state[2] = state[2], state[4]
    update('swap', state)


def rotate():
    state[1:] = state[-1:] + state[1:-1]
    update('rotate', state)


def new_game():
    state[:] = [1, 2, 3, 4, 5]
    update('new_game', state)
```

```python
new_game()
```

```python
rotate()
```

```python
from IPython.display import display
from ipycanvas import MultiCanvas
from ipywidgets import Output


layout = {'border': '1px solid black'}
out = Output(layout=layout)
```

```python
@out.capture(clear_output=True)
def on_key_down(key, *flags):
    print(f'key pressed: {key}')
    if key == 'ArrowUp':
        swap()
    if key == 'ArrowRight':
        rotate()
    if key == 'n':
        new_game()


# @out.capture()
# def update(event, data):
#     print(f'event: {event}, data: {data}')


r = 40
cx, cy = 100, 140

pts = [(cx, cy-2*r),
       (cx, cy - r),
       (cx + r, cy),
       (cx, cy+r),
       (cx-r, cy),
       ]


@out.capture()
def update(event, data):
    print(f'event: {event}, data: {data}')
    fg.clear()
    for i, n in enumerate(state):
        x, y = pts[i]
        fg.fill_text(str(n), x, y)
```

```python
def draw_bg(bg):
    bg.fill_text('Swap (Up) and Rotate (Right)', 20, 20)
    bg.stroke_style = 'blue'
    bg.stroke_circle(cx, cy, r)
    bg.stroke_style = 'red'
    bg.stroke_lines([*pts[0], *pts[1]])
    bg.stroke_lines([*pts[2], *pts[4]])

    bg.stroke_style = 'black'
    r1 = 10
    for x, y in pts:
        bg.clear_rect(x-r1, y-r1, 2*r1)
        bg.stroke_circle(x, y, r1)
```

```python
mcanvas = MultiCanvas(2, width=200, height=200, layout=layout)
bg, fg = mcanvas
fg.text_align = 'center'
fg.text_baseline = 'middle'

draw_bg(bg)
mcanvas.on_key_down(on_key_down)

display(mcanvas, out)
```
