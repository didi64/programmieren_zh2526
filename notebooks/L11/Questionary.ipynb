### Fragebogen auswerten
Mitarbeitende haben einen Fragebogen ausgefüllt, der die Zufriedenheit mit
der Arbeit, dem Lohn und einem internen Tool auf einer Skala von 1 bis 5 bewertet.
Manche Fragen wurden nicht beantwortet.

Wir wollen bei jeder Frage angeben, wieviele Antworten eingingen,
wieviele unglücklich sind (<= 2), wieviele glücklich sind (>= 4) sowie den Durchschnitt 
(leere Antworten werden nicht berücksichtigt).

Hier das Resultat der Umfrage als Tabelle.
```python
[['Arbeit', 'Lohn', 'Tool'],
 ['2', '3', '3'],
 ['4', '5', '4'],
 ['4', '', ''],
 ['', '2', '1'],
 ]

HAPPY = 4
UNHAPPY = 2
```

1. Erstelle einen Dict `frage_idx`, der jedem Spaltenheader den Index zuordnet.  
   `{'Arbeit': 0, 'Lohn': 1, 'Tool': 2}`
3. Schreibe eine Funktion `get_values(frage)`, die eine Liste mit den Zahlen in der entsprechenden Spalte als
   Integer zurück gibt (leere Antworten werden ignoriert).  
   Beispiel: `get_values('Arbeit') -> [2, 4, 4]`
5. Schreibe eine Funktion `summarize(values)`, die ein Tupel zurück gibt mit
    - Anzahl Werte,
    - Durchschnitt, gerundet auf eine Nachkommastelle,
    - Anzahl Werte, die grösser gleich `HAPPY` sind,
    - Anzahl Werte, die kleiner gleich `UNHAPPY` sind.
      
    `summarize(get_values('Arbeit')) -> (3, 3.3, 1, 2)`
7. Schreibe eine Funktion `get_results(table)`, die
   eine Liste mit den Tupel der zusammengefassten Werten zu jeder Frage enthält.  
   `get_results(table) -> [(3, 3.3, 1, 2), (3,  3.3, 1, 1), (3, 2.7, 1, 1)]`

````python
table = [['Arbeit', 'Lohn', 'Tool'],
         ['2', '3', '3'],
         ['4', '5', '4'],
         ['4', '', ''],
         ['', '2', '1'],
         ]

HAPPY = 4
UNHAPPY = 2
````

````python
frage_idx = {frage: i for i, frage in enumerate(table[0])}
frage_idx
````

````python
def get_col(frage):
    idx = frage_idx[frage]
    values = [int(row[idx]) for row in table if row[idx].isdigit()]
    return values
````

````python
for frage in frage_idx:
    print(get_col(frage))
````

````python
def summarize(values):
    n = len(values)
    avg = sum(values) / n
    n_happy = len([v for v in values if v <= UNHAPPY])
    n_unhappy = len([v for v in values if HAPPY <= v])
    return (n, round(avg, ndigits=1), n_happy, n_unhappy)
````

````python
values = get_col('Arbeit')
summarize(values)
````

````python
def get_results(table):
    results = [summarize(get_col(frage)) for frage in frage_idx]
    return results
````

````python
get_results(table)
````
