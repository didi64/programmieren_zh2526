### Mehr Fullhouses oder mehr Straights?
Es wird 5 Würfeln gewürfelt.
Die Wurfbilder *12345* und *23456* nennen wir ein *Straight*.
Ein Wurfbild der Form *22255* (ein Paar und ein Triple) nennen wir Fullhouse.
Was ist wahrscheinlicher: Ein *Straight* oder ein *Fullhouse*? 

1. Schreibe eine Funktion `wuerfle(n=5)`, die ein Tuple mit n Zufallszahlen von 1 bis 6 zurück gibt
(benutze die Funktion `randint` im Modul `random`).
1. Schreibe eine Funktion `count_dict(wurf)`, die einen Dictionary erstellt, der zu jeder Zahl in der Liste `wurf` angibt, wie oft diese Zahl vorkommt.
1. Ist Wurf ein *Straight*, so hat `d` Länge 5 (5 verschiedene Zahlen) und enthält nicht sowohl 1 und 6.
Ist Wurf ein *Fullhouse*, so hat Count-Dict Länge 2 und 3 ist ein Wert (eine Zahl kommt 3-Mal vor).  
Schreibe Funktionen `is_straight(count_dict)` und `is_fullhouse(count_dict)`, die auf
*Straight* bez. *Fullhouse* testen.
Teste die Funktionen mit den Count-Dicts zu folgenden Würfen.
```python
wuerfe = [(1, 2, 3, 4, 5), (2, 3, 4, 5, 6), (1, 2, 4, 5, 6),
          (2, 2, 2, 4, 4), (4, 2, 2, 2, 4), (1, 1, 2, 2, 3)]
```
1.  Würfle 10'000 Mal und zähle, wieviele Straights und Fullhouses gewürfelt wurden.

````python
from random import randint


def wuerfle(n=5):
    return tuple(randint(1, 6) for _ in range(n))
````

````python
wurf = wuerfle()
wurf
````

````python
def make_count_dict(items):
    d = {}
    for item in items:
        d[item] = d.get(item, 0) + 1
    return d
````

````python
d = make_count_dict(wurf)
d
````

````python
def is_fullhouse(count_dict):
    return len(count_dict) == 2 and 3 in count_dict.values()


def is_straight(count_dict):
    return len(count_dict) == 5 and not (1 in count_dict and 6 in count_dict)
````

````python
# is_full_house und is_straight testen
wuerfe = [(1, 2, 3, 4, 5), (2, 3, 4, 5, 6), (1, 2, 4, 5, 6),
          (2, 2, 2, 4, 4), (4, 2, 2, 2, 4), (1, 1, 2, 2, 3)]

for w in wuerfe:
    d = make_count_dict(w)
    print(w, is_straight(d), is_fullhouse(w))
````

````python
N = 10_000

n_straight = 0
n_fullhouse = 0

for _ in range(N):
    wurf = wuerfle(5)
    count_dict = make_count_dict(wurf)

    if is_straight(count_dict):
        n_straight = n_straight + 1
    if is_fullhouse(count_dict):
        n_fullhouse = n_fullhouse + 1

n_straight, n_fullhouse
````
