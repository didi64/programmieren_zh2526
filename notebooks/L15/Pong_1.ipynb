```python
import pong as game
import darstellung_pong as D
from ipywidgets import Output
from IPython.display import display

canvas = D.canvas  # um nicht immer D.canvas schreiben zu muessen
layout = {'border': '1px solid black'}
out = Output(layout=layout)

key_move = {'ArrowUp': -5,
            'ArrowDown': 5,
            }


@out.capture(clear_output=True)
def on_key_down(key, *flags):
    print(key)
    if key in key_move:
        dy = key_move[key]
        game.move(dy)
    if key == 'n':
        game.new_game()


def update(event, **kwargs):
    print(event, kwargs)
    if event == 'new_game':
        D.new_game(**kwargs)
    if event == 'game_over':
        D.game_over()
    if event == 'move':
        D.move(**kwargs)


game.update = update
canvas.on_key_down(on_key_down)
display(canvas, out)
canvas.focus()
```
