### Pong
Nachstehend zeigen wir, wie sich ein Siel wie Pong mit einem Eventloop implementieren
lässt. Dabei muss ein sich ständig bewegender Ball mit einem Schläger abgefangen werden.
Der Schläger bewegt sich auf oder ab, in Richung der zuletzt gedrücken Pfeiltaste.

Die Spielkomponente hat eine Funkion `move(dy)`, die den Schläger um `dy` nach oben
bewegt, und den Ball einen Schritt weiter bewegt.


```python
import pong as game
import darstellung_pong as D
from ipywidgets import Output
import eventloop as E
from IPython.display import display


E.move = game.move

layout = {'border': '1px solid black'}
out = Output(layout=layout)

canvas = D.canvas


@out.capture(clear_output=True)
def on_key_down(key, *flags):
    print(key)
    if key in E.move.startswith('Arrow'):
        E.event_queue.insert(0, key)
    if key == 'n':
        game.new_game()


def update(event, **kwargs):
    # out.append_stdout(f'{event}, {kwargs}')
    if event == 'new_game':
        D.new_game(**kwargs)
        E.stop_event.clear()
        E.event_loop()
    if event == 'game_over':
        D.game_over()
        E.stop_event.set()
    if event == 'move':
        D.move(**kwargs)


game.update = update
canvas.on_key_down(on_key_down)

display(canvas, out)
canvas.focus()
```
