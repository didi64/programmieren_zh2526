### HBox und VBox
Die Widgets HBox und VBox sind Container-Widgets, die
andere Widgets enthalten können. 
Siehe auch [hier](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html#hbox).

Mit HBox und VBox lässt sich das Layout einer simplen Applikation auf einfache Weise gestalten.


Nachstehend platzieren wir einmal 3 Buttons **über** einem Output-Widget (das über einem
anderen Output-Widget liegt), und einmal 3 Buttons **neben** einem Output-Widget.

```python
from ipywidgets import VBox, HBox, Button, Output
from IPython.display import display


layout_out = {'border': '1px solid black'}
layout_bt = {'border': '1px solid blue'}


out = Output(layout=layout_out | {'height': '200px'})

# setzten wir im Dict layout den Schluessel 'flex' auf '1 1 auto', so
# wird Widget bestmoeglichst eingepasst.
# out = Output(layout=layout_out | {'flex': '1 1 auto', 'height': '200px'})


button_1 = Button(description='Action 1', layout=layout_bt)
button_2 = Button(description='Action 2', layout=layout_bt)
button_3 = Button(description='clear', layout=layout_bt)

# Buttons uebereinander stapeln
buttons_h = HBox(children=[button_1, button_2, button_3])

# Buttons nebeneinander legen
buttons_v = VBox(children=[button_1, button_2, button_3])


@out.capture()
def action1(bt):
    print('Action 1')


@out.capture()
def action2(bt):
    print('Action 2')


@out.capture()
def action3(bt):
    out.clear_output()


button_1.on_click(action1)
button_2.on_click(action2)
button_3.on_click(action3)


display(buttons_h, out)
```

```python
app = HBox(children=[buttons_v, out])
app
```
