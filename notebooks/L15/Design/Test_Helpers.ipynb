### Hilfsfunktionen der Darstellungskomponente testen
Das Modul `helpers` enthält die Funktionen
- `draw_grid(canvas, board_spec, line_width=2, color='blue')`: zeichnet Gitter
- `get_midpoint(col, row, board_spec)`: liefert Feldmittelpunkt
- `clear_field(canvas, pos, board_spec)`: löscht Feld
- `fill_field(canvas, pos, board_spec, color=None)`: füllt Feld

`board_spec` ist ein Tuple `(x_upper_left, y_upper_left, Feldbreite, Feldhöhe, Anzahl_Spalten, Anzahl_Zeilen)`.

```python
import helpers as H
from ipycanvas import MultiCanvas


board_spec = (20, 20, 30, 30, 7, 5)
layout = {'border': '1px solid black'}
mcanvas = MultiCanvas(2, width=250, height=190, layout=layout)
bg, fg = mcanvas

mcanvas
```

```python
H.draw_grid(fg, board_spec)
H.fill_field(bg, (0, 0), board_spec, color='yellow')
x, y = H.get_midpoint(0, 0, board_spec)
fg.fill_circle(x, y, 2)
```
