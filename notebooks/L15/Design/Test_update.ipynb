### Update-Funktionen testen
Wir schreiben und testen Funktionen, welche
die von der `update` Funktion angesagten Modifikationen vornimmt.
Das Argument `canvas` ist die Leinwand auf die wir malen,
`bs` ist die Spezifikation des Gitters.

```python
import helpers as H


def new_game(canvas, bs, player, boxes):
    canvas.clear()
    x0, y0 = H.get_midpoint(*player, bs)
    canvas.fill_style = 'red'
    canvas.fill_circle(x0, y0, 4)

    for pos in boxes:
        H.fill_field(canvas, pos, bs, color='brown')


def move(canvas, bs, old, new):
    H.clear_field(canvas, old, bs)
    x, y = H.get_midpoint(*new, bs)
    canvas.fill_style = 'red'
    canvas.fill_circle(x, y, 4)


def push(canvas, bs, old, new):
    H.clear_field(canvas, old, bs)
    H.fill_field(canvas, new, bs, color='brown')
```

```python
def update(canvas, bs, event, **kwargs):
    if event == 'new_game':
        new_game(canvas, bs, **kwargs)
    if event == 'move':
        move(canvas, bs, **kwargs)
    if event == 'push':
        push(canvas, bs, **kwargs)
```

```python
from ipycanvas import MultiCanvas

targets = ((5, 0), (6, 0), (5, 1), (6, 1))
board_spec = (20, 20, 30, 30, 7, 5)
layout = {'border': '1px solid black'}
mcanvas = MultiCanvas(2, width=250, height=190, layout=layout)
bg, fg = mcanvas

H.draw_grid(fg, board_spec, color='blue')
for col, row in targets:
    x, y = H.get_midpoint(col, row, board_spec)
    fg.fill_circle(x, y, 2)

mcanvas
```

```python
player = (2, 1)
boxes = ((1, 1), (2, 2))
new_game(bg, board_spec, player, boxes)
```

```python
player_new = (2, 0)
move(bg, board_spec, player, player_new)
```

```python
push(bg, board_spec, (1, 1), (1, 0))
```

```python
update(bg, board_spec, event='new_game', player=player, boxes=boxes)
```

```python
update(bg, board_spec, event='move', old=player, new=player_new)
```

```python
update(bg, board_spec, event='push', old=(1, 1), new=(1, 0))
```
