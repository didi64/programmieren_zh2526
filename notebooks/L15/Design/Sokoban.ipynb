### Protoype Sokoban
Wir erstellen eine MultiCanvas mit 2 Layern, einem Hinter- undd Vordergrund.
Weiter spreiben wir eine Funktion `on_key_down`, welche auf Tastendrücke hört und
die Funktion `move` der Spielkomponente aufruft.

Der Dict `key_direction` ordnet einer Pfeiltaste die Positionsänderung zu.

```python
import game
import darstellung as D
import helpers as H
from ipywidgets import Output
from ipycanvas import MultiCanvas
from IPython.display import display


layout = {'border': '1px solid black'}
out = Output(layout=layout)


key_direction = {'ArrowUp': (0, -1),
                 'ArrowDown': (0, 1),
                 'ArrowLeft': (-1, 0),
                 'ArrowRight': (1, 0),
                 }


@out.capture(clear_output=True)
def on_key_down(key, *flags):
    print(key)
    if key in key_direction:
        dx, dy = key_direction[key]
        game.move(dx, dy)


board_spec = (20, 20, 30, 30, 7, 5)
mcanvas = MultiCanvas(2, width=250, height=190, layout=layout)
bg, fg = mcanvas


H.draw_grid(fg, board_spec, color='blue')
for col, row in game.TARGETS:
    x, y = H.get_midpoint(col, row, board_spec)
    fg.fill_circle(x, y, 2)


# neue update-Funktion der Spielkomponente
def update(event, **kwargs):
    D.update(bg, board_spec, event, **kwargs)


game.update = update
mcanvas.on_key_down(on_key_down)

display(mcanvas, out)
```
