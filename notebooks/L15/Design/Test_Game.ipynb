### Spielkomponente testen
Die Spielkomponente definiert
- `NCOLS`: Anzahl Spalten des Gitters,
- `NROWS`: Anzahl Zeilen des Gitters,
- `BOXES`: Tuple mit den Positionen der Boxen,
- `TARGETS`: Tuple mit den Zielpositionen der Boxen,
- `MOVES`: Tuple mit den möglichen Zügen,
- `state`: Dict zum Speichern der aktuellen Spieler- und Boxpositionen.

Die Spielkomponenten definiert die Funktionen
- `update(event, **kwargs)`: gibt Modifikation des Spielzustandes aus,
- `add(pos, dx, dy)`: Hilfsfunktion zum updaten einer Position,
- `new_game()`: initialisisert den Startzustand,
- `move(dx, dy)`: Bewegt Spieler und Boxen.
