```python
def deal_cards(deck, nhands, n=-1):
    hands = tuple([] for _ in range(nhands))
    while deck and n != 0:
        n = n - 1
        for hand in hands:
            hand.append(deck.pop())
    return hands
```

```python
deck = new_deck(ranks=range(4, 4+8), suits=(1,))
hands = deal_cards(deck, nhands=2)
for i, hand in enumerate(hands):
    print(f'Player {i+1}: {cards2str(hand)}')
stack = []
history = []

steps = 100
while hands[0] and hands[1] and steps > 0:
    steps = steps - 1
    card_1 = hands[0].pop()
    card_2 = hands[1].pop()
    stack.extend((card_1, card_2))
    history.extend((card_1, card_2))

    if card_1[0] > card_2[0]:
        hands[0][:] = stack[::-1] + hands[0]
        print(f'player 1: captures stack {cards2str(stack)}')
        print(f'Player 1: {cards2str(hands[0])}, Player 2: {cards2str(hands[1])}')
        stack.clear()
    if card_1[0] < card_2[0]:
        hands[1][:] = stack[::-1] + hands[1]
        print(f'player 2: captures stack {cards2str(stack)}')
        print(f'Player 1: {cards2str(hands[0])}, Player 2: {cards2str(hands[1])}')
        stack.clear()

len(hands[0]), len(hands[1])
```

```python
def get_starting_hands(history):
    first_card = history[0]
    second_card = history[1]
    hands = ([first_card], [second_card])

    idx = 2
    n = len(history)
    while idx < n and first_card != history[idx]:
        for hand in hands:
            card = history[idx]
            hand.append(card)
            idx = idx + 1
    return hands
```

```python
starting_hands = get_starting_hands(history)
for i, h in enumerate(starting_hands):
    print(f'{i+1}: {cards2str(h)}')
```
