```python
from dict_tools import peek
```

```python
def make_count_dict(table, i, idxs):
    d = {}
    for row in table:
        key = row[i]
        value = tuple(row[i] for i in idxs)
        if key in d:
            d[key].append(value)
        else:
            d[key] = [value]

    d = {k: sorted(v) for k, v in d.items()}
    d = dict(sorted(d.items(), key=lambda x: len(x[1]), reverse=True))
    return d
```

```python
d_plz6 = make_count_dict(table3, 1, (0, 2))
peek(d_plz6, 1)
```

```python
d_bfs = make_count_dict(table3, 2, (0, 1))
peek(d_bfs, 1)
```
