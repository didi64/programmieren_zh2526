```python
def step(numbers):
    if numbers[-2] == 0:
        numbers[-2:] = [numbers[-1] + 1]
    elif numbers[-1] == 0:
        numbers[-2:] = [numbers[-2] - 1, 1]
    else:
        numbers[-2:] = [numbers[-2] - 1, numbers[-2], numbers[-1] - 1]
```

```python
def fight_hydra(numbers):
    steps = 0
    while len(numbers) > 1:
        step(numbers)
        steps = steps + 1
    return steps
```

```python
xs = [3, 2]
fight_hydra(xs), xs
```

```python
numbers = [3, 2]
for i in range(50):
    step(numbers)
    print(numbers)
```
