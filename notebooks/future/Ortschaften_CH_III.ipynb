```python
def make_index(s):
    index = [0]
    current = s[0]

    for i, c in enumerate(s):
        if c != current:
            index.append(i)
            current = c

    return index
```

```python
def is_disjoint(items):
    n = len(items)
    for i in range(n-1):
        if items[i] == items[i+1]:
            return False
    return True


def is_one(items):
    n = len(items)
    for i in range(n-1):
        if items[i] != items[i+1]:
            return False
    return True
```
