```python
def get_counts(sorted_items):
    first = sorted_items[0]
    unique_items = [first]
    counts = []

    previous_i = 0
    current_item = first

    for i, item in enumerate(sorted_items):
        if current_item != item:  # neuer Punkt
            unique_items.append(item)
            counts.append(i - previous_i)
            current_item = item
            previous_i = i

    counts.append(i+1 - previous_i)
    return unique_items, counts
```

```python
letters = 'aabbcccdefffx'
get_counts(letters)
```
