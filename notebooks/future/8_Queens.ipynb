### Queens
On a chessboard, a queen at (x1, y1) attacks a queen to its right (x2, y2) (where x2 > x1), if
- both queens are in the same row,
- the smallest rectangle that contains both queens is a square, thus  y2-y1 = |x2-x1|.

```python
def is_save(queens, y, n):
    '''queens: list[int] of length at least n
       returns False if a queen at (n, y)  attacks a queen at
       (0, queens[0]), ..., (n-1, queens[n-1])
    '''
    for i in range(n):
        if queens[i] == y or n-i == abs(y-queens[i]):
            return False
    return True
```

```python
N = 8
queens = [0] * N
stack = [(0, 0)]

i = 0
while i != N-1:
    i, y = stack.pop()
    queens[i] = y
    options = [y for y in reversed(range(N)) if is_save(queens, y, i+1)]
    stack.extend([(i+1, row) for row in options])

queens
```

```python
def rotate_90(queens, N=8):
    qs = [0] * N
    for i in range(N):
        qs[N-1-queens[i]] = i
    return tuple(qs)


def canonic(queens, N=8):
    q = tuple(queens)
    qs = [q, q[::-1]]
    for i in range(3):
        q = rotate_90(q, N)
        qs.extend((q, q[::-1]))
    return min(qs)
```

```python
def get_relative_position(pos1, pos2):
    dx = pos1[0] - pos2[0]
    dy = pos1[1] - pos2[1]

    r = None
    if pos1 == pos2:
        r = '='
    elif dx == 0:
        r = 'x'
    elif dy == 0:
        r = 'y'
    elif dx == dy:
        r = 'D'
    elif dx == -dy:
        r = 'd'

    return r


def is_save(queens, pos):
    for queen in queens:
        if get_relative_position(queen, pos) is not None:
            return False
    return True
```
