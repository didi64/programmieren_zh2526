```python
def is_save(queens, row, col):
    save = True
    k = 0
    while k < col and save:
        save = queens[k] != row and abs(k-col) != abs(queens[k]-row) and row < 8
        k = k + 1
    return save
```

```python
def is_save(queens, row, col):
    save = True
    for k in range(col):
        save = queens[k] != row and abs(k-col) != abs(queens[k]-row) and row < 8
        if not save:
            break
    return save
```
