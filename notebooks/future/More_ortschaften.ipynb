### Recap 07
Nachstehen lesen wir ein File mit
Angaben zu Orten in der Schweiz und Liechtenstein.
Wir zuerst erstellen wir eine sortierte Liste die jeden Ortschaftsnamen nur einmal enthält.



```python
filename = '../L07/PLZO_CSV_LV95.txt'
with open(filename, mode='r') as f:
    lines = [line.rstrip() for line in f]

lines[0]
```

```python
# Alle Zeilen ausser die Kopfzeile beim ; in Listen spalten und
# diese Listen als Reihen in die Tabelle einfuegen.
SEP = ';'
table = [line.split(SEP) for line in lines[1:]]
table[:2]
```

```python
# Fehlende Werte, d.h. leere Eintraege?
# gib nur die ersten 6 Spalten aus
[row[:6] for row in table if '' in row]
```

```python
# Eintraege mit Ortnamen Zimmerwald
[row[:6] for row in table if row[0] == 'Zimmerwald']
```

```python
# Eintraege mit Ortnamen Buchs
[row[:6] for row in table if 'Buchs' in row[0]]
```

```python
# Eintraege mit PLZ 1945
[row[:6] for row in table if row[1] == '1945']
```

```python
# Liste der Ortsnamen erstellen
ortsliste = [row[0] for row in table]
ortsliste[:3]
```

```python
# Positionen, wo 'Zürich' in der Liste auftritt
name = 'Zürich'
indices = [i for i, ort in enumerate(ortsliste) if ort == name]
indices[:5]
```

```python
# zugehoerige Tabellenzeilen ausgeben
for i in indices[:5]:
    row = table[i]
    print(row[:6])
```

```python
# sortierte Liste der Ortsnamen ohne Dubletten 
orte = sorted(list(set(ortsliste)))
orte[:3] + orte[-3:]
```

```python
# Orte der Laenge 2
[ort for ort in orte if len(ort) == 2]
```

```python
# Orte die mit Abkuerzung beginnen
q = [ort for ort in orte if '.' in ort.split()[0]]
q[:2] + q[-2:]
```

```python
# Orte die mit Abkuerzung im Innern
q = [ort for ort in orte if ort.find(' ') < ort.find('.')]
q[:2] + q[-2:]
```

```python
suffixes = (' Dorf', ' Hospiz', ' Station', ' Stadt', ' Bad', ' Platz', ' Kulm')
[ort for ort in orte if ort.endswith(suffixes)]
```

```python
def contains_digits(s):
    for c in '01234567689':
        if c in s:
            return True


[ort for ort in orte if contains_digits(ort)]
```

```python
def contains_palindrom(s):
    for p in s.split():
        p = p.lower()
        if len(p) > 1 and p == p[::-1]:
            return True


[ort for ort in orte if contains_palindrom(ort)]
```

```python
def contains_XX(s):
    '''gibt True zurueck, falls s
       zwei aufeinanderfolgende Grossbuchstaben enthaelt
    '''
    for x, y in zip(s, s[1:]):
        if x.isupper() and y.isupper():
            return True
```

```python
q = [ort for ort in orte if contains_XX(ort)]
q[:2] + q[-2:]
```

```python
def get_ktz(s):
    for x, y in zip(s, s[1:]):
        if x.isupper() and y.isupper():
            return x+y


# Kantonskuerzel
kts = set(get_ktz(ort) for ort in orte_XX)
q = sorted(kts)
q[:2] + q[-2:]
```

```python
orte_mit_plz = [(row[0], row[1]) for row in table]
q = orte_mit_plz = sorted(set(orte_mit_plz))
q[:2] + q[-2:]
```

```python
def find_XX(s):
    for i, (x, y) in enumerate(zip(s, s[1:])):
        if x.isupper() and y.isupper():
            return i
    return -1


orte_short_mit_plz = []
for ort, plz in orte_mit_plz:
    if contains_digits(ort):
        continue
    i = ort.find(' (')
    if i != -1:
        ort = ort[:i]
    i = find_XX(ort)
    if i != -1:
        ort = ort[:i-1]

    for sub in (' an', ' am', ' b.', ' im', ' ob'):
        i = ort.find(sub)
        if i != -1:
            ort = ort[:i]

    orte_short_mit_plz.append((ort, plz))

orte_short_mit_plz[:2] + orte_short_mit_plz[-2:]
```

```python
ort_plzs = {}
for ort, plz in orte_short_mit_plz:
    if ort in ort_plzs:
        ort_plzs[ort].append(int(plz))
    else:
        ort_plzs[ort] = [int(plz)]

ort_plzs = {k: sorted(v) for k, v in ort_plzs.items()}
ort_plzs = dict(sorted(ort_plzs.items(), key=lambda x: len(x[1]), reverse=True))
```
