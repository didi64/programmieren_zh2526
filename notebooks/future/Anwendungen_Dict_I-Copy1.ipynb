### Aufgaben die man mit Hilfe von Dictionaries löst
Als nächstes wollen wir wissen, an welchen Positionen die Zahlen in der ursprünglichen Liste auftreten.  
Wir erstellen einen Dictionary `index_dict`, der zu jeder Zahl angibt, an welchen Positionen diese auftritt.
```python
index_dict = {}
for i, number in enumerate(numbers):
    if number in index_dict:
        index_dict[number].append(i)
    else:
        index_dict[number] = [i]

index_dict
```

```
{5: [0, 19],
 1: [1, 4, 9, 14],
 3: [2, 16, 18],
 2: [3],
 4: [5, 8, 10, 13, 15, 17],
 7: [6, 11],
 6: [7, 12]}
```
**Aufgabe**: Teste obigen Code.

````python
def make_count_dict(items, cond=None):
    count_dict = {}

    for item in items:
        if cond and not cond(item):
            continue
        if item in count_dict:
            count_dict[item] += 1
        else:
            count_dict[item] = 1

    return count_dict
````

````python
make_count_dict('alles, gute', str.isalpha)
````

````python
def lf_distance(d_ref, d):
    diffs = [(d_ref[c] - d.get(c, 0)) ** 2 for c in d_ref]
    dist = (sum(diffs) / len(diffs)) ** 1/2
    return dist
````
