```python
def startswith_1(s, prefix):
    return s[:len(prefix)] == prefix


def endswith_1(s, suffix):
    if not suffix:
        return True
    return s[-len(suffix):] == suffix

def startswith(s, prefix, start=0, end=None):
    if end is None:
        end = len(s)
    s = s[start: end]
    if type(prefix) is str:
        return s[:len(prefix)] == prefix
    elif type(prefix) is tuple:
        for pre in prefix:
            if type(pre) is not str:
                raise TypeError(f'tuple for startswith must only contain str, not {type(pre).__name__}')
            match = s[:len(pre)] == pre
            if match:
                return True
        return False
    else:
        raise TypeError(f'startswith first arg must be str or a tuple of str, not {type(prefix).__name__}')


def endswith(s, suffix, start=0, end=None):
    if not suffix:
        return True

    if end is None:
        end = len(s)

    s = s[start: end]
    if type(suffix) is str:
        suffix = (suffix,)
    if type(suffix) is not tuple:
        raise TypeError(f'startswith first arg must be str or a tuple of str, not {type(suffix).__name__}')

    for suf in suffix:
        if type(suf) is not str:
            raise TypeError(f'tuple for startswith must only contain str, not {type(suf).__name__}')
        match = len(suf) == 0 or s[:len(suf)] == suf
        if match:
            return True
    return False


```

```python
def join(s, items):
    result = ''
    for item in items:
        result = result + item + s
    if not s:
        return result
    return result[:-len(s)]
```

```python
def lstrip(s, chars=None):
    if chars is None:
        chars = ' '

    n = len(s)
    i = 0
    while i < n and s[i] in chars:
        i = i + 1
    return s[i:]


def rstrip(s, chars=None):
    if chars is None:
        chars = ' '

    i = len(s) - 1
    while i >= 0 and s[i] in chars:
        i = i - 1
    return s[:i+1]


def strip(s, chars=None):
    s = lstrip(s, chars)
    s = rstrip(s, chars)
    return s
```

```python
s = '  foobar  '
strip(s, ' fo') == s.strip(' fo')
```

```python
def split(s, sep=','):
    words = []
    n = len(s)
    i = find(s, sep)
    while 0 <= i < n:
        word = s[:i]
        words.append(strip(word))
        s = s[i+1:]
        i = find(s, sep)
    words.append(strip(s))
    return words
```

```python
def count(s, sub, start=0, end=None):
    total = 0
    m = len(sub)
    i = find(s, sub, start, end)
    while i != -1:
        total = total + 1
        i = find(s, sub, i+m+1)
    return total
```
