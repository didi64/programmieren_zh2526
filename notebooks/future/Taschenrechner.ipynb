Berechne den Wert des Terms  
$$\frac{(16-1)\cdot (-3)}{2\cdot 3 - 1}$$.

```python
def calculator():
    stack = []

    while True:
        x = input('positive Zahl oder Operation (~,-,+,*,/):')
        if not x:
            break
        elif x.isnumeric():
            stack.append(int(x))
        elif x == '~':
            y = stack.pop()
            stack.append(-y)
        elif x == '-':
            z = stack.pop()
            y = stack.pop()
            stack.append(y - z)
        elif x == '+':
            z = stack.pop()
            y = stack.pop()
            stack.append(y + z)
        elif x == '*':
            z = stack.pop()
            y = stack.pop()
            stack.append(y * z)
        elif x == '/':
            z = stack.pop()
            y = stack.pop()
            stack.append(y / z)
        else:
            pass

        print(stack)


    if len(stack) > 1 or len(stack) == 0:
        print('unvollendete Berechnung!')
    else:
        print(f'Das Resultat ist {stack.pop()}')
```

```python
calculator()
```
