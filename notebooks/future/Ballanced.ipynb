```python
def is_opening(token):
    return token[0] == '<' and token[1] != '/'


def is_closing(token):
    return token[:2] == '</'


def get_name(token):
    SPACE = ' '
    end = token.find(SPACE)
    if is_opening(token):
        return token[1:end]
    else:
        return token[2:end]


def is_balanced(tokens):
    stack = []
    n = len(tokens)
    for i in range(n):
        if is_opening(tokens[i]):
            stack.append(i)
        elif is_closing(tokens[i]):
            if not stack:
                print(f'{tokens[i]} is not closing any tag')
                return False
            j = stack[-1]
            name_i = get_name(tokens[i])
            name_j = get_name(tokens[j])
            if name_i == name_j:
                stack.pop()
            else:
                print(f'{tokens[i]} does not close token {j} with name {name_j}')
                return False
    return True
```
