```python
# Laenge der Zahl n bestimmen
n = 1
length = 0
while n > 0:
    n = n // 10
    length = length + 1

length
```

```python
def numberlength(n):
    '''nbr: int
       return: int
       returns the number of digits of a number or 0 if the number is 0
    '''
    length = 0
    while n > 0:
        n = n // 10
        length = length + 1
    return length
```

```python
# teste, ob der String text cas Zeichen character enthaelt
text = 'agagdfsgdsf'
character = 'x'

n = len(text)
idx = 0
contains_character = False

while idx < n and not contains_character:
    if text[idx] == character:
        contains_character = True
    idx = idx + 1

contains_character
```

```python
def contains_char(s, char):
    '''s: str
       char: str of length 1
       return: bool
       returns True if the s contains the symbol char
    '''
    n = len(s)
    idx = 0
    contains_char = False

    while idx < n and not contains_char:
        if s[idx] == char:
            contains_char = True
        idx = idx + 1  # add first!
    return contains_char
```

```python
contains_char('abcde', 'c')
```

```python
# Teste, ob die Zahl n die Ziffer digit enthaelt
n = 12345689
n0 = n
digit = 7
found = False
while n > 0 and not found:
    z = n % 10
    if z == digit:
        found = True
    n = n // 10

if found:
    print(digit, 'occurs in', n0)
if not found:
    print(digit, 'does not occur in', n0)
```

```python
def contains_digit(nbr, digit):
    '''nbr: int
       digit: int (a single digit)
       return: bool
       returns True if nbr contains digit
    '''
    found = False
    while nbr > 0 and not found:
        z = nbr % 10
        if z == digit:
            found = True
        nbr = nbr // 10
    return found
```

```python
contains_digit(12345, 6)
```

```python
def count(s, char):
    '''s: str
       char: str of length 1
       return: int
       returns the number of times char occurs in s
    '''
    n = len(s)
    idx = 0
    count = 0
    while idx < n:
        if s[idx] == char:
            count = count + 1
        idx = idx + 1  # add first!
    return count
```

```python
def is_prime(n):
    '''n: int
       return: bool
       returns True if n is prime
    '''
    i = 2
    is_prime = True
    while i*i <= n and is_prime:
        if n % i == 0:
            is_prime = False
        i = i + 1
    return is_prime
```

```python
def split(s, idx):
    '''s: str
       idx: int with 0<= idx <len(s)
       return: tuple[str, str]
       returns a tuple (head, tail), there head is consists of the first idx characters of s
               and tail of the last len(s)-idx charachters of s
       '''
    head = ''
    tail = ''
    n = len(s)
    i = 0
    while i < idx:
        head = head + s[i]
        i = i + 1
    while i < n:
        tail = tail + s[i]
        i = i + 1
    return head, tail
```

```python
# Fuege die Zeichen zweier sortierter Strings so zusammen, das ein
# sortierter String resultiert
word1 = 'adef'
word2 = 'bc'
merged_word = ''

m = len(word1)
n = len(word2)

i = 0
j = 0

while i+j < m+n:
    from_word1 = (j >= n) or (i < m and word1[i] <= word2[j])
    if from_word1:
        next_char = word1[i]
        i = i + 1
    if not from_word1:
        next_char = word2[j]
        j = j + 1

    merged_word = merged_word + next_char

merged_word
```

```python
def merge_sorted_words(word1, word2):
    '''word1, word2: str (sorted strings)
       retrun: str
       returns a sorted string consisting of the characters of word1 and word2
    '''
    merged_word = ''
    m = len(word1)
    n = len(word2)
    i = 0
    j = 0
    while i+j < m+n:
        from_word1 = (j >= n) or (i < m and word1[i] <= word2[j])
        if from_word1:
            next_char = word1[i]
            i = i + 1
        if not from_word1:
            next_char = word2[j]
            j = j + 1
        merged_word = merged_word + next_char
    return merged_word
```

```python
merge_sorted_words('ac', 'bbdeg')
```

```python
# sortiere einen String word
def sort_word(word):
    '''word: str
       return: str
       returns the characters of word in ascending order
    '''
    sorted_word = ''
    n = len(word)
    i = 0
    while i < n:
        sorted_word = merge_sorted_words(sorted_word, word[i])
        i = i + 1
    return sorted_word
```
