**if-Anweisungen**

```python
# if-Anweisungen
# Eingeruecker Code nach if <Ausdruck> wird nur ausgefuehrt, falls <Ausdruck> wahr ist
x = 6

if x == 5:
    print(x, 'ist gleich 5')
if x != 5:
    print(x, 'ist ungleich 5')
    print('Abweichung:', 5 - x)
```

```python
# teste, ob Anwort ja oder nein
answer = 'jein'
is_ja_or_nein = answer == 'ja' or answer == 'nein'

if is_ja_or_nein:
    print('Die Antwort ist', answer)
if not is_ja_or_nein:
    print(answer, 'ist keine gueltige Antwort')
```

**While-Schleife**

```python
# while Schleife
i = 0

while i < 3:
    print('i ist gleich', i)
    i = i + 1  # nicht vergessen, sonst Endlosschleife!
```

```python
# mit Whileschleife ueber Zeichen eines Strings iterieren
word = 'hallo'
i = 0

while i < len(word):
    print(word[i])
    i = i + 1
```

```python
# while Schleife: Besuche Zeichen, bis Endmarker . kommt
dot_terminated_word = 'hallo.'
i = 0

while dot_terminated_word[i] != '.':
    print(dot_terminated_word[i])
    i = i + 1
```

**Aus String neuen String konstruieren**

```python
# String kann man nicht modifizieren!
word = 'hallo'
n = len(word)
inner_word = ''
i = 1

while i < n-1:
    inner_word = inner_word + word[i]
    i = i + 1
inner_word
```
