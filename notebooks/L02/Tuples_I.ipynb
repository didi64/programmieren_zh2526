### Tuples I
Wir werfen einen ersten Blick auf Tuples.  
&Auml;hnlich wie Strings, die eine Folge von Zeichen speichern,
kann ein Tuple eine Folge von Werten speichern. Die Werte k&ouml;nnen dabei 
einen unterschiedlichen Typ haben.

Wie Strings, kann man zwei Tuple aneinanderh&auml;ngen und mit einem Integer vervielfachen.  
Ebenso liefert `len(<tuple>)` die L&auml;nge (Anzahl Elemente) des Tuples und
`<tuple>[i]` das i-te Element.
- `names = ('Alice', 'Bob')` Tuple mit 2 Strings,
- `prime_numbers = (2, 3, 5, 7, 11, 13)` Tuple mit Zahlen,
- `person = ('Max', 'Muster', 1995)` Tuple mit Vorname, Name und Jahrgang.  
  `person[0]` liefert den Vornamen, `person[2]` den Jahrgang.

Spezialf&auml;lle:
- `empty_tuple = ()` leeres Tuple
- `einelementiges_tuple = ('Bob',)` Das Komma ist **obligatorsich**

**Bemerkung**:
Die &auml;usseren Klammern sind beim Erzeugen eines Tuples fakultativ.  
Das ist manchmal bequem, kann aber zu Fehlern f&uuml;hren und ist nur selten guter Stil.



```python
x = 1
y = 2
x, y  # Erzeut das Tuple (x, y), da die Klammern fakultativ sind!
```

```python
x = 1,  # ein Tuple! Besser x = (1,)
x + 2  # Tuple und Integer kann man nicht addieren
```

```python
number = (1)  # kein Tuple, sondern ein Integer
number
```

```python
name = ('Alice')  # kein Tuple, sondern ein String
name
```

```python
number = (1,)  # ein 1-elementiges Tuple
number
```

```python
len(number)
```

```python
number[0]  # erstes (und einziges) Element des Tuples
```

```python
ones = (1,) * 10
ones
```

```python
len(ones)
```

```python
ones + ones
```

```python
# Element in einem Tuple suchen
names = ('Alice', 'Bob', 'Carl')
name = 'Bob'

n = len(names)
i = 0
found = False

while i < n and not found:
    if name == names[i]:
        found = True
    if name != names[i]:
        i = i + 1

print(name, 'gefunden an Position', i)
```

```python
# Solange nach Woertern fragen, bis das leere Wort eingegeben wird.
# Woerter in einem Tuple words speichern
words = ()
word = input('Gib ein Wort ein:')

while word:  # Schleife wird ausgefuehrt solange Wort nicht leer (ungleich '')
    words = words + (word, )
    word = word = input('Gib ein Wort ein:')

print('Du hast', len(words), 'WÃ¶rter eingeben')
print(words)
```
