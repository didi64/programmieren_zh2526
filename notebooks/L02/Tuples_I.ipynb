### Tuples I
Wir werfen einen ersten Blick auf Tuples.  
&Auml;hnlich wie Strings, die eine Folge von Zeichen speichern,
kann ein Tuple eine Folge von Werten speichern. Die Werte k&ouml;nnen dabei 
einen unterschiedlichen Typ haben.

Wie Strings, kann man zwei Tuple mit `+` aneinanderh&auml;ngen und mit einem Integer mit `*` vervielfachen.  
Ebenso liefert `len(<tuple>)` die L&auml;nge (Anzahl Elemente) des Tuples und
`<tuple>[i]` das i-te Element.  
Zwei Tuple sind gleich bez. `==` falls ihre entsprechenden Elemente bez. `==` gleich sind.  
Und wie bei Strings entscheidet das 1. Element, bei dem sich die Tupel unterscheiden, welches das kleinere Tuple ist.  
- `names = ('Alice', 'Bob')` Tuple mit 2 Strings,
- `prime_numbers = (2, 3, 5, 7, 11, 13)` Tuple mit Zahlen,
- `person = ('Max', 'Muster', 1995)` Tuple mit Vorname, Name und Jahrgang.  
  `person[0]` liefert den Vornamen, `person[2]` den Jahrgang.

Spezialf&auml;lle:
- `empty_tuple = ()` leeres Tuple
- `einelementiges_tuple = ('Bob',)` Das Komma ist **obligatorsich**

**Bemerkungen**:  
- Tuples kann man **nicht modifizieren**.  
- Die &auml;usseren Klammern sind beim Erzeugen eines Tuples fakultativ.  
  Das ist manchmal bequem, kann aber zu Fehlern f&uuml;hren und ist nur selten guter Stil.

```python
x = 1
y = 2
x, y  # Erzeugt das Tuple (x, y), da die Klammern fakultativ sind!
```

```python
x = 1,  # ein Tuple wegen dem Komma! Besser x = (1,)
x + 2  # Tuple und Integer kann man nicht addieren
```

```python
(2, 2, 3) < (3,)  # erstes Element, wo sich die Tupel unterscheiden gibt Ausschlag, 2 < 3
```

```python
number = (1)  # kein Tuple, sondern ein Integer
number
```

```python
name = ('Alice')  # kein Tuple, sondern ein String
name
```

```python
number = (1,)  # ein 1-elementiges Tuple
number
```

```python
len(number)
```

```python
number[0]  # erstes (und einziges) Element des Tuples
```

```python
ones = (1,) * 10
ones
```

```python
len(ones)
```

```python
ones + ones
```

```python
# Ein Tuple kann auch andere Tuple als Elemente enthalten
dreieck = ((-1, -2), (2, -2), (2, 2))
i = 0

while i < 3:
    pt1 = dreieck[i]
    pt2 = dreieck[(i+1) % 3]  # 3 % 3 ist 0
    distanz = ((pt1[0] - pt2[0])**2 + (pt1[1] - pt2[1])**2)**0.5  # Pythagoras
    i = i + 1
    print('Länge der Seite', pt1, pt2, 'ist', distanz)
```

```python
# Solange nach Wörtern fragen, bis das leere Wort eingegeben wird.
# Wörter in einem Tuple words speichern
words = ()
word = input('Gib ein Wort ein:')

while word:  # Schleife wird ausgefuehrt solange Wort nicht leer (ungleich '')
    words = words + (word, )
    word = input('Gib ein Wort ein:')

print('Du hast', len(words), 'Wörter eingeben')
print(words)
```

```python
# Element in einem Tuple names suchen
# Wird das Element gefunden, i der Index des Elements sein.
names = ('Alice', 'Bob', 'Carl')
name = 'Bob'

n = len(names)
i = -1
found = False

while i < n and not found:
    i = i + 1
    if name == names[i]:
        found = True


if found:
    print(name, 'gefunden an Position', i)
    print(names[i])
if not found:
    print(name, 'nicht gefunden!')
```
