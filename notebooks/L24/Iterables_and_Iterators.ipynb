### Iteratoren und Iterables

Ein **Iterator** ist ein Objekt `obj`, welches eine `__next__` und `__iter__` Methode bereitstellt.
Die `__iter__` Methode muss einen Iterator liefern und gibt typischerweise sich selbst zurück.
`next(obj)` und `iter(obj)` sind *syntactic sugar* für `obj.__next__()` und `obj.__iter__()`.

Insbsondere ist jedes Generator-Objekt ein Iterator.


Ein **Iterable** ist Objekt `obj`, welches eine Methode `__iter__` bereitstellt, die einen
Iterator liefert.  

So sind z.B. Listen, Tuple, Mengen und Dictionaries Iterables,
aber eben auch  jedes Generator-Objekt.

Über jedes Iterable kann mit einem For-Loop iteriert werden, denn
```python
for x in iterable_obj:
    print(x)
```

übersetzt zu
```python
it = iter(iterable_obj)

while True:
    try: 
        x = next(it)
    except StopIteration:
        break
    print(x)
```  

**Nützliche Anwendung**:

Manchman möchte man die ersten und letzten paar Elemente eines grossen Dicts `d` ansehen.
```
list(d.items())[:2]
list(d.items())[-2:]
```
tun dies, jedoch wird dabei eine grosse Liste erstellt.  
Folgendes liefert Iteratoren. Auf diese Iteratoren kann `next` anwenden, um 
erste bez. letzte Key-Value Paare des Dicts zu bekommen.
```python
iter(d.items())
iter(reversed(d.items()))
```


````python
gen = (x for x in range(3))
iter(gen) is gen  # gen ist bereits ein Iterator, iter liefert sich selbst
````

````python
items = [1, 2, 3]
next(items)  # kein Iterator
````

````python
it = iter(items)  # it ist nun ein Iterator!
````

````python
next(it)
````

````python
d = dict(zip((chr(65+i) for i in range(26)), range(26)))
firsts = iter(d.items())
lasts = iter(reversed(d.items()))
````

````python
next(firsts), next(lasts)
````

````python
class BadIter:
    def __next__(self):
        print('Ich bin ein Iterator?')
````

````python
baditer = BadIter()
next(baditer)
````

````python
for x in baditer:
    print(x)
````

````python
class GoodIter:
    def __next__(self):
        print('Ich bin ein Iterator!')

    def __iter__(self):
        return ('haha' for _ in range(3))
````

````python
for x in GoodIter():
    print(x)
````

````python
class Iterable:
    def __iter__(self):
        print('Ich bin ein iterable')
        return ('ich liefere einen Generator (und das ist acu ein Iterator)' for _ in range(3))
````

````python
iterable = Iterable()
it = iter(iterable)
it
````

````python
for x in iterable:
    print(x)
````
