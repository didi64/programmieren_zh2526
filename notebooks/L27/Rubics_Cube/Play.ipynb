### Der 2x2x2 Rubiks-Cube
Der 2x2x2 Würfel besteht aus 8 Würfelchen, 0 bis 7. Von nun an schreiben wir W. für Würfelchen.
Von jedem W. sind 3 Seiten sichtbar, die alle unterschiedliche Farben tragen.


Den Würfel kann man auf jeder seiner 6 Seiten stellen, dann hat man noch 4 Möglichkeiten eine
Vorderseite zu wählen. Wir wollen diese 24 Zustände nicht unterscheiden.
Deshalb positioniern wir den Würfel immer so, dass **das W. mit den Seiten orange, grün und gelb hinten links liegt, mit der gelben Seite nach unten**.  




Hier sehen wir den Würfel von oben im Ausgangszustand.
Links ist die obere Seite. W. 0 ist oben hinten links.
Rechts ist die untere Seite. Das W. 7 ist unten hinten links und **ist dort fixiert**.  

Jedes W. bekommt eine feste Nummer.
Die Position 0 ist dort wo W. 0 im Ausganszustand ist (analog für die anderen W.s).

Wir weisen jedem W. eine **Orientierung**  in (0, 1, 2), bez. (=, +, -) zu:   
Ist die obere bez. unter Fabe eines W. weiss oder gelb, hat es Orientierung 0 (=).
Rotiert man ein W. mit Orientierung 0 um 1x$120^{\circ}$ (2x$120^{\circ}$) im Uhrzeigersinn um die Achse
vom Würfelzentrum zur Ecke des W.s, dann hat das W. Orientierung $1$ ($+$) bez. $2$ ($-$).

Damit W. 7 an  seinem Ort bleibt, drehen wir nur die obere, vordere und rechte Seite.
Diese Drehungen notieren wir so:
- U (obere Seite im Uhrzeigersinn), u (im Gegenuhrzeigersinn), U2 (Doppeldrehung)
- F (vordere Seite im Uhrzeigersinn), f (im Gegenuhrzeigersinn), F2 (Doppeldrehung)
- R (rechte Seite im Uhrzeigersinn), r (im Gegenuhrzeigersinn), R2 (Doppeldrehung)

**Beachte**: U ändert keine Orientierungen. Doppeldrehungen ändern keine Orientierungen.
F und R ändern die Summe der Orientierungen modulo 3 nicht.

Lassen wir W. 7 fest, haben wir für W. 0  6 Plätze, für W. 1 noch 5, ...
Insgesamt gibt es $7! = 5040$ Anordnungen. Weiter können wir 
für 6 W.s noch eine Orientierung wählen. Total gibt es also $5040 \cdot 3^6 = 3'674'160$ Zustände.



Jeder Zustand s des Würfels fassen wir auch als **Operation** aus, nämlich als die Veränderungen, die wir ausführen müssen, um
diesen **Zustand von Ausgangszustand her zu erreichen**.
Diese Veränderungen können wir aber auch auf einen anderen Zustand anwenden.

Zustand/Operation  $op = (0, 2, 5, 3, 4, 6, 1, 7) + (0, 1, 2, 0, 0, 1, 2, 0)$ heisst:
>2 -> 1+, 5 -> 2-, 6 -> 5+, 1 -> 6-  (allg. op[i]-> i mit O.änderung op[i+8])

W. auf Pos. 2 geht auf Pos. 1 und wird dann noch um $1*120^{\circ}$ gedreht, ...
W. die nicht bewegt werden sind nicht erwähnt.


Zu jedem Zustand $s$ gibt es auch einen inversen Zustand $s^{-1}$:
Jede Operation lässt sich rückgängig machen. In obigen Fall durch,

> 1 -> 2-, 2-> 5+, 5 -> 6-, 6 -> 1+ 

was dem inversern Zusand $s^{-1} = (0, 6, 1, 3, 4, 2, 5, 7) + (0, 1, 2, 0, 0, 1, 2, 0)$ entspricht.

```python
# op auf state anwenden
new_state[i] = state[op[i]]  # W. von Pos. op[i] geht nach Pos i
new_state[i+8] = (state[op[i]+8] + op[i+8]) % 3  # neue O.von W. an Pos. op[i]: alte O. + O.änderung

# inv_state
state_inv[state[i]] = i
state_inv[state[i]+8] = -state[8] % 3
```

````python
import babycube as cube


def get_transition(s, t):
    x = cube.apply_op(t, cube.inv_op(s))
    return cube.OP_KEY.get(x)
````

````python
word = 'RUF2U2rUrUR2fU2FurfU2'
s = cube.apply_word(word)
t = cube.apply_op(cube.KEY_OP['R'], s)

get_transition(s, t)
````
