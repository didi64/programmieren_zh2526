### Wieso liefert BFS kürzeste Pfade
Für jeden Knoten $k$ in der Liste `nodes_to_visit`
liefert `get_path_home(k, goback)[::-1]` einen Pfad vom Ausgangsknoten `start` zu  `k`.
Mit `|k|` bezeichnen wir die Länge diese Pfades.

1. Die Liste `[|k| for k in nodes_to_viste]` hat die Form `[n, ..., n]` oder `[n+1, ..., n+1, n, ..., n]`.

   Dies gilt zu Beginn, wenn die Liste erst 1 Element enthält.
   Auch bleibt diese Eigenschaft erhalten:
   Wird ein Knoten `k` mit zugehörigem Pfad `(start, ..., k)` von rechts entfernt,
   werden neue Nachbarn `k'` von `k`  links hinzugefügt.
   Der Pfad zu diesen Knoten ist `(start, ..., k, k')`.
   
   
2. Für jeden Knoten `k` in der Liste `nodes_to_visit`,
   liefert `get_path_home(k, goback)[::-1]` einen kürzesten Pfad von `start` zu `k`.

   Dies gilt zu Beginn. Auch bleibt diese Eigenschaft erhalten:
   Sei `k` der letzte Knoten in `nodes_to_visit`.
   Nach Voraussetzung ist `(start, ..., k)` ein kürzester Pfad der Länge |`k`| = n.
   Wird `k` entfernt, gehört zum neu von links hinzugefügten Knoten `k'` ein Pfad
    `(start, ..., k, k')` mit  |`k'`| = n+1. 

   Wir zeigen, dass es keinen Pfad der Länge m &le; n von `start` nach `k'` gibt.

   Sei `(start=k_0, ..., k_i, k_i+i, ..., k')` so ein Pfad. Weiter sei
   `k_i+i` der erste Knoten, der nie aus `nodes_to_visit` entfernt wurde
   (`start` wurde entfernt, `k'` noch nicht, also gibt es so einen Knoten).

    Auch `(k_0, ..., k_i, k_i+i)` ist ein kürzester Pfad. Da `k_i` in `nodes_to_visit` ist,
    gilt nach Voraussetzung, dass `get_path_home(k_i, goback)[::-1]` ein kürzester Pfad von `start`
    zu `k_i` ist.
    Damit ist auch `get_path_home(k_i+1, goback)[::-1]` ein kürzester Pfad von `start` zu `k_i+1`.

    1. `k_i+1` = `k'`. Dann wäre `|k_i|` = m-1 und `|k'|` = m, im Widerspruch zu oben, wo `|k'|` = n > m$!
    2. `k_i+1` &ne; `k`.  
        Dann wäre `|k'_i+1|` < m.
        Da `k_i` entfernt wurde und `k_i+i` ein Nachbar von `k_i` ist, ist `k_i+1` in `nodes_to_visit`.
        Aber `k_i+1` ist rechts von `k`, da `|k|`= n &gt; m > `|k_i+1|`.
        Das kann nicht sein, da `k` vorher entfernt wurde. Somit kann es diesen kürzeren Pfad nicht geben!

### Minimales Beispiel, wo Smart-Search den kürzesten Pfad nicht findet

```python
import search_strategies as S


graph = {0: [1, 2],
         1: [4],
         2: [3],
         3: [4],
         4: [5],
         5: [6],
         }


#      1
#   /    \
# 0       4 - 5 - 6
#  \     /
#   2 - 3


def get_neighbors(node):
    return graph.get(node, ())
```

```python
start = 0
goal = 6
node, go_back, dist_dict = S.search_bf(start, get_neighbors,  goal)
path = S.get_path_home(goal, go_back)
path[::-1]
```

```python
h = lambda n: 3 if n == 1 else 0
node, go_back = S.search_smart(start, get_neighbors, h, goal)
path = S.get_path_home(6, go_back)
path[::-1]
```

```python
path = S.get_path_home(6, go_back)
path[::-1]
```
