DFS: Start ist bei (0, 0), Ziel bei (9, 9). Das Suchteam irrt umher bis es auf das Ziel stösst.  
Das Suchteam verirrt sich zuerst in die linke untere Ecke (0, 9) und steckt fest.
Immer wenn dies geschieht, wird versucht, die Suche vom letzten Knoten in `nodes_to_visit` fortzusetzen.
Bei (3, 8), und dann von (2, 3) gelingt das.


BFS: Alle vom Start ausgehenden Pfade werden gleich lang gehalten.
Ein kürzester Pfad ins Ziel wird gefunden.

### BFS vs bidirektionale Suche
<div>
    <figure style='float:left'>
        <figcaption>BFS mit Knightmoves</figcaption>
       <img src='/files/images/search_bf.png' width='200px'>
    </figure>
   <figure style='float:left'>
        <figcaption>Bidirektionale Suche mit Knightmoves</figcaption>
       <img src='/files/images/bidirectional.png' width='200px'>
    </figure>
    </div>     
<div style="clear: both;"></div>  

- **BFS**: Alle Knoten mit Abstand kleiner gleich 2 werden besucht.
- **Bidirektionale BFS**: Nur Knoten mit Abstand 1 werden besucht.  
Schwarze Knoten zeigen die Suchfront des 1. Suchteams, graue die des 2.  
Hellblaue Knoten gehört zu beiden Fronten.  
Beim blau umkreisten Knoten treffen sich die Teams.

**Smart-Search**: Bei der geg. Priority werden die hellbrauenen Felder

    <figure style='float:left'>

### Smart-Search Failures:
Smart-Search besucht alle relevanten Knoten, findet bei irreführender Heuristik aber keinen
kürzesten Pfad.  

h ist nur  bei (2, 0) ungleich 0.  
Priority(pos) = Pfadlänge bis pos + h(pos)

- Links: Priority((2, 0)) = 7. Nachdem die Felder auf der Nebendiagonale mit Priority 6 besucht wurden,
  ist dann (2, 0) an der Reihe. Von dort geht es direkt wieter bis (5, 2).
- Rechts: Priority((2, 0)) = 8. Hier wird (2, 0) einen Schritt später besucht.
  Zu diesem Zeitpunkt ist (5, 2) bereits in `nodes_to_visit` und von (2, 0) führt kein Weg mehr ins Ziel.
