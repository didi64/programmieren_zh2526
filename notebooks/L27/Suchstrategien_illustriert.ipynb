       <img src='/files/images/greedy_vs_smart_g2.png' width='200px'>
    </figure>
    </div>     
<div style="clear: both;"></div>

Links: Nach dem 1. Schritt sieht `nodes_to_visit` so aus:  
`[((10, 0), (2, 1)), ((10, 1), (1, 2)), ((11, 2), (0, 1)), ((11, 3), (1, 0))]`  
Der "richtige" Knoten hat die kleinste Priority, und von da an leitet
die Heuristik den Pfad direkt ins Ziel.

Rechts: Nach dem 1. Schritt sieht `nodes_to_visit` so aus:  
`[((10, 0), (1, 2)), ((10, 1), (2, 1)), ((11, 2), (0, 1)), ((11, 3), (1, 0))]`  
Der "falsche" Knoten hat die kleinste Priority.
Das wird erst an Position (1, 6) bemerkt, aber da gibt es schon kein Zurück mehr:
Alle neu hinzugefügten Knoten haben kleinere Priority als `10`.
***
<div>
    <figure style='float:left'>
        <figcaption>Smart-Search: Priority(pos)=Länge Pfad bis pos + Manhattendistanz zum Ziel</figcaption>
       <img src='/files/images/greedy_vs_smart_s1.png' width='200px'>
    </figure>
</div>     
<div style="clear: both;"></div>  

Alle Knoten mit Abstand 1 von Rand haben
Priority 10, da sie aus Sicht der Heuristik auf  einem  Pfad  der Länge 10 liegen.  
Nach dem 1. Schritt sieht `nodes_to_visit` so ev. aus:  
`[((10, 0), (1, 2)), ((10, 1), (2, 1)), ((11, 2), (0, 1)), ((11, 3), (1, 0))]`  

Zwar wir hier der "falsche Knoten" entfernt, aber
die Priority aller neu hinzugefügten Knoten ist grösser als (10, 1).
Die Suche wird als nächstes von (2, 1) aus fortgesetzt.
