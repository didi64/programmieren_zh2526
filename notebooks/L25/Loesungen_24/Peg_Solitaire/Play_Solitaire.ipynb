### Peg-Solitaire ([wiki/Solitär](https://de.wikipedia.org/wiki/Solit%C3%A4r_(Brettspiel)))

Wir wollen unsere Grid-Controller nutzen, um ein einfaches
Solitaire zu implementieren.
Damit nachstehender Code läuft, sind noch die
Spiel- und Viewkomponenten zu schreiben. Dabei orientieren wir uns
an den entsprechenden Komponenten des Spiels Sokoban in `L23/Sokoban_L`.

Um die Spielsteine mit der Maus bewegen zu können, registrieren wir eine Funktion `touch`
für das Event `on_mouse_down`.

Das Klicken auf einen Spielstein soll diesen auswählen oder die Auswahl rückgängig machen.
Klicken wir auf eine zweite Position, soll versucht werden, mit dem ausgewählten Stein dorthin zu springen.
Um den ausgewählten Stein zu speichern, nutzen wir das Default-Argument
`move=[None, None]` der Funktion `touch`.
Diese Liste wird mit der Funktionsdefinition kreiert und lässt sich innerhalb der Funktion
modifizieren.

Später wollen wir dann noch die in `solution.py` gespeicherte Lösung vorspielen können.

```python
import importlib
import grid_controller as C
import game as game
import view
from solution import get_move_iter
importlib.reload(game)
importlib.reload(view)


def touch(pos, move=[None, None]):
    if move[0] is None:
        move[0] = pos
    elif move[0] == pos:
        move[0] = None
    elif move[0]:
        move[1] = pos
        print(f'move({move[0]}, move[1])')
        game.move(*move)
        move[:] = None, None

    print(f'move: {move}')


def preview_move(jpeg, hole):
    x1, y1 = view.G.cr2xy(*jpeg, view.grid_spec, center=True)
    x2, y2 = view.G.cr2xy(*hole, view.grid_spec, center=True)
    view.canvas.stroke_style = 'blue'
    view.canvas.stroke_line(x1, y1, x2, y2)
    view.canvas.stroke_style = 'black'


def show_and_move(move=[None, None]):
    global moves
    if move[0] is None:
        try:
            move[:] = next(moves)
        except StopIteration:
            moves = get_move_iter()
            game.new_game()
            move[:] = next(moves)

        preview_move(*move)
    else:
        game.move(*move)
        move[:] = None, None


callbacks = {'n': game.new_game,
             # 'u': game.undo,
             'ArrowUp': show_and_move,
             'mouse_down': touch,
             }

moves = get_move_iter()

view.init(game)
C.init(view.canvas, callbacks, view.grid_spec)
C.on_key_down('n')
C.show(debug=True)
```
