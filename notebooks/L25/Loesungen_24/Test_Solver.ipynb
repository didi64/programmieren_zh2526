### Teste den Sudoku Solver

```python
fn = 'hardest_puzzles.txt'
lines = []
paragraphs = []
paragraph = []

with open(fn, mode='r') as f:
    for line in f:
        lines.append(line.rstrip())

for line in lines:
    if line:
        paragraph.append(line)
    else:
        paragraphs.append(paragraph)
        paragraph = []

header = paragraphs[0]
problems = paragraphs[1:]
print('\n'.join(problems[-1]))
```

```python
import sudoku_solver as S


def solve_problem(problem):
    comment = '\n'.join(problem[:-1])
    s = problem[-1]
    print(comment)
    S.print_sudoku_str(s)
    print()
    if (solution := S.solve_sudoku(s, good_neighbors=True)):
        S.print_sudoku_str(solution)
```

```python
for problem in problems:
    solve_problem(problem)
```
