### Priority-Queue/Min-Heap und Greedy-Search
Bei einer Greedy-Search speichert man in der Liste `nodes_to_visit` jeweils Paare der Form
`(priority, node)`. Die Priorität `priority` ist dabei eine Zahl.
Die Idee ist, das dem aussichtsreichsten Knoten die kleinste Zahl zugewiesen ist.
Statt das erste oder letzte Element entfernen wir jeweils das Tuple mit der kleinsten Priorität.

Um Effizient das kleinste Element entfernen zu können wird ein Priority-Queue verwendet.


Ein Priority-Queue oder Min-Heap ist eine Datenstruktur mit einer `pop`-Methode,
die effizient immer das kleinste Element entfernt und zurück gibt.


Min-Heap als Baum:   
Jeder Knoten (bis zu einer vergegebenen Tiefel) hat genau 2 Kinder, mit Ausnahme des rechtesten Blatts (vollständiger binärer Baum).

Min-Heap Eigenschaft:  
**Der Wert eines Elternknoten ist immer kleiner oder gleich als die Werte der Kinderknoten**.

Wir benutzen Pythons Modul `heapq`, siehe [python.org/3/library/heapq.html](https://docs.python.org/3/library/heapq.html)


Der Einfachheit halber speichert man  einen vollständigen binären Baum als flache Liste.
Der Wurzelknoten hat Index 0, dann folgen die beiden Kinder, dann alle Grosskinder, u.s.w.

**Indexrechnerei**:  
Ist i der Index des Elternkotens, dann gilt:
- Index des linken Kindes: 2i + 1.
- Index des rechten Kindes: 2i + 2.
  
Ist i der Index eines Kindes, dann gilt:
- Index des Elterknotens: (i-1)//2

Die wichtigsten Methoden des Moduls `heapq`:
- `heapify(items)`: Ordnet die Listenelemente so, dass die Heap-Eigenschaft gilt (falls man die Liste wie oben als binären Baum liest). Insbesondere ist nun das erste Element minimal.

- `heappop(items)`: Entfernt und liefert das kleinste (erste) Element der Liste `items` und ordnet die Liste geeignet um.

- `heappush(items, item)`: Fügt ein Element `item` in die Liste `items` ein und ordnet die Liste geeignet um.

**Manipuliert man eine heapified Liste nur mit
`heappop` und `heappush`, dann ist das kleinstes Element immer zuvorderst.**


Oft bestehen die Listenelemente aus (Priority, Wert) Paaren.
Der Wert ist dabei irgend ein Objekt, die Priority ist eine Zahl, und 
das Paar mit der kleinsten Priority wird jeweils gepoppt.  

Sind die Priorities zweier Elemente gleich, dann
werden die Werte verglichen.
Sind die Werte nicht vergleichbar, wird ein Fehler erzeugt.

Um dies zu verhindern, macht man die Priority selber zu einem Paar der Form
> `(eigentliche_Priority, counter_als_Tiebreaker)`.


Die Suchstrategie, bei der jeweils der Knoten mit der kleinsten Priority aus der Liste
`nodes_to_visit` entfernt wird, kann man nun wie folgt implementieren.

```python
import heapq


def search_greedy(node, get_neighbors, get_priority, goal):
    count = 0
    priority = (get_priority(node), count)
    nodes_to_visit = [(priority, node)]
    go_back = {node: None}

    while nodes_to_visit:
        _, node = heapq.heappop(nodes_to_visit)
        if node == goal:
            print(f'Success! count: {count}')
            return node, go_back

        for new_node in get_neighbors(node):
            if new_node in go_back:
                continue

            go_back[new_node] = node
            count += 1
            priority = (get_priority(new_node), count)
            heapq.heappush(nodes_to_visit, (priority, new_node))

    print(f'failure: {count}')
    return node, go_back
```
