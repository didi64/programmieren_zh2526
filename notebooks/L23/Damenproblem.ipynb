### Das Damenproblem (siehe [https://de.wikipedia.org/wiki/Damenproblem](https://de.wikipedia.org/wiki/Damenproblem))

Ziel ist es, alle Möglichkeiten zu finde, $n$ Damen so auf einem $nxn$-Schachbrett zu platzieren,
so dass keine Dame eine andere schlagen kann. Das kleinste Brett auf dem das Problem lösbar ist, ist ein 
4x4 Brett. Eine Lösung ist unten gezeigt.

<img src='/files/images/4queens.png'>  



`queens = (1, 3, 0, 2)`, also Damen auf den Feldern a2, b4, c1 und d3.

Um eine Lösung su suchen, starten wir mit dem leeren Tuple
```
() -> (0,) -> (0, 2) 
           -> (0, 3) 
   -> (1,) -> (1, 3) -> (1, 3, 0) -> (1, 3, 0, 2)    
```

Der Suchpfad ist ein Baum.

Fertig, sobald wir Tief `n_queens` erreichen.

Können aber auch Suche fortsetzen.
Geben alle Blätter mit Tiefe `n_queens` aus.

Suchgraph ist bereits ein Baum.
Brauchen also nicht zu Prüfen, ob eine Knoten bereits besucht wurde.
Da in einem Baum jeder Knoten höchsten einen Elternknoen hat, brauchen wir auch keine Hilfe zum Finden des 
Wegs nach Hause.

````python
# Unweighted graph with 10 nodes (A through J)
graph = {
    'A': ('B', 'C', 'D'),
    'B': ('A', 'E', 'F'),
    'C': ('A', 'G'),
    'D': ('A', 'H'),
    'E': ('B', 'I'),
    'F': ('B', 'J'),
    'G': ('C', 'J'),
    'H': ('D',),
    'I': ('E',),
    'J': ('F', 'G')
}
````

````python
def search_df(node, get_successors, is_goal):
    stack = [node]

    while stack:
        node = stack.pop()
        if is_goal(node):
            return node

        stack.extend(get_successors(node))


def search_df_all(node, get_successors, is_goal):
    stack = [node]

    while stack:
        node = stack.pop()
        if is_goal(node):
            yield node

        stack.extend(get_successors(node))
````

````python
def is_queen_move(p, q):
    '''p, q: Schachbrettfelder (col row)
       returns True falls eine Dame von p nach q ziehen kann
    '''
    return (p[0] == q[0]
            or p[1] == q[1]
            or abs(p[0] - q[0]) == abs(p[1] - q[1])
            )


def extend_queens(queens, n_queens=8):
    '''queens: Tuple der Form (0, 3, 5), steht fuer Damen auf a1, b4, c6
       liefert alle Tupel der Form queens + (row,), von sich nicht bedrohenden Damen
    '''
    col = len(queens)
    if col == n_queens:
        return

    for row in range(n_queens):
        if any(is_queen_move(p, (col, row)) for p in enumerate(queens)):
            continue

        yield queens + (row,)


import widgets_helpers as W
import grid_helpers as G
from ipywidgets import Image


def draw_solution(queens):
    img = Image.from_file('wQ.svg', width=16)
    n = len(queens)
    grid_spec = (10, 10, 20, 20, n, n)
    canvas = W.new_canvas(width=20*(n+1), height=20*(n+1))
    G.draw_chessboard(canvas, grid_spec)

    for col, row in enumerate(queens):
        x, y = G.cr2xy(col, row, grid_spec)
        canvas.draw_image(img, x+2, y+2, 16, 16)

    return canvas
````

````python
n_queens = 5
get_successors = lambda q: extend_queens(q, n_queens)
is_goal = lambda node: len(node) == n_queens
solution = search_df((), get_successors, is_goal)
print(solution)
draw_solution(solution)
````
