### Graphen und Spannbäume

Ein Spannbaum/Gerüst eines Graphen ist ein Baum, der alle Knoten des Graphen enthält.
Besucht man mit depth-first oder breadth-first Search alle Knoten eines Baumes, dann
beschreibt der Dict `go_back` einen Spannbaum, indem er den Elternknoten eines jeden Kindes angibt.

In der Regel beschreibt man einen Baum als gerichteten Graphen.
Zu jedem Knoten gibt man die Kinder an. Zudem nennt man die Wurzel. 
Nachstehende Funktion `cp2rtree(child_parent)` liefert die Wurzel und
einen Dict `parent_children`. Das Argument ist ein `go_back` Dict, den wir hier `child_parent` nennen.




Anbei ein Graph und darunter 2 Spannbäume. 
Der linke wurde mit depth-first Search, der rechte mit depth-first Search erstellt.
(um Spannbäume mit einer anderen Wurzel als A anzuzeigen,

Snapshots der Liste `notes_to_visit` bei fortschreitender Suche:
- DFS: [A] -> [B, C, D] -> [B, C, H] -> [B, G] -> [B, J] -> [B, F] -> [E] -> [I]
- BFS: [A] -> [D, C, B] -> [F, E, D, C] -> [G, F, E, D] -> [H, G, F, E] -> [I, H, G, F] -> [J, I, H, G]  
  Baum wird **Level um Level** aufgebaut:  
  Alle Knoten mit Abstand 1 von der Wurzel, dann alle Knoten mit Abstand 2, ...


Hier noch der linke Baum beschreiben durch einen Dict:
```python
{'A': ['B', 'C', 'D'],
  'D': ['H'],
  'C': ['G'],
  'G': ['J'],
  'J': ['F'],
  'B': ['E'],
  'E': ['I']})
```

````python
from collections import deque


graph = {
    'A': ('B', 'C', 'D'),
    'B': ('A', 'E', 'F'),
    'C': ('A', 'G'),
    'D': ('A', 'H'),
    'E': ('B', 'I'),
    'F': ('B', 'J'),
    'G': ('C', 'J'),
    'H': ('D',),
    'I': ('E',),
    'J': ('F', 'G')
}


def search_df(start, graph, target=None):
    nodes_to_visit = [start]
    go_back = {start: None}

    while nodes_to_visit:
        node = nodes_to_visit.pop()
        if target and node == target:
            break

        for neighbor in graph.get(node, ()):
            if neighbor in go_back:
                continue
            go_back[neighbor] = node
            nodes_to_visit.append(neighbor)

    return node, go_back


def search_bf(start, graph, target=None):
    nodes_to_visit = deque([start])
    go_back = {start: None}

    while nodes_to_visit:
        node = nodes_to_visit.pop()
        if target and node == target:
            break

        for neighbor in graph.get(node, ()):
            if neighbor in go_back:
                continue
            go_back[neighbor] = node
            nodes_to_visit.appendleft(neighbor)

    return node, go_back


def get_path_home(node, go_back):
    path = []
    while node:
        path.append(node)
        node = go_back[node]
    return path


def cp2rtree(child_parent):
    '''return root, tree: dict'''
    parent_children = {p: [] for p in child_parent.values() if p is not None}
    root = None

    for child, parent in child_parent.items():
        if parent is None:
            root = child
            continue
        parent_children[parent].append(child)

    return root, parent_children
````

````python
# kuerzester Weg von D nach F
target, child_parent = search_bf('D', graph, 'F')
get_path_home(target, child_parent)
````

````python
target, child_parent = search_df('A', graph)
child_parent
````

````python
root, tree = cp2rtree(child_parent)
print(f'root: {root}')
tree
````

````python
target, child_parent = search_bf('A', graph)
child_parent
````

````python
root, tree = cp2rtree(child_parent)
print(f'root: {root}')
tree
````
