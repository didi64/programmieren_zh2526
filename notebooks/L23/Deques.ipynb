### Deque (double endet queue)

Bei Listen sind  diese Operationen schnell: 
- `append(item)`: neues Element am Ende anhängen,
- `pop()`: letztes Element entfernen und zurückgeben.

Insbesondere bei langen Listen sind dagegen diese Operationen langsam:  
- `insert(0, item)`: neues Element am Anfang anhängen,
- `pop(0)`: erstes Element entfernen und zurückgeben.

Muss man an beiden Enden Elemente anhängen und/oder einfügen, empfiehlt es sich einen
sog. `deque` zu verwenden. Diese Datenstruktur stellt die gleichen Methoden wie der Typ Liste zur
Verfügung, plus noch
- `appendleft(item)`: schnelles `insert(0, items)`,
- `popleft()`: schnelles `pop(0)`,
- `extendleft(items)`: schnelles Anhängen von Elementen,
- `reverse()`: Deque Umdrehen,
- `rotate(n=1)`: jedes Element zyklisch um n Positionen weiterschieben.

Ein Deque wird so erstellt:
1. `items = deque(Iterable)`
1. `items = deque(Iterable, maxlen=n)`: sobald der Deque Länge n hat, wird beim Einfügen eines Elementes am
  einen Ende ein Element am anderen Ende entfert.
  
Offizielle Pythondokumentation zu [Deque](https://docs.python.org/3/library/collections.html#collections.deque).

```python
from collections import deque
```

```python
items = deque(range(5), maxlen=5)
print(items)

items.append(5)
print(items)

items.appendleft(0)
print(items)
```

```python
items = deque((1, 2, 3))
for i in range(3):
    items.rotate()
    print(items)
```

***
Performance von `list.pop()` und `list.pop(0)`
***

```python
%%timeit -n1
items = list(range(100_000))
while items:
    items.pop()
```

```python
%%timeit -n1
items = list(range(100_000))
while items:
    items.pop(0)
```

***
Performance von `deque.pop()` und `deque.pop(0)`
***

```python
%%timeit -n1
items = deque(range(100_000))
while items:
    items.pop()
```

```python
%%timeit -n1
items = deque(range(100_000))
while items:
    items.popleft()
```

***
Performance von `list.append(item)` und `list.insert(0, item)`
***

```python
%%timeit -n1
items = []
for i in range(100_000):
    items.append(i)
```

```python
%%timeit -n1
items = []
for i in range(100_000):
    items.insert(0, i)
```

***
Performance von `deque.append(item)` und `deque.insert(0, item)`
***

```python
%%timeit -n1
items = deque()
for i in range(100_000):
    items.append(i)
```

```python
%%timeit -n1
items = deque()
for i in range(100_000):
    items.appendleft(i)
```
