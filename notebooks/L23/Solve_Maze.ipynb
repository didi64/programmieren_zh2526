### Weg durch Labyrinth finden
Wir wollen depth-first search nutzen, um einen Weg vom Feld (0, 0) zum Feld (19, 9) in folgendem Labyrinth zu finden.  
<img src='/files/images/aufgabe_maze.png'>

Der Graph ist im File `maze.py` als Dictionary gespeichert. Weiter sind dort die Variabeln
`dims = (20, 10)`,  `start = (0, 0)` und `target = (19,9)` definiert.

In der gegebenen Situation brauchen wir nicht eine Funktion `search_df(start, get_neighbors, is_goal)`,
welche Funktionen `get_neighbors(node)` und `is_goal(node)` als Argumente nimmt,
sondern benutzen `search_df(start, graph, target)`, wo `graph` der Graph-Dict und `target` die Zielposition ist.

Im Funktionsbody erhalten wir dann die Nachbarn mit `graph.get(node, ())` (Knoten ohne Nachbarn sind nicht im Dict `graph`) und testen mit `node == target`, ob wir den Zielknoten gefunden haben.

```python
from maze import dims, start, target, graph


start, target, dims
```

```python
def search_df(root, tree, target):
    nodes_to_visit = [root]
    go_back = {root: None}

    while nodes_to_visit:
        parent = nodes_to_visit.pop()
        if parent == target:
            return parent, go_back

        for child in tree.get(parent, ()):
            if child in go_back:
                continue

            go_back[child] = parent
            nodes_to_visit.append(child)


def get_path_home(node, go_back):
    path = []
    while node:
        path.append(node)
        node = go_back[node]

    return path
```

```python
target, child_parent = search_df(start, graph, target)
target
```

```python
path = get_path_home(target, child_parent)
path
```

```python
import widget_helpers as W
import grid_helpers as G


def get_canvas_and_grid_spec(ncol, nrow, dx=20, dy=20):
    canvas = W.get_canvas(width=ncol*dx, height=nrow*dy)
    grid_spec = (0, 0, dx, dy, ncol, nrow)
    G.draw_grid(canvas, grid_spec, color='black')

    canvas.stroke_style = 'blue'
    canvas.line_width = 3
    return canvas, grid_spec


def connect(canvas, src, dst, grid_spec):
    p = G.cr2xy(*src, grid_spec, center=True)
    q = G.cr2xy(*dst, grid_spec, center=True)
    canvas.stroke_line(*p, *q)


def draw_maze(canvas, grid_spec, graph):
    canvas.stroke_style = 'blue'
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            connect(canvas, node, neighbor, grid_spec)


def mark_path(canvas, grid_spec, path):
    canvas.stroke_style = 'lightblue'
    for i in range(1, len(path)):
        connect(canvas, path[i-1], path[i], grid_spec)
```

```python
canvas, grid_spec = get_canvas_and_grid_spec(*dims)
canvas
```

```python
draw_maze(canvas, grid_spec, graph)
```

```python
mark_path(canvas, grid_spec, path)
```
