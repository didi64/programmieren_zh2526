### Projektvorschläge

Ziel ist es, anhand eines konkreten Suchproblems 
verschiedene Suchstrategien zu testen.
Der Suchraum des Problems sollte skalierbar sein.  
Als erstes sollte das Problem so aufbereitet werden, dass ein kleiner Suchraum mit
DFS und BFS durchsucht werden kann.

Dann ist der Suchraum zu vergrössern. 
Benutze dann eine greedy-Search mit einer geeigneten Heuristik, und falls anwendbar,
bidirektionale Breitensuche.


**Mögliche Suchprobleme**:
- Sudoku-Varianten wie
    - [Killer-SUdoku](https://de.wikipedia.org/wiki/Killer-Sudoku),
    - [Hitori](https://de.wikipedia.org/wiki/Hitori),
    - [Futoshiki](https://de.wikipedia.org/wiki/Futoshiki).<br><br>
- [Wortleiter-Problem](https://de.wikipedia.org/wiki/Wortleiter):
  Gegen ist eine Liste mit Wörtern der Länge 5 (z.B. `WORDS_CH5_umlautfrei.txt`). Zwei Wörter sind benachbart, falls sie sich nur an einer Position unterscheiden.
  Finde Zusammenhangskomponenten in diesem Graph. Finde Wörter, die weit voneinander entfernt sind.
  Finde lange Kreise (geschlossene Wege).
- Schiebepuzzle wie
    - [Rush Hour](https://de.wikipedia.org/wiki/Rush_Hour_(Spiel))  
      Das rote Auto muss herausgefahren werden.  
      Fahrzeuge können sich nur vor- und rückwärts bewegen.<br><br>

    - [Klotski](https://en.wikipedia.org/wiki/Klotski)
 
    - Peg-Solitaire
    - Top Spin<br><br>
       <img src='/files/images/topspin.jpg' width='200px'> 
  
 
- 2x2x2 Rubiks-Cube:
  - Finde statt nur einer Lösung alle kürzesten Lösungen (suche nicht nach einem Zustand, sonder nach allen
    Wegen dorthin). 
  - Finde Weg zu einer Seite, einem Stockwerk.
  - Finde alle kürzesten Operationen, die z.B. 2 Würfelchen vertauschen.

 - Sokoban  
   Finde die Lösung kleiner Levels (`Sokobanlevels_Anomaly.txt` und `Sokobanlevels_Albe.txt`).
