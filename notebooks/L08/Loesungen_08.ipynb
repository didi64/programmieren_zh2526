```python
def unique(items):
    '''gib eine Liste zurueck, die jedes Element von xs
       in der gleichen Reihenfolge enthalt, aber nur einmal.
       xs: list
    '''
    unique_items = []
    seen = set()
    for item in items:
        if item not in seen:
            unique_items.append(item)
            seen.add(item)
    return unique_items
```

```python
s = 'abakadabra'
unique(s)
```

```python
from random import randint


def random_numbers():
    numbers = set()
    while len(numbers) < 5:
        n = randint(1, 9)
        numbers.add(n)

    return numbers
```

```python
random_numbers()
```

```python
def set_intersection(xs, ys):
    '''gib  eine Menge mit den Elementen, die in  xs UND ys sind zurueck'''
    return set(x for x in xs if x in ys)
```

```python
for i in range(3):
    xs = random_numbers()
    ys = random_numbers()
    print(f'{xs} geschnitten mit {ys}: {set_intersection(xs, ys)}')
```

```python
for i in range(3):
    xs = random_numbers()
    ys = random_numbers()
    if xs & ys != set_intersection(xs, ys):
        print(f'Test 1 failed with {xs} and {ys}!')

    if xs.intersection(ys) != set_intersection(xs, ys):
        print(f'Test 2 failed with {xs} and {ys}!')

print('all tests passed')
```

```python
def set_diff(xs, ys):
    '''gib eine Menge mit den Elemente, die in xs aber nicht in ys sind zurueck'''
    return set(x for x in xs if x not in ys)
```

```python
for i in range(3):
    xs = random_numbers()
    ys = random_numbers()
    print(f'{xs} ohne {ys}: {set_diff(xs, ys)}')
```

```python
for i in range(3):
    xs = random_numbers()
    ys = random_numbers()
    if xs - ys != set_diff(xs, ys):
        print(f'Test 1 failed with {xs} and {ys}!')

print('all tests passed')
```

```python
def set_union(xs, ys):
    '''gib  eine Menge mit den Elementen, die in  xs ODER ys sind zurueck'''
    union = set(xs)
    for y in ys:
        union.add(y)

    return union
```

```python
for i in range(3):
    xs = random_numbers()
    ys = random_numbers()
    print(f'{xs} vereint mit {ys}: {set_union(xs, ys)}')
```

```python
for i in range(3):
    xs = random_numbers()
    ys = random_numbers()
    if xs | ys != set_union(xs, ys):
        print(f'Test 1 failed with {xs} and {ys}!')

    if xs.union(ys) != set_union(xs, ys):
        print(f'Test 2 failed with {xs} and {ys}!')

print('all tests passed')
```

```python
def make_count_dict_1(items):
    '''gibt einen dict zurueck, der zu jedem Element in items
       angibt, wie oft es in vorkommt
    '''
    count_dict = {}
    for item in items:
        if item in count_dict:
            count_dict[item] = count_dict[item] + 1
        else:
            count_dict[item] = 1

    return count_dict
```

```python
def make_count_dict_2(items):
    '''gibt einen dict zurueck, der zu jedem Element in items
       angibt, wie oft es in vorkommt
    '''
    count_dict = {}
    for item in items:
        count_dict[item] = count_dict.get(item, 0) + 1

    return count_dict
```

```python
def make_count_dict(items):
    '''gibt einen dict zurueck, der zu jedem Element in items
       angibt, wie oft es in vorkommt
       absteigend soritert nach Werten
    '''
    count_dict = {}
    for item in items:
        count_dict[item] = count_dict.get(item, 0) + 1

    kv_pairs = sorted(count_dict.items(), key=lambda x: (x[1], x[0]), reverse=True)
    return dict(kv_pairs)
```

```python
numbers = [2, 1, 4, 5, 6, 4, 5, 4, 5, 6, 4, 5, 4, 3, 4, 3]
make_count_dict(numbers)
```
