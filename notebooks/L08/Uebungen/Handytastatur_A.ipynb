Bei alten Handytastaturen sind auf den Zifferntasten 2-9 auch noch 
die Buchstaben des Alphabets zu finden. Der Dictionary
`digit_letters` ordnet jeder dieser Ziffern Buchstaben zu. 
Rufnummern können so mit Wörter repräsentiert werden z.B. 
`'hallo'` -> `'42556'`.
1. Erstelle einen Dictionary,
    ```python
    letter_digit = {'a': '2', 'b': '2', 'c': '2', 'd': '3', ..., 'z': '9'}
    ```  
    der jedem Buchstaben die entsprechende Ziffer (als String) zuordnet.
1. Schreibe eine Funktion `word2numbers(word)`, 
die die Nummer  zurück gibt, die beim Tippen des Worts `word` gewählt wird.
`word2numbers('hallo')` -> `'42556'`.

1. Wir wollen alle Wörter finden, die beim Tippen eine bestimmte Nummer ergeben.
   1. Schreibe eine Funktion `extend(words, chars)`, welche die Liste aller Wörter
      zurückgibt, die man erhält indem man ein Wort aus
      der Liste `words` mit einem Buchstaben aus `chars` verlängert.  
      `extend(['foo', 'bar'], 'xy')` -> `['foox', 'fooy', 'barx', 'bary']`.
   1. Schreibe eine Funktion `number2words(number)`, die eine Liste mit allen
      Wörter zurückgibt, welche der Rufnummber `number` entsprechen.
1. Das File `words5.txt` enthält Wörter der Länge 5. Erstelle eine Liste `words5` mit allen Wörtern dieses Files. Finde alle Wörter in dieser Liste, die
   den Nummern $74663$, $43436$ und $53463$ entsprechen. 

````python
digit_letters = {
    '2': 'abc',
    '3': 'def',
    '4': 'ghi',
    '5': 'jkl',
    '6': 'mno',
    '7': 'pqrs',
    '8': 'tuv',
    '9': 'wxyz',
}
````

````python
letter_digit = {}
for digit, letters in digit_letters.items():
    for c in letters:
        letter_digit[c] = digit
letter_digit
````

````python
def word2number(word):
    number = ''.join(letter_digit[c] for c in word)
    return number
````

````python
word2number('hallo')
````

````python
def extend(words, chars):
    return [word+c for word in words for c in chars]
````

````python
words = ['foo', 'bar']
extend(words, 'xy')
````

````python
def number2words(number):
    words = ['']
    for digit in number:
        chars = digit_letters[digit]
        words = extend(words, chars)
    return words
````

````python
words = number2words('268')
words[:3] + words[-3:]
````

````python
fn = 'words5.txt'
with open(fn, 'r') as f:
    words5 = [line.strip().lower() for line in f]

words5[:5], words5[-5:]
````

````python
numbers = ['74663', '43436', '53463']
for number in numbers:
    words = number2words(number)
    words = [word for word in words if word in words5]
    print(words)
````
