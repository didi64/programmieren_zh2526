### Wie gross ist der durchschnittliche Gewinn?
Jemand schlägt dir 3 Varianten eines Spiels vor.  
Der Einsatz beträgt 3 Franken.  
Du bekommst 5 Franken, falls du



- mit 6  Würfeln  mindestens eine 6 würfelst,
- mit 12 Würfeln  mindestens zwei 6er würfelst,
- mit 18 Würfeln  mindestens drei 6er würfelst.

Spiele jede Variante  100'000-mal. Wie gross ist der durchschnittliche Gewinn?  
Implementiere zum würfeln die  Funktion `wuerfle(n)`. Die Funktion

````python
import random


def wuerfle(n):
    return [random.randint(1, 6) for i in range(n)]
````

````python
N = 100_000
einsatz = 3
auszahlung = 5
n = 1  # Variante: mit 6*n Wuerfeln mind. n 6er wuerfeln (teste mit n=1, 2, 3)
balance = 0

for _ in range(N):
    balance -= einsatz
    wurf = wuerfle(n * 6)
    if wurf.count(6) >= n:
        balance += auszahlung

print(f'Durchschnittlicher Gewinn: {round((balance/N), ndigits=3)}')
````

````python
def durchschnittlicher_gewinn(n, N=100_000, einsatz=3, auszahlung=5):
    balance = 0

    for _ in range(N):
        balance -= einsatz
        wurf = wuerfle(n * 6)
        if wurf.count(6) >= n:
            balance += auszahlung

    return round((balance/N), ndigits=3)
````

````python
gewinne = [(durchschnittlicher_gewinn(n), n) for n in range(1, 4)]
gewinn, variante = max(gewinne)
print(f'groesster Gewinn bei Variante {variante}: {gewinn}')
````
