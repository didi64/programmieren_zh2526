1. Ã–ffne in Jupyterlab ein neues File `file_1.txt` und
copy/paste folgende Zeilen in dieses File.
```
 foo , bar ,   baz
Anna , Bob ,  Carl
```
2. Schreibe Python-Code, der dieses File liest und dann ein
   neues File `file_2.txt` schreibt, wo nun das
   Komma durch ein Semilonon ersetzt ist, und aller Space entfernt ist,
   bis auf einen Leerschlag nach dem Semikolon.
3. Benutze Python-Code, um ein File `file_3.txt` zu erstellen,
   mit dem Inhalt beider Files. Stellen den beiden Fileinhalten jeweils ein Header voran.
   `file_3.txt` sieht nun so aus:
```   
----------------
File: file_1.txt
----------------
 foo , bar ,   baz
Anna , Bob ,  Carl
----------------
File: file_2.txt
----------------
foo; bar; baz
Anna; Bob; Carl
```

````python
header_1 = '''\
----------------
File: file_1.txt
----------------
'''

header_2 = '''\
----------------
File: file_2.txt
----------------
'''
````

````python
# File 'file_1.txt' lesen
file_1 = 'file_1.txt'

with open(file_1, mode='r') as f:
    lines_1 = [line.rstrip() for line in f]

lines_1
````

````python
# Trennzeichen ersetzten und SPACE entfernen
old_sep = ','
new_sep = '; '


lines_2 = []
for line in lines_1:
    values = line.split(old_sep)
    values = [value.strip() for value in values]
    new_line = new_sep.join(values)
    new_line = new_line + '\n'
    lines_2.append(new_line)

lines_2
````

````python
# Zeilen in 'file_2.txt' schreiben
file_2 = 'file_2.txt'

with open(file_2, mode='w') as f:
    f.writelines(lines_2)
````

````python
# Zeilen der beiden Files kombinieren, jeweils header voranstellen
with open(file_1, mode='r') as f:
    lines_1 = [line.rstrip() for line in f]

with open(file_2, mode='r') as f:
    lines_2 = [line.rstrip() for line in f]


lines_3 = header_1.splitlines() + lines_1 + header_2.splitlines() + lines_2
lines_3
````

````python
# fuege \n' an die Zeilenenden
lines_3 = [line + '\n' for line in lines_3]
lines_3
````

````python
# Zeilen in 'file_3.txt' schreiben
file_3 = 'file_3.txt'

with open(file_3, mode='w') as f:
    f.writelines(lines_3)
````
