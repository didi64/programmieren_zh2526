```python
def insert(s, i, word):
    '''fuege word an Position i in den String s ein'''
    return s[:i] + word + s[i:]
```

```python
insert('foobar', 3, 'xx') == 'fooxxbar'
```

```python
def digits(start, end, sep=','):
    '''gibt einen String mit durch sep getrennten Zahlen
       von start bis vor end zurueck
    '''
    result = ''
    for i in range(start, end):
        result = result + str(i) + sep
    return result[:-1]
```

```python
digits(5, 10, sep=':') == '5:6:7:8:9'
```

```python
digits(1, 1) == ''
```

```python
def digits(start, end, sep=','):
    '''gibt einen String mit durch sep getrennten Zahlen
       von start bis vor end zurueck
    '''
    if end <= start:
        return ''

    result = ''
    for i in range(start, end-1):
        result = result + str(i) + sep
    return result + str(end-1)
```

```python
digits(5, 10, sep=':') == '5:6:7:8:9'
```

```python
digits(1, 1) == ''
```

```python
def digits(start, end, sep=','):
    '''gibt einen String mit durch sep getrennten Zahlen
       von start bis vor end zurueck
    '''
    digits = [str(i) for i in range(start, end)]
    return sep.join(digits)
```

```python
digits(5, 10, sep=':') == '5:6:7:8:9'
```

```python
digits(1, 1) == ''
```

```python
def merge(v, w):
    '''fuege die Buchstaben der Woerter v und w
       reissverschlussmaessig zusammen bis dem kuerzeren Wort
       die Buchstaben ausgehen
    '''
    result = ''
    n = min(len(v), len(w))
    for i in range(n):
        result = result + v[i] + w[i]
    return result
```

```python
merge('abc', 'xyzZZ') == 'axbycz'
```

```python
def merge(v, w):
    '''fuege die Buchstaben der Woerter v und w
       reissverschlussmaessig zusammen bis dem kuerzeren Wort
       die Buchstaben ausgehen
    '''
    return ''.join([x + y for x, y in zip(v, w)])
```

```python
merge('abc', 'xyzZZ')
```

```python
def unmerge(s):
    '''macht merge_two_words(v, w) rueckgaengig
       (falls v und w gleich lang)
    '''
    v = s[::2]  # jeder 2.te Buchstabe
    w = s[1::2]
    return v, w
```

```python
word1 = 'foo'
word2 = 'bar'
unmerge(merge(word1, word2)) == (word1, word2)
```

```python
def alphabet(n):
    '''gibt einen String mit den ersten n Buchstaben des
       Alphabets zurueck
    '''
    result = ''
    for i in range(n):
        result = result + chr(65+i)
    return result
```

```python
alphabet(5) == 'ABCDE'
```

```python
def alphabet(n, lower=True):
    '''gibt einen String mit den ersten n Buchstaben des
       Alphabets zurueck (Kleinbuchstaben falls lower=True, sonst
       Grossbuchstaben)
    '''
    letters = [chr(65+i) for i in range(n)]
    s = ''.join(letters)
    return s
```

```python
alphabet(5) == 'abcde'
```
