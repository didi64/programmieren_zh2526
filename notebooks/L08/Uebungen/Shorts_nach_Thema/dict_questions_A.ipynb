````python
# small_dict_1
small_dict_1 = {'cat': 'Katze', 'bird': 'Vogel'}
small_dict_1
````

````python
# small_dict_2
words_en = ['cow', 'dog']
words_de = ['Kuh', 'Hund']
small_dict_2 = dict(zip(words_en, words_de))
small_dict_2
````

````python
# big_dict mit |
big_dict = small_dict_1 | small_dict_2
big_dict
````

````python
# big_dict mit For-Loops
big_dict = {}
for k, v in small_dict_1.items():
    big_dict[k] = v

for k, v in small_dict_2.items():
    big_dict[k] = v
````

````python
# count_dict fuer zu numbers
numbers = [2, 1, 4, 5, 6, 4, 5, 4, 5, 6, 4, 5, 4, 3, 4, 3]

count_dict = {}
for n in numbers:
    count_dict[n] = count_dict.get(n, 0) + 1

numbers_unique = list(count_dict)
numbers_unique
````

Zahlen werden in der richtigen Reihenfolge als Schlüssel in den
`count_dict` aufgenommen und jeweils nur einmal.
Deshalb bleibt die Reihenfolge erhalten.

````python
# index_dict
index_dict = {}
for i, n in enumerate(numbers):
    if n in index_dict:
        index_dict[n].append(i)
    else:
        index_dict[n] = [i]

index_dict
````

````python
# Uebersetzungsdict fuer grosse und Kleine Umlaute
subst = {'ä': 'ae',  'ö': 'oe', 'ü': 'ue'}
subst = subst | {k.upper(): v.capitalize() for k, v in subst.items()}
subst
````

````python
# alle Umlaute ersetzen
s = 'Äpfel, Öl, Übung, äusserst, schön, über'

for old, new in subst.items():
    s = s.replace(old, new)

s
````
