```python
def inverse(xs):
    '''xs ist eine Liste mit Zahlen.
       gib eine Liste mit den reziproken Werten (1/x) dieser Zahlen zurueck.
       Ignoriere 0.
    '''
```

```python
def popn(xs, n):
    '''Entferne die letzten n Elemente aus der Liste
       gib eine Liste mit den entfernten Elementen zurueck
       (stoppe, wenn die Liste leer ist)
    '''
```

```python
xs = [1, 2, 3]
ys = popn(xs, 2)
xs == [1] and ys == [3, 2]
```

```python
xs = [1, 2, 3]
ys = popn(xs, 5)
xs == [] and ys == [3, 2, 1]
```
