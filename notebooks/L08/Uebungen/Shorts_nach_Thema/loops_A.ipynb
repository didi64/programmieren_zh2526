```python
def find_all(s, char):
    '''gibt eine Liste zurueck mit den
       Positionen, wo char in s auftritt
    '''
    indices = []
    for i, c in enumerate(s):
        if c == char:
            indices.append(i)
    return indices
```

```python
find_all('abakadabra', 'a') == [0, 2, 4, 6, 9]
```

```python
def find_all(s, char):
    '''gibt eine Liste zurueck mit den
       Positionen, wo char in s auftritt
    '''
    indices = []
    n = len(s)
    i = 0

    while i < n:
        if s[i] == char:
            indices.append(i)
        i = i + 1
    return indices
```

```python
find_all('abakadabra', 'a') == [0, 2, 4, 6, 9]
```

```python
def find_all(s, char):
    '''gibt eine Liste zurueck mit den
       Positionen, wo char in s auftritt
    '''
    indices = []
    i = s.find(char)
    while i != -1:
        indices.append(i)
        i = s.find(char, i+1)
    return indices
```

```python
find_all('abakadabra', 'a') == [0, 2, 4, 6, 9]
```

```python
def delete_chars(s, chars):
    '''loesche alle Vorkommen der Zeichen chars aus dem String s'''
    result = ''
    for c in s:
        if c in chars:
            continue
        result = result + c
    return result
```

```python
s = 'a1b2c3'
chars = '123'
delete_chars(s, chars) == 'abc'
```

```python
def delete_chars(s, chars):
    '''loesche alle Vorkommen der Zeichen chars aus dem String s'''
    result = ''
    n = len(s)
    i = 0

    while i < n:
        if s[i] not in chars:
            result = result + s[i]
        i = i + 1
    return result
```

```python
s = 'a1b2c3'
chars = '123'
delete_chars(s, chars) == 'abc'
```

```python
def delete_chars(s, chars):
    '''loesche alle Vorkommen der Zeichen chars aus dem String s'''
    for c in chars:
        s = s.replace(c, '')
    return s
```

```python
s = 'a1b2c3'
chars = '123'
delete_chars(s, chars) == 'abc'
```
