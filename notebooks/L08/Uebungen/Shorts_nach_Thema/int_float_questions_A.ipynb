### Integer und Floats
1. Erstelle eine Liste mit 5 Zufallszahlen zwischen 100 und 1000.
1. Finde zu jeder Zahl in der Liste Integers d und r mit 0 <= r < 13, so dass n gleich d*13 + r,
   und gib dann diese Gleichung aus.Z.B. bei der Zahl 226 ist
   ```
       226 ist 17*13 + 5
   ```
   auszugeben.
1. Gib für jede Zahl in der Liste die Wurzel der Zahl aus, auf 3 Nachkommastellen gerunde.
1. Erstelle eine Liste mit den durch 137 teilbaren Zahlen zwischen 500 und 1000.
   Produziere dann den Output
    ```
    548 ist 4*137
    685 ist 5*137
    822 ist 6*137
    959 ist 7*137
    ```
1. Führe die beiden folgenden Zeilen Code aus.
    ```python
    1/5 + 2/5 == 3/5  # False
    1/5 + 2/5 == 1351079888211149 / 2251799813685248  # True
    ```
    Was ist der Grund für diese Verhalten?
    Schreibe eine Funktion `is_close(x, y)`, die `True` zurück gibt, falls
    x und y weniger als $10^{-15}$ auseinander liegen. 
    
   

````python
from random import randint
````

````python
numbers = [randint(100, 1000) for _ in range(5)]
numbers
````

````python
for n in numbers:
    print(f'{n} ist {n // 13}*13 + {n % 13}')
````

````python
for n in numbers:
    print(f'Die Wurzel aus {n} ist {round(n**0.5, ndigits=3)}')
````

````python
multiples_of_137 = [n for n in range(500, 1000) if n % 137 == 0]

for n in multiples_of_137:
    print(f'{n} ist {n // 137}*137')
````

````python
def is_close(x, y, err=1E-16):
    return abs(x - y) < err
````

````python
is_close(1/5 + 2/5, 3/5)
````

Wie sich im Dezimalsystem 1/3=0.333... nicht exakt als endlicher Dezimalbruch darstellen lässt, so  
kann man 1/5 und 2/5 im Binärsystem nicht exakt darstellen.
Bei der Addition der genäherten Werte addiert sich auch der Fehler und wird sichtbar.
