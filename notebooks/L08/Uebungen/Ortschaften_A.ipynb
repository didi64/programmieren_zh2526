### Ortschaften  
Nützliche Stringmethoden: `rstrip`, `startswith`, `endswith`, `split`, `lower`.  


Das File `ortsnamen.txt` enthält Zeilen mit den Ortschaftsnamen der Schweiz und Liechtenstein.
1. Erstelle eine Liste mit allen  Ortschaftsnamen des Files. Wieviele sind es?
1. Gib alle Ortschaften aus, die nur aus einem Wort bestehen (kein SPACE enthalten) und auf `'tal'` enden.
1. Wieviele Ortschaften beginnen mit einem französischen Artikel (`'Le '`, `'La '`, `'Les '`)?
1. Wieviele Ortschaften beginnen mit einer Abkürzung (`'.'` vor dem ersten SPACE)?
1. Welches sind die 5 längsten Ortsnamen ohne SPACE? (Erstelle eine Liste und sortiere nach der Länge.)
1. Welches sind die 5 längsten Ortsnamen ohne eines de Zeichen in `' -/'`? Benutze zum Aussondern eine Hilfsfunktion `contains_one_of(word, chars)`, die `True` zurück gibt, falls `word` eines der Zeichen im String `chars` enthält.
1. Forme eine Menge mit den Anfangsbuchstaben aller Ortsnamen. Mache daraus einen sortierten String.
1. Mit welchem Buchstaben beginnt kein Ortsname? 
1. Erstellen einen Dictionary, der zu jedem Buchstaben angibt, wieviele Ortschaftsnamen damit beginnen.
1. Erstellen einen Dictionary, der jedem Buchstaben eine Liste zuordnet mit den 3 kürzesten Ortsnamen, die mit diesem Buchstaben beginnen.

```python
# Orte in Liste orte lesen
fn = 'ortsnamen.txt'
with open(fn, 'r') as f:
    orte = [line.rstrip() for line in f]
len(orte)
```

```python
# Orte ohne SPACE die auf 'tal' enden
[ort for ort in orte if ' ' not in ort and ort.endswith('tal')]
```

```python
# Orte, die mit franz. Artikel beginnen
prefixes = ('Le ', 'La ', 'Les ')
len([ort for ort in orte if ort.startswith(prefixes)])
```

```python
# Orte, die Akuerzung beginnen (St. Gallen)
len([ort for ort in orte if ort.split()[0].endswith('.')])
```

```python
# Die 5 laensten Ortsnamen ohne Space
q = [ort for ort in orte if ' ' not in ort]
q = sorted(q, key=lambda x: len(x), reverse=True)
q[:5]
```

```python
def contains_one_of(word, chars):
    for c in chars:
        if c in word:
            return True
```

```python
contains_one_of('abc', 'xyz')
```

```python
# Die 5 laensten Ortsnamen ohne ' -/'
q = [ort for ort in orte if not contains_one_of(ort, ' -/')]
q = sorted(q, key=lambda x: len(x), reverse=True)
q[:5]
```

```python
# Sortierter Buchstaben mit denen Ortsnamen beginnen
anfangsbuchstaben = ''.join(sorted(set(ort[0] for ort in orte)))
anfangsbuchstaben
```

```python
# BUchstaben, mit denen kein Ortsname beginnt
ABC = ''.join(chr(65+i) for i in range(26))
set(ABC) - set(anfangsbuchstaben)
```

```python
# Dict, der jedem Buchstaben die Anzahl Orte mit diesem Anfangsbuchstaben zuordnet
d = {}
for ort in orte:
    k = ort[0]
    d[k] = d.get(k, 0) + 1

count_dict = dict(sorted(d.items(), key=lambda x: (x[1], x[0]), reverse=True))
count_dict
```

```python
# Dict, der jedem Buchstaben die 3 kuerzesten Ortsnamen mit diesem Anfangsbuchstaben zuordnet
d = {}
for k in sorted(count_dict):
    q = [ort for ort in orte if ort.startswith(k)]
    q = sorted(q, key=lambda x: len(x))
    d[k] = q[:3]

d
```
