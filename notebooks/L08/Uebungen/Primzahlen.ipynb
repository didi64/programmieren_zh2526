   Entferne das letzte Element $p$ aus `numbers` und nimm es in die Liste
   `primes` auf. Erstelle eine neue Liste `numbers` mit den Zahlen aus der alten Liste, die nicht durch $p$ teilbar sind.
1.  Wiederhole obigen Schritt solange, bis die Liste `numbers` leer ist.
    Die Liste `primes` sollte nun alle Primzahlen kleiner 30 enthalten.
