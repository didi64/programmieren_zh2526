### Ortschaften  
Nützliche Stringmethoden: `rstrip`, `startswith`, `endswith`, `split`, `lower`.  


Das File `ortsnamen.txt` enthält Zeilen mit den Ortschaftsnamen der Schweiz und Liechtenstein.
1. Erstelle eine Liste mit allen  Ortschaftsnamen des Files. Wieviele sind es?
1. Gib alle Ortschaften aus, die nur aus einem Wort bestehen (kein SPACE enthalten) und auf `'tal'` enden.
1. Wieviele Ortschaften beginnen mit einem französischen Artikel (`'Le '`, `'La '`, `'Les '`)?
1. Wieviele Ortschaften beginnen mit einer Abkürzung (`'.'` vor dem ersten SPACE)?
1. Welches sind die 5 längsten Ortsnamen ohne SPACE? (Erstelle eine Liste und sortiere nach der Länge.)
1. Welches sind die 5 längsten Ortsnamen ohne eines de Zeichen in `' -/'`? Benutze zum Aussondern eine Hilfsfunktion `contains_one_of(word, chars)`, die `True` zurück gibt, falls `word` eines der Zeichen im String `chars` enthält.
1. Forme eine Menge mit den Anfangsbuchstaben aller Ortsnamen. Mache daraus einen sortierten String.
1. Mit welchem Buchstaben beginnt kein Ortsname? 
1. Erstellen einen Dictionary, der zu jedem Buchstaben angibt, wieviele Ortschaftsnamen damit beginnen.
1. Erstellen einen Dictionary, der jedem Buchstaben eine Liste zuordnet mit den 3 kürzesten Ortsnamen, die mit diesem Buchstaben beginnen.
