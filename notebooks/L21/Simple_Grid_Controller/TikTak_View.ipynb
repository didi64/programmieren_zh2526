### Darstellungskomponente (View) zu TikTakToe


Die Funktion `update(event, **kwargs)` der Spielkomponente wird
von folgenden Funktionen wie folgt aufgerufen:

- `new-game()`:
  - `update('new_game')`
- `check_result(player)`:
  - `update('winner', win_line=<Positionen pos in der win-line>, winner=<'X' oder 'O'>)`
  - `update('draw')`
- `play(pos)`:
  - `update('play', player=<'X' oder 'O'>, pos=<Position>)`

Wir überschreiben die Funktion `game.update`, so dass
das Spielfeld graphisch auf einem Canvas-Widget dargestellt wird.  
Das Modul `view_0` definiert eine Funktion `update`, welche 
abhänging vom Argument `event` eine Funktion aufruft, die
printet, was getan werden sollte.
Ziel ist es, diese Funktionen so zu modifieren, dass der Spielzustand graphisch auf einen
Canvas-Widget dargestellt wird.

```python
import tiktak_L19 as game
import view_0 as view


game.update = view.update
```

```python
game.new_game()
game.show()
```

```python
game.new_game()
for i in (4, 0, 1, 7, 3, 5, 2, 6, 8):
    game.play(game.idx2pos(i))
game.show()
```

Nachstehende zeigen wir, wie der Spielzustand dargestellt werden kann.
***

```python
# Canvas-WIdget erstellen und grid_spec definieren
import widget_helpers as W
import grid_helpers as G


width, height = 100, 100
canvas = W.get_canvas(width, height)
grid_spec = (width//8, height//4 - 2, width//4, height//4, 3, 3)

canvas
```

```python
# Gitter zeichnen
canvas.clear()
G.draw_grid(canvas, grid_spec, color='blue')
```

```python
# Resultat ausgeben
canvas.text_align = 'center'
canvas.text_baseline = 'ideographic'
canvas.font = f'{height//5}px sans-serif'
canvas.fill_text('draw', width/2, grid_spec[1])
```

```python
# Spielfeld markieren
player = 'X'
pos = (1, 1)
G.fill_text(canvas, player, pos, grid_spec, color='black')
```

```python
# win_line markieren
# canvas.save()  # speichert den Zustand der Leinwand
win_line = [(0, 0), (1, 1), (2, 2)]
canvas.global_alpha = 0.3
for pos in win_line:
    G.fill_rect(canvas, pos, grid_spec, color='yellow')
# canvas.restore()  # stellt den gespeicherten Zustand wieder her
```

- Um das Modul reimportieren zu können ohne immer den Kernel neu starten zu müssen,
benutzen wir die Methode `reload` des Moduls  `importlib`.
