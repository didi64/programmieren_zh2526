### TikTakToe 
Mit Spielkomponente (Model), graphischer Darstellung (View) und Maussteuerung (Controller).

Die für die Steuerung zuständige Komponente nennen wir Controller.
Der Controller nimmt Benutzerinput entgegen und ruft dann die entsprechenden Funktionen
der Spielkomponente auf. Wir benutzen ein Canvas-Widget, um auf Mausklicks zu hören.

````python
import widget_helpers as W
import grid_helpers as G
import tiktak_L19 as game
import view_tiktaktoe as view
from IPython.display import display


view.init(game, width=200, height=200)

# Controller-Code
out = W.get_out()


@out.capture(clear_output=True)
def on_mouse_down(x, y):
    pos = G.xy2cr(x, y, view.grid_spec)
    game.play(pos)


@out.capture(clear_output=True)
def on_key_down(key, *flags):
    if key == 'n':
        game.new_game()


view.canvas.on_mouse_down(on_mouse_down)
view.canvas.on_key_down(on_key_down)
display(view.canvas, out)
view.canvas.focus()
````

### Konfigurierbarer Grid-Controller
Bei vielen Spielen werden Steine auf einem Brett/Gitter verschoben.  
Die Controllerkomponente sieht in diesen Fällen immer ähnlich aus.  

Ziel ist es, eine konfigurierbare Controllerkomponente zu schreiben.
Ein erster Ansatz ist `grid_controller.py`. Dieses Modul hat eine Funktion  
`init(canvas, grid_spec, callbacks)`, wobei
- `canvas`: Widget, das auf die Mausklicks hört,
- `grid_spec`: spezifiziert das Gitter,
- `callbacks`: Dict der Form
    ```python
    {'mouse_down': game.play),
     'n': game.new_game),
     }
    ```
    Schlüssel ist ein Mausevent oder ein Tastenname.
    Der Wert ist die Funktion, die in diesem Fall aufgerufen wird.  
    Ist der Schlüssel ein Mausevent, dann wird der Funktion
    die Argument `pos` übergeben. `pos` ist ein Tuple mit der Gitterspalte und Zeile, in die geklickt wurde (kann ausserhalb des Gitters liegen). 

````python
import tiktak_L19 as game
import view_tiktaktoe as view
import grid_controller as C


view.init(game, width=200, height=200)

callbacks = {'mouse_down': game.play,
             'n': game.new_game,
             }


C.init(view.canvas, callbacks, view.grid_spec)
C.show(debug=True)
````
