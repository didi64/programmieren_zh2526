```python
import grid_helpers as G


def reveal_next():
    n = game.N
    for col in range(n):
        for row in range(n):
            if (col, row) in game.mines or (col, row) in game.save_fields:
                continue
            x, y = G.cr2xy(col, row, view.grid_spec)
            C.on_mouse_down(x, y)
            break
```
