### Spielsteine mit der Maus auswählen

```python
from ipywidgets import Output
from ipycanvas import MultiCanvas
import helpers as H
from IPython.display import display

board_spec = (20, 20, 20, 20, 8, 8)
points = H.get_midpoints(board_spec)

layout = {'border': '1px solid black'}
out = Output(layout=layout)
mcanvas = MultiCanvas(2, width=200, height=200, layout=layout)
bg, fg = mcanvas


@out.capture(clear_output=True)
def on_mouse_down(x, y):
    i = H.get_closest((x, y), points)
    col, row = H.xy2cr(x, y, board_spec)
    print(f'Klickposition: {(round(x), round(y))},'  # Strings werden zusammengefuegt
          f'Feld: ({col}, {row}), naechste Feldmitte {points[i]}')


H.draw_board(bg, board_spec)
for x, y in points:
    fg.fill_circle(x, y, 2)


mcanvas.on_mouse_down(on_mouse_down)
display(mcanvas, out)
```
