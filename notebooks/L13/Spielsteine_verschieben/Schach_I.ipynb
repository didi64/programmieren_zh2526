### Schach
Auf einen Schachbrett sollen Figuren mit der Maus bewegt werden
können. Vorerst lassen sich auch illegale Züge ausführen.

Die Spielposition wird als Tabelle (Liste mit Listen) mit
8 Zeilen und 8 Spalten gespeichert. Ein leeres Feld wird mit Space (`' '`)
makiert, weisse Figuren mit grossen Buchstaben
(B(auer), T(urm), S(pringer), L(äufer), D(ame), K(önig)), und schwarze Figuren mit den entsprechenden Kleinbuchstaben.

Im Modul `schach_0.py` wird eine 8x8 Tabelle `board` mit lauter leeren Feldern erstellt. Weiter sind folgende Funktionen definiert:

- `set_startpos()`: Modifiziert `board`, so dass die Startposition speichert.
- `get_field(col, row)`: Gibt `board[row][col]` zurück.
- `set_field(col, row, value)`: Setzt `board[row][col] = value`.
- `move(src, target)`: Bewegt die Figur von `src=(col0, row0)` nach `target=(col1, row1)`. Gibe ein Tuple mit den Änderungen der Position zurück, in der Form  
  ((neuer Eintrag, Spalte, Reihe), ...).

- `get_pieces()`: Gibt eine Liste mit den Figuren und ihren Positionen zurück, in der Form
  \[(Symbol, Spalte, Reihe), ...\].
- `ld2cr(notation)`: Konvertiert Schachnotation wie `a1` nach `(0, 7)` (
  Das Feld `a1` lieft in der 0-ten Spalte und er 7 Reihe.

- `hmove(src, target)`: (Human readable move): wie `move` aber
  `src` und `target` sind z.B. `e2`, `e4`'.
   
- `new_game()`: Ruft vorerst nur `set_startpos()` auf.

```python
import schach_0 as schach
import schach_darstellung as D
from ipycanvas import MultiCanvas
from time import sleep


layout = {'border': '1px solid black'}
mcanvas = MultiCanvas(2, width=200, height=200, layout=layout)
bg, fg = mcanvas

D.draw_chessboard(bg)
schach.new_game()
changes = schach.get_pieces()
D.apply_changes(fg, changes)

mcanvas
```

```python
moves = [('e2', 'e4'), ('e7', 'e5'), ('f1', 'c4'), ('f8', 'c5'),
         ('d1', 'h5'), ('g8', 'f6'), ('h5', 'f7')]

schach.new_game()
for move in moves:
    changes = schach.hmove(*move)
    D.update(fg, 'move', changes=changes)
    sleep(1)
```
