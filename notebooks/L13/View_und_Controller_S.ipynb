### View und Controller (simple)
Aufgabe der Controller-Komponente ist es,
Benutzerinput (Mausklicks und Tastendrücke) entgegenzunehmen, diese zu interpretieren und dann die passenden Funktionen der Game-Komponente aufzurufen.

```python
import game
from ipywidgets import Output


LAYOUT = {'border': '1px solid black'}
out = Output(layout=LAYOUT)


@out.capture(clear_output=True)
def foreward_cmd(key, game):
    print(f'key pressed: {key}')
    if key == 'ArrowUp':
        game.swap()
    if key == 'ArrowRight':
        game.rotate()
    if key == 'n':
        game.new_game()
```

```python
from IPython.display import display
from ipycanvas import MultiCanvas


mcanvas = MultiCanvas(2, width=200, height=200, layout=LAYOUT)
bg, fg = mcanvas


def draw_bg():
    bg.fill_text('Schiebespiel', 20, 20)


def on_mouse_down(key, *flags):
    foreward_cmd(key, game)


def update(event, state):
    print(event, state)
    fg.clear()
    fg.fill_text(f'state: {state}', 20, 50)


draw_bg()
mcanvas.on_key_down(on_mouse_down)
game.update = update

display(mcanvas, out)
```
