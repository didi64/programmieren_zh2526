### Currying

```python
n = 3


def f(x, y, z):
    print(f'x: {x}, y: {y}, z: {z}')


def f_z_fixed_1(x, y):
    f(x, y, n)


def f_z_fixed_2(x, y, z=n):
    f(x, y, z)


def f_z_fixed_3(x, y, z=n):
    f(x, y, n)
```

```python
f_z_fixed_1(1, 2)
f_z_fixed_2(1, 2)
f_z_fixed_3(1, 2)
```

```python
n = 42
f_z_fixed_1(1, 2)
f_z_fixed_2(1, 2)
f_z_fixed_3(1, 2)
```

```python
def draw_dot(x, y, canvas, color='blue', r=10):
    canvas.fill_style = color
    canvas.fill_circle(x, y, r)
```

```python
from ipycanvas import MultiCanvas


mcanvas = MultiCanvas(2, width=100, height=100,
                      layout={'border': '1px solid black'})
bg, fg = mcanvas
mcanvas
```

```python
draw_dot(50, 50, bg, color='grey', r=20)
draw_dot(50, 50, fg, color='yellow')
```

```python
def on_mouse_down(x, y, canvas=fg):
    draw_dot(x, y, canvas)


mcanvas.on_mouse_down(on_mouse_down)
```
