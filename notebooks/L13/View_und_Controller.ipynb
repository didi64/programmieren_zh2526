### View und Controller
Aufgabe der Controller-Komponente ist es,
Benutzerinput (Mausklicks und Tastendrücke) entgegenzunehmen, diese zu interpretieren und dann die passenden Funktionen der Game-Komponente aufzurufen.

```python
import game
import view_helpers
from ipywidgets import Output
from IPython.display import display
from ipycanvas import MultiCanvas


LAYOUT = {'border': '1px solid black'}
out = Output(layout=LAYOUT)


@out.capture(clear_output=True)
def foreward_cmd(key, game, mcanvas=None):
    print(f'key pressed: {key}')
    if key == 'ArrowUp':
        game.swap()
    if key == 'ArrowRight':
        game.rotate()
    if key == 'n':
        game.new_game()

    if key == 'd':
        view_helpers.change_theme(mcanvas, 'dark')
    if key == 'l':
        view_helpers.change_theme(mcanvas, 'light')


mcanvas = MultiCanvas(2, width=200, height=200, layout=LAYOUT)
bg, fg = mcanvas

view_helpers.draw_bg(bg)


def on_key_down(key, *flags, game=game, mcanvas=mcanvas):
    foreward_cmd(key, game, mcanvas)


def update(event, data, canvas=fg):
    view_helpers.update_fg(canvas, data)


mcanvas.on_key_down(on_key_down)
game.update = update

display(mcanvas, out)
```

```python
view_helpers.settings['theme'] = 'dark'
```
