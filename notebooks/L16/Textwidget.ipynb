### Text-Widget
Das Text-Widget `Text` erstellt ein Eingabefeld.  
Dabei löst die `on_submit` Methode eine Deprecation Warning aus, die sich mit 
```python
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
```
unterdrücken lässt (meine Empfehlung).

Offiziell wird empfohlen, das Text-Widget so zu verwenden:
```python
eingabe = Text(placeholder='deine Eingabe', description='Eingabe:')
eingabe.continuous_update = False
eingabe.observe(on_value_change, names="value")  
```
Das ruft `on_value_change(change)` auf, wenn `change.value` ändert.  
`change` ist dabei ein Dict-artiges Objekt, auf dessen Schlüssel auch mit der dot-Notation
zugegriffen werden kann.  
Das `change`-Objekt speichert folgende Schlüssel-Wert Paare:
```
{'name': 'value', 
 'old': <alter Wert>, 
 'new': <neuer Wert>, 
 'owner': <Text-Objekt>), 
 'type': 'change',
 }
```
**Nachteile**:
- `on_value_change` wird aufgerufen, wenn das Eingabefeld den Fokus verliert,
- eine leere Eingabe kann nicht gemacht werden,
- `eingabe.value = ''` ruft auch `on_value_change` auf.

***
Verwendung von `Text` mit `on_submit`:
***

````python
import warnings
from ipywidgets import Output, Text
from IPython.display import display


# Deprecation Warning unterdruecken
warnings.filterwarnings('ignore', category=DeprecationWarning)


out = Output(layout={'border': '1px solid black'})
eingabe = Text(placeholder='deine Eingabe', description='Eingabe:')


@out.capture(clear_output=True)
def on_submit(text):
    print(f'submitted text: "{text.value}"')
    text.value = ''


eingabe.on_submit(on_submit)
display(eingabe, out)
````

***
Verwendung von `Text` mit `observe`:
***

````python
from ipywidgets import Output, Text
from IPython.display import display

out = Output(layout={'border': '1px solid black'})
eingabe = Text(placeholder='deine Eingabe', description='Eingabe:')


@out.capture(clear_output=False)
def on_value_change(change):
    print(change)
    print(f'old: "{change.old}", new: "{change.new}"')
    print(f'change.owner.value: "{change.owner.value}"')
    print(f'eingabe.value: "{eingabe.value}"')
    print('-'*30)

    change.owner.value = ''  # triggers on_value_change!


eingabe.continuous_update = False
eingabe.observe(on_value_change, names="value")

display(eingabe, out)
````
