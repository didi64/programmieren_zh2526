### Pong mit Eventloop
Das Modul `eventloop` enthält folgende Attribute:
- `key_buffer = []`: Liste aus der die Tasten gelesen werden.
- `key_handler(key) = None`: Ist mit einer Funktion, welche die Tasten gehandelt zu überschreiben.
- `default_action = None`: Ist mit einer Funktion zu überschreiben.
- `delay = 1`
- `out = None`: Wird `out` mit einem Output-Widget überschreiben, wird der aktuelle Thread dort ausgegeben.
- `stop_event = threading.Event()`
  
Weiter enthält das Modul `eventloop` eine Funktion
- `start()`: leert `key_buffer` und setzt das `stop_event` auf `False`.

```python
import eventloop
import game
import darstellung as D
from ipywidgets import Output
from ipycanvas import MultiCanvas
from IPython.display import display


layout = {'border': '1px solid black'}
out = Output(layout=layout)
mcanvas = MultiCanvas(2, width=game.WIDTH, height=game.HEIGHT, layout=layout)
bg, fg = mcanvas


@out.capture(clear_output=True)
def on_key_down(key, *flags):
    # print(key)
    if key == 'n':
        game.new_game()
        eventloop.start()
    else:
        eventloop.key_buffer.insert(0, key)


def key_handler(key):
    if key == 'ArrowUp':
        game.move_pad(-5)
        game.move_ball()
    elif key == 'ArrowDown':
        game.move_pad(5)
        game.move_ball()

    D.draw_ball_and_pad(fg, game.state)


def default_action():
    game.move_ball()
    D.draw_ball_and_pad(fg, game.state)


D.draw_background(bg)
D.draw_ball_and_pad(fg, game.state)

mcanvas.on_key_down(on_key_down)

eventloop.out = out
eventloop.delay = 0.5
eventloop.key_handler = key_handler
eventloop.default_action = default_action


display(mcanvas, out)
mcanvas.focus()
```

```python
# stoppt den Eventloop
eventloop.stop_event.set()
```
