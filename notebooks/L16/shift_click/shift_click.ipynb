### Mausklick: mehrere Aktionen tätigen
Das Canvas-Widget löst das `on_mouse_down`-Event aus, wenn eine Maustaste gedrückt wird.
Leider wird dabei nicht zwischen den Maustasten unterschieden.
Um trotzdem verschiedene Aktionen tätigen zu können, 
kann z.B. unmittelbar vor dem Mausklick eine Taste gedrückt werden,
die angibt, was beim Mausklick passieren soll.


Nachstehend wird ein Kreis oder ein Dreick in ein Feld gezeichnet, je nach dem, ob vor dem Mausklick Shift gedrückt wurde.

```python
import darstellung as D
from ipycanvas import MultiCanvas
from ipywidgets import Output
from IPython.display import display
import canvas_helpers as H


state = {'flag': False}

layout = {'border': '1px solid black'}
out = Output(layout=layout)
BOARD_SPEC = (20, 20, 20, 20, 10, 10)


@out.capture(clear_output=True)
def on_mouse_down(x, y):
    col, row = H.xy2cr(x, y, BOARD_SPEC)
    print(f'clicked at ({col}, {row})')
    if state['flag']:
        D.draw_flag(bg, BOARD_SPEC, (col, row))
    else:
        D.draw_mine(bg, BOARD_SPEC, (col, row))


@out.capture()
def on_mouse_up(x, y):
    state['flag'] = False
    print(f'flag: {state['flag']}')


@out.capture()
def on_key_down(key, *flags):
    if key == 'Shift':
        state['flag'] = True
    else:
        state['flag'] = False
    print(f'flag: {state['flag']}')


mcanvas = MultiCanvas(2, width=240, height=240, layout=layout)
bg, fg = mcanvas


mcanvas.on_mouse_down(on_mouse_down)
mcanvas.on_mouse_up(on_mouse_up)
mcanvas.on_key_down(on_key_down)

H.draw_grid(fg, BOARD_SPEC)
display(mcanvas, out)
```
