### TikTakToe Battle
Eine TikTakToe-Strategie ist eine Funktion
```python
def find_best_move(s):
    '''s: str der Laenge 9 der Form "X..OX.O..", welcher
       folgende Spielsituation repraesentiert:
       X..
       OX.
       O..

       Zurueckgegeben wird eine Zahl in (0,...,8), die Position
       an welcher der Spieler am Zug seinen Stein platzieren soll.
    '''
```

Nachstehende Funktion spielt falls möglich in der Mitte, in einer Ecke oder Seite (in dieser Reihenfolge)
auf ein freies Feld.

```python
def dummy_strategy(s):
    options = (4, 0, 2, 6, 8, 1, 3, 5, 7)  # center, corners, sides
    return [i for i in options if s[i] == '.']
```

In Module `tiktak_battle` sind weiter folgende Strategien implementiert:
- `wta_strategy`: spielt wenn möglich einen Winner, pariert falls nötig eine Drohung und
  stellt anderfalls, falls möglich, eine Drohnung auf.
- `default_stragety`: spielt den besten Zug.


In Module `tiktak_battle` hat eine Funktion `battle(f, g=None, n=10)`, welche die
Strategien `n`-Mal gegeneiander antreten lässt und die 
gewonnenen und verlorenen Partien zurück gibt.
`f` beginnt. `None` wird durch die Default-Strategy ersetzt.

Die Klasse `Gameviewer` aus dem Modul `tiktakviewer` erlaubt, eine Liste mit
Partien nachzuspielen.
Mit ArrowUp und ArrowDown lässt sich eine Partie auswählen, und mit
ArrowRight und ArrowLeft die Partie nachspielen.

````python
import tiktak_battle as B
from tiktakviewer import GameViewer
````

````python
def show(s):
    print(f'{s[:3]}\n{s[3:6]}\n{s[6:]}')
````

````python
s = 'X..OX.O..'
show(s)
````

````python
B.dummy_strategy(s)
````

````python
B.wta_strategy(s)
````

````python
wins, losses = B.battle(B.dummy_strategy, B.wta_strategy)
wins, losses
````

````python
# beginnt die wta_strategy, haelt sie immer ein Unentschieden
wins, losses = B.battle(B.wta_strategy)
wins, losses
````

````python
# wta_strategy im Nachzug gegen die Default-Strategy
# nochmals ausfuehren, falls keine Partien mit Resultat
wins, losses = B.battle(None, B.wta_strategy)
wins, losses
````

````python
# vom ersten Spieler gewonnene Partien anschauen
gv = GameViewer(wins)
gv
````
