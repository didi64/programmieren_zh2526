**Neu** sind folgende Ergänzungen:
- DocString für ein Modul: Ist der erste Ausdruck in einem Modul `foo.py` ein DocString der Form
  ```python
  '''
  Dies ist ein DocString fuer ein Modul
  '''
  ```
  wird dieser angezeigt, wenn **nach erfolgtem Import**
  ```python
  help(foo)
  ```
  ausgeführt wird.
  Diese Funktion gibt zudem alle im Modul definierten Funktionen in alphabetischer Reihnefolge mit Signatur und DocString aus, ebenso die im Modul definierten Variabeln.
- Tests zu einem Modul hinzufügen, bez. das Pattern
  ```python
  if __name__ == '__main__':
      # do stuff (z.B. teste Funktionen dieses Moduls)
  ``` 
  &nbsp;  
  Wird ein Modul `foo.py` importiert, wird der (globalen) Variable `__name__` dieses Moduls der String `'foo'` zugewiesen.
  Wird das File/Modul dagegen (z.B. mit der Zeilen-Magic `%run [-m] foo`) ausgeführt, so wird
  `__name__` der String `'__main__'` zugewiesen.
  - `%run foo` führt das File `foo.py` **im aktuellen Verzeichnis** aus,
  - `%run -m foo` sucht in den Ordnern in der Liste `sys.path` nach `foo.py`.
