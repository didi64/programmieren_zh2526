### Bäume (Trees)

`Tree ::= (value: any, children: List[Tree])`

In Baum ist ein Tuple `(value, children)` der Länge 2, wobei
`value` irgend ein Python-Objekt und `children` eine 
möglicherweise leere Liste von Bäumen ist.

```python
def tree(value, children=None):
    if children is None:
        children = []
    return (value, children)
```

```python
def level_by_level1(tree):
    d = {}
    stack = [(0, tree)]
    while stack:
        depth, node = stack.pop(0)
        d.setdefault(depth, []).append(node[0])
        for child in node[1]:
            stack.append((depth+1, child))
    return d
```

```python
def preorder(tree, depth=0, indent='    '):
    print(f'{indent*depth}value: {tree[0]}')
    for child in tree[1]:
        preorder(child, depth+1)
```

```python
def preorder1(tree, depth=0, indent='    ', store=None):
    if store is None:
        store = []
    store.append(tree[0])
    for child in tree[1]:
        preorder1(child, depth+1, store=store)
    return store
```

```python
def print_tree(tree, code=''):
    value = tree[0]
    if code == '':
        print(f'@ {value}')
    else:
        prefix = ''.join(f'{c}   ' for c in code[:-1])
        print(f'{prefix}|')
        print(f'{prefix}+---@ {value}')

    if (children := tree[1]):
        *firsts, last = children
        for child in firsts:
            print_tree(child, code + '|')
        print_tree(last, code + ' ')
```

```python
def is_save(queens, row0, col0):
    '''queens: str of length col
       returns False if a queen at (col, row)  attacks a queen at
       (0, queens[0]), ..., (col-1, queens[col-1])
    '''
    for col, row in enumerate(queens):
        row = int(row)
        in_same_row = row == row0
        in_same_diag = col0-col == abs(row0-row)
        if in_same_row or in_same_diag:
            return False
    return True
```

```python
def build_tree(value):
    children = []
    if len(value) < 8:
        col = len(value)
        children = [build_tree(f'{value}{row}')
                    for row in range(8) if is_save(value, row, col)
                    ]
    # else:
    #     print(value)
    return (value, children)
```

```python
def keep_longpaths_only(tree, n):
    if n == 0:
        return tree
    children = [t for child in tree[1] if (t := keep_longpaths_only(child, n-1))]
    if children:
        return (tree[0], children)
```
