```python
import random


N = 5


def get_mines():
    mines = ['0'] * N*N
    for i in random.sample(range(N*N), N):
        mines[i] = '1'
    return ''.join(mines)


def game_over():
    print('game over!')


def you_win():
    print('congrats!')


def reveal_cell(i):
    if mines[i]:
        game_over()
    else:
        save_fields.append(i)
        if len(save_fields) == N*(N-1):
            you_win()


def count_neighbors(i):
    count = 0
    x0, y0 = divmod(i, N)
    for di in (-(N+1), -N, -(N-1), -1, 1, N-1, N, N+1):
        j = i+di
        if not (0 <= j < N*N):
            continue
        x, y = divmod(j, N)
        count += max(abs(x-x0), abs(y-y0)) <= 1 and mines[j] == '1'
    return count


def print_mines():
    print('\n'.join(mines[i*N:(i+1)*N] for i in range(N)))


mines = get_mines()
save_fields = []
```
