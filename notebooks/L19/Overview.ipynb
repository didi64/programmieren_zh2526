### Überblick Lösungsalgorithmen/Programmieren (Teil 2)
Der Fokus in diesem 2. Teil der Vorlesung liegt auf Softwareentwicklung.
Wie plant, entwickelt und testet man etwas grössere Programme.  
Wie zerlegt man ein Programm in geeignete, möglichst unabhängige und separate testbare Komponenten (Funktionen, Module/Files, Klassen). 


Wir werden auch eine Anzahl eigener Module erstellen,
um bereits geschriebenen und getesteten Code wiederverwenden  zu können. 


**Einige Themen**:

- Module, Module testen.
- Klassen, objektorientiertes Programmieren
- Fehler auslösen und behandeln
- dynamische Datenstrukturen (Bäume, ...)
- Rekursion
- Generators, Decorators,
- Asynchrones Programmieren mit `asyncio`.  
  Mehrere Aufgaben gleichzeitig (concurrent) ausführen.  
- Desing-Pattern für einfache Applikationen: Model-View-Controller (MCV)
- Software Design Prinzipien: [SOLID](https://de.wikipedia.org/wiki/Prinzipien_objektorientierten_Designs)

**Benotung**:
- 2 kleinere Projektarbeiten mit mündlicher Prüfung (total 4 Noten)
- Bonus für bearbeitete und abgegebene Übungsaufgaben.
