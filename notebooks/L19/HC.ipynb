```python
class InstanceDict(dict):
    def __getitem__(self, key):
        if key in self:
            return super().__getitem__(key)

        elif key in self['__class__']:
            attr = self['__class__'][key]
            if not callable(attr):
                return attr
            else:
                return lambda *args, **kwargs: attr(self, *args, **kwargs)

        else:
            raise AttributeError(f'Attribute {key} not found!')
```

```python
class ClassDict(dict):
    def __call__(self, *args, **kwargs):
        d = InstanceDict({'__class__': self})

        __init__ = self.get('__init__')
        if __init__:
            __init__(d, *args, **kwargs)

        return d

    def __repr__(self):
        return f'Class {id(self)}'
```

```python
Person = ClassDict()
Person
```

```python
def __init__(self, name, age):
    self['name'] = name
    self['age'] = age


def birthday(self):
    self['age'] += 1


def show(self):
    print(f'Person(name={self['name']}, age={self['age']}')
```

```python
Person['__init__'] = __init__
Person['birthday'] = birthday
```

```python
P = {}
P['__init__'] = __init__
P['birthday'] = birthday
```

```python
def make_instance(P, *args, **kwargs):
    d = {'__class__': P} 
    __init__ = P.get('__init__')
    if __init__:
        __init__(d, *args, **kwargs)

    return d


def get_attr(d, key):
    assert '__class__' in d, 'dict is not an insance of a class!'

    if key in d:
        return d[key]

    elif key in d['__class__']:
        attr = d['__class__'][key]
        if not callable(attr):
            return attr
        else:
            return lambda *args, **kwargs: attr(d, *args, **kwargs)

    else:
        raise AttributeError(f'Attribute {key} not found!')
```

```python
alice = make_instance(P, 'Aice', 18)
bob = make_instance(P, 'Bob', 20)
```
