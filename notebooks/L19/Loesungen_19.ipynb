**100 Lampen Problem**  
Jede der 100 Lampen 1,2,...,100 wird mit einem Lichtschalter 1,2,...,100 ein, bez. ausgeschaltet.  
Zu Beginn sind alle Lampen aus.
Nun kommen 100 Leute und schalten Lampen ein und aus.  
- Die 1. Person legt alle Lichtschalter um und schaltet alle Lampen ein.
- Die 2. Person legt jeden 2.ten Lichtschalter um und schaltet die Lampen 2, 4, 8, ..., 98, 100 wieder aus.
- Die n-te Person legt jeden n-ten Lichtschalter um, den n-ten, (2n)-ten, (3n)-ten, ...

Welche Lampen brennen, nachdem alle 100 Personen Lichtschalter umgelegt haben

**Aufgabe**: Löse obiges Problem zuerst für N Lampen und N Personen für N = 10.
- Modelliere die N Lampen als Liste `are_on = [False] * N`.
- Schreibe eine Funktion `toggle(k, items)`, welche die Wahrheitswerte mit Indices 
  `k - 1`, `2k - 1`, `3k - 1`, ... in der Liste `items` negiert. 
- Erstelle eine Liste mit den Indices der Lampen, die am Ende brennen:  
  Führe `toggle(1, are_on)`, ..., `toggle(N, are_on)` aus.  
  Beginne mit dem Zählen bei 1.

````python
N = 10
are_on = [False] * N
are_on
````

````python
def toggle(k, items):
    i = k - 1
    while i < len(items):
        items[i] = not items[i]
        i += k
````

````python
N = 100
are_on = [False] * N

for i in range(N):
    toggle(i+1, are_on)

[i+1 for i, is_on in enumerate(are_on) if is_on]
````

````python
def fill_triangle(canvas, pos, grid_spec, anchor=0, color='black'):
    '''verbinde 3 Ecken des Gitterfeldes pos=(col, row) zu einem Dreieck.
       Der rechte Winkel ist bei der Ecke anchor
       (die linke obere Ecke ist Ecke 0, aufgezaehlt wir im Gegenuhrzeigersinn).
    '''
    pts = [(0, 0), (1, 0), (1, 1), (0, 1)]
    pts.pop((anchor+2) % 4)
    G.fill_polygon(canvas, pos, pts, grid_spec, color=color)
````
