```python
from my_vec2d import Vec
from ipycanvas import MultiCanvas


def get_mcanvas(n=2, width=200, height=200):
    layout = {'border': '1px solid black'}
    return MultiCanvas(n, width=width, height=height, layout=layout)
```

```python
def draw_koch(canvas, p, q, n):
    if n == 0:
        canvas.stroke_line(p.x, p.y, q.x, q.y)
        return

    pq = q-p
    pt1 = p + (1/3)*pq
    pt2 = p + (2/3)*pq
    pt = (1/2)*(p+q) - (1/(2*3**.5))*pq.perp()

    draw_koch(canvas, p, pt1, n-1)
    draw_koch(canvas, pt1, pt, n-1)
    draw_koch(canvas, pt, pt2, n-1)
    draw_koch(canvas, pt2, q, n-1)
```

```python
mcanvas = get_mcanvas()
bg, fg = mcanvas
mcanvas
```

```python
bg.clear()

A = Vec(10, 100)
B = Vec(190, 100)
draw_koch(bg, A, B, 4)
```

```python
def koch_curve(canvas):
    size = max(canvas.width, canvas.height)
    center = Vec(size/2, size/2)
    v = Vec(0, -0.9*size/2)
    pts = [center + v.rotate(i*120) for i in range(4)]
    for i in range(3):
        draw_koch(bg, pts[i], pts[i+1], n)
```

```python
def sirpinsky(canvas, A, B, C, n):
    if n == 0:
        canvas.fill_polygon([(A.x, A.y), (B.x, B.y), (C.x, C.y)])
        return

    sirpinsky(canvas, A, A+(B-A)/2, A+(C-A)/2, n-1)
    sirpinsky(canvas, B, B+(C-B)/2, B+(A-B)/2, n-1)
    sirpinsky(canvas, C, C+(A-C)/2, C+(B-C)/2, n-1)
```

```python
mcanvas = get_mcanvas()
bg, fg = mcanvas
bg.sync_image_data = True
mcanvas
```

```python
bg.clear()
center = Vec(100, 100)
v = Vec(0, -90)
A = center + v
B = center + v.rotate(120)
C = center + v.rotate(-120)

sirpinsky(bg, A, B, C, 6)
```

```python
def divide_and_conquor(target, items, lower=0, upper=None, attempt=0):
    if upper is None:
        upper = len(items)

    i = (lower + upper) // 2
    if (item := items[i]) == target:
        return i, attempt
    elif target > item:
        return divide_and_conquor(target, items, i+1, upper, attempt+1)
    else:
        return divide_and_conquor(target, items, lower, i, attempt+1)
```

```python
items = list(range(1000))
target = 6
divide_and_conquor(target, items)
```

```python
def divide_and_conquor(target, items):

    def _divide_and_conquor(target, items, lower, upper, attempt):
        if lower == upper:
            return False, attempt

        i = (lower + upper) // 2
        if (item := items[i]) == target:
            return i, attempt
        elif target > item:
            return _divide_and_conquor(target, items, i+1, upper, attempt+1)
        else:
            return _divide_and_conquor(target, items, lower, i, attempt+1)

    if not items or target < items[0] or target > items[-1]:
        return False, 0
    else:
        return _divide_and_conquor(target, items, 0, len(items), 0)
```

```python
items = list(range(1000))
target = 37
divide_and_conquor(target, items)
```
