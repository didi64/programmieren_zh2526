```python
from ipycanvas import MultiCanvas


def get_mcanvas(width=200, height=200):
    layout = {'border': '1px solid black'}
    return MultiCanvas(2, width=width, height=height, layout=layout)
```

```python
import math


def sign(x):
    return (x > 0) - (x < 0)


class Vec:

    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __neg__(self):
        return Vec(-self.x, -self.y)

    def __rmul__(self, s):
        return Vec(s * self.x, s * self.y)

    def __truediv__(self, s):
        return Vec(self.x / s, self.y / s)

    def __mul__(self, other):
        if isinstance(other, int):
            return other*self
        elif isinstance(other, Vec):
            return self.x * other.x + self.y * other.y

    def __add__(self, w):
        return Vec(self.x + w.x, self.y + w.y)

    def __sub__(self, w):
        return Vec(self.x - w.x, self.y - w.y)

    def perp(self):
        return Vec(-self.y, self.x)

    def norm(self):
        return (self.x**2 + self.y**2)**.5

    def area(self, other):
        return self.x*other.y - self.y*other.x


    def angle(self, other):
        cos = (self*other) / (self.norm()*other.norm())
        angle = math.acos(cos) * 180 / math.pi
        return sign(self.area(other))*angle

    def rot(self, alpha):
        alpha *= math.pi/180
        return Vec(self.x * math.cos(alpha) - self.y * math.sin(alpha),
                   self.x * math.sin(alpha) + self.y * math.cos(alpha))

    def __repr__(self):
        return f'Vec({self.x}, {self.y})'
```

```python
def draw_koch(canvas, p, q, n):
    if n == 0:
        canvas.stroke_line(p.x, p.y, q.x, q.y)
        return

    pq = q-p
    pt1 = p + (1/3)*pq
    pt2 = p + (2/3)*pq
    pt = (1/2)*(p+q) - (1/(2*3**.5))*pq.perp()

    draw_koch(canvas, p, pt1, n-1)
    draw_koch(canvas, pt1, pt, n-1)
    draw_koch(canvas, pt, pt2, n-1)
    draw_koch(canvas, pt2, q, n-1)
```

```python
center = Vec(100, 100)
v = Vec(0, -90)
A = center + v
B = center + v.rot(120)
C = center + v.rot(-120)

A, B, C
```

```python
pts = (center, A, B, C)
colors = ('black', 'red', 'blue', 'orange')

bg.stroke_circle(center.x, center.y, v.norm())
for pt, color in zip(pts, colors):
    bg.fill_style = color
    bg.fill_circle(pt.x, pt.y, 5)
```

```python
n = 4
draw_koch(bg, A, B, n)
draw_koch(bg, B, C, n)
draw_koch(bg, C, A, n)
```

```python
def koch_curve(canvas):
    size = max(canvas.width, canvas.height)
    center = Vec(size/2, size/2)
    v = Vec(0, -0.9*size/2)
    pts = [center + v.rot(i*120) for i in range(4)]
    for i in range(3):
        draw_koch(bg, pts[i], pts[i+1], n)
```
