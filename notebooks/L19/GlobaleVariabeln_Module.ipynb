```python
 Each of these return a dictionary:

    globals() always returns the dictionary of the module namespace
    locals() always returns a dictionary of the current namespace
    vars() returns either a dictionary of the current namespace (if called with no argument) 
            or the dictionary of the argument.
dir() returns the names in the module namespace
```

```python
def f(x, y, z=42):
    print([k for k in globals() if not k.startswith('__')])
    print(f'lokales Variabelverzeichnis von f: {locals()}')
    u = 'foo'
    v = 'bar'
    
    print(f'lokales Variabelverzeichnis von f: {locals()}')
```
