### Die Operatoren `and` und `or`, Rechnen mit Booleans

**Rechnen mit Booleans**:
Was liefert die Funktion `f(x)`, vorausgesetzt das Argument `x` ist eine Zahl?
```python
def f(x):
    return (x > 0) - (x < 0)
```

Wertet man die Klammern aus, so erhĂ¤lt man Wahrheitswerte (booleans), also entweder `True` oder `False`.
`True` oder `False` verhalten sich beim Rechnen wie die Integer `1` und `0`.
Das Resultat einer Addition oder Subraktion von Booleans ist also ein Integer.

**Alle FĂ¤lle**:
- x > 0: True - False = 1, 
- x = 0: False - False = 0,
- x < 0: False - True = -1

Obige Funktion `f` liefert also das Vorzeichen einer Zahl, und trĂ¤gt 
ĂĽblicherweise den Namen `sign`.  
***
**Wie verhalten sich die Operatoren `and` und `or` auf beliebigen Objekten**?  
Zur Erinnerung: ein Wert `x` nennen wir *truthy*/*wahr* bez. *falsy*/*falsch*,  
falls `bool(x)` gleich `True`, bez. `False` ist.  
Nur `0`, `0.0`, `None`, `False` und leere Iterables (`''`, `()`, `[]`, `{}`, ...) sind *falsy*.  

`x and y` ist gleich
- `x`, falls `x` *falsy* ist (das 2. Argument `y` wird nicht ausgewertet)
- `y`, falls `x` *truthy* ist.
  
**Beispiele**:
- `False and False` und `False and True` ist `False`,
- `True and False` ist `False` und  `True and True` ist `True`.
  

`x or y` ist gleich
- `x`, falls `x` *truthy* ist (das 2. Argument `y` wird nicht ausgewertet)
- `y`, falls `x` *falsy* ist.

**Beispiele**:
- `True or False` ist `True` und  `True or True` ist `True`,
- `False or False` ist `False` und  `False and True` ist `True`.

***

**Typische Verwendungen**:
- Teste, ob der letzte Wert einer Liste einen gewissen Wert hat:
  Ist die Liste `items` leer, dann ist `items and items[-1]` gleich `[]` (*falsy*). Dabei wird `items[-1]` nicht ausgewertet (was einen Fehler verursachen wĂĽrde).
  ```python
    items = [3, 0, 2, 2]
    while items and items[-1] == 2:  # ist die Liste leer,
        print(items.pop())
  ```

- `None`-Wert eines Default-Arguments ĂĽberschreiben.
    ```python
    def f(items=None):
        items = items or []
        # if items is None:
        #    items = []
        print(items)
    ```

**Caveats**:
Die Zuweisung `value = value or default_value` ĂĽberschreibt `value` nicht nur, falls `value` den Wert `None` hat,
sondern auch in den doch relativ vielen FĂ¤llen, wo `value` *falsy* ist (siehe unten). 
Manchmal ist dies das gewĂĽnschte Verhalten und manchmal nicht.

````python
def sign(x):
    return (x > 0) - (x < 0)
````

````python
# sign-Funktion testen
for x in (2, 0, -3):
    print(f'sign({x}) = {sign(x)}')
````

````python
False and (1 + 'x')  # liefert kein Fehler, 1 + 'x' wird nicht ausgewertet
````

````python
True or (1 + 'x')  # liefert kein Fehler
````

````python
items = [3, 0, 2, 2]
while items and items[-1] == 2:
    print(items.pop())
````

````python
def f(items=None):
    items = items or []
    print(items)
````

**Caveats**:

````python
def g(value=None):
    default_value = 42
    value = value or default_value
    print(value)
````

````python
for value in [0, 0.0, '', None, False, (), [], {}, set()]:
    g(value)
````
