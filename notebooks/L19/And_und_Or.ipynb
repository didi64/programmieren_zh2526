### Repetition

**Rechnen mit Booleans**:
Was liefert die Funktion `f(x)`, vorausgesetzt das Argument `x` ist eine Zahl?
```python
def f(x):
    return (x > 0) - (x < 0)
```

Wertet man die Klammern aus, so erhält man Wahrheitswerte (booleans), also entweder `True` oder `False`.
`True` oder `False` verhalten sich beim Rechnen wie die Integer `1` und `0`.
Das Resultat einer Addition oder Subraktion von Booleans ist also ein Integer.

**Alle Fälle**:
- x > 0: True - False = 1, 
- x == 0: False - False = 0,
- x < 0: False - True = -1

Die Funktion `f` liefert also das Vorzeichen einer Zahl, und wird 
üblicherweise sign-Funktion genannt.  
***
**Was verhalten sich die Operatoren `and` und `or` auf beliebigen Objekten**?

`x and y` ist gleich
- `x`, falls `bool(x)` `False` ist,
- `y`, falls `bool(x)`, `True` ist.
  
**Beispiele**:
- `False and False` und `False and True` ist `False` (das 2. Argument wird nicht ausgewertet),
- `True and False` ist `False` und  `True and True` ist `True`.
  

`x or y` ist gleich
- `x`, falls `bool(x)` `True` ist,
- `y`, falls `bool(x)`, `False` ist.

**Beispiele**:
- `True or False` ist `True` und  `True or True` ist `True` (das 2. Argument wird nicht ausgewertet),
- `False or False` ist `False` und  `False and True` ist `True`.

````python
def sign(x):
    return (x > 0) - (x < 0)
````

````python
# sign-Funktion testen
for x in (2, 0, -3):
    print(f'sign({x}) = {sign(x)}')
````

````python
False and 1 + 'x'  # liefert kein Fehler, 1 + 'x' wird nicht ausgewertet
````

````python
True or 1 + 'x'  # liefert kein Fehler
````

````python
def f(items=None, new_item=None):
    items = items or [1, 2, 3]
    new_item = new_item or 4
    items.append(new_item)
    print(f'items: {items}')
````

````python
f()
````

````python
def g(items=[1, 2, 3], new_item=4):
    items.append(new_item)
    print(f'items: {items}')
````

````python
g()  # fuehre die Zelle mehrmals aus
````

**Erklärung**:
Sobald die Funktionsdefinition
```python
def g(items=[1, 2, 3], new_item=4):
```
ausgeführt wurde, speichert die lokale Variable `items` (eine Referenz auf) die Liste `[1, 2, 3]`.
Wird diese Liste später modifiziert, so 
speicher die lokale Variable `items` immer noch diese gleiche Referenz auf die nun modifizierte Liste.

Es gilt als schlecher Stiel, Default-Argumenten modifizierbare (mutable) Objekte als Default-Wert zuzuweisen.


Um einem Default-Argument die Liste `[1, 2, 3]` zuzuweisen, wird stattdessen folgendes Pattern verwendent:
```python
def h(items=None):
    if items is None:
        items = [1, 2, 3]
    ...
```

**Achtung**: Was ist problemtisch an folgendem Pattern?

```python
def h1(items=None):
    items = items or [1, 2, 3]  # rechte Seite hat den Wert [1, 2, 3] falls 
    ...
```

````python
def h(items=None, new_item=None):
    items = items or [1, 2, 3]
    new_item = new_item or 4
    items.append(new_item)
    print(f'items: {items}')
````

````python
h([3, 4], 5)
````

````python
h([])
````

````python
h([3, 2, 1], 0)
````
