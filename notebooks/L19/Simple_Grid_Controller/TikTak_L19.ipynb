### TikTakToe
- Der Spielzustand wird in einer  Liste `board` der Länge 9 gespeichert.
  Leere Felder sind mit `'.'` markiert, ein `'X'` oder `'O'` ist ein Stein des 1. bez. 2. Spielers.  
- Die Liste `win_lines` enthält Triple von Indices mit allen möglichen TikTaks.  
  `(0, 1, 2)` steht für ein TikTak in der 1. Zeile. Ist  
  `board[0] == board[1] == board[2] == 'X'`,
  dann hat der Spieler `X` dort ein TikTak.
- Die globale Variable `result` speichert das Resultat, einer der Werte `None`, `'X'`, `'O'`, oder `'draw'`.

Die Funktion `update(event, **kwargs)` gibt aus, was sich am Spielzustand ändert.
Die Funktion `update` wird von der Darstellungskomponente überschreiben.
Die Idee ist, dass die übergebenen Argumente genug Informationen liefern um die
Darstellung bequem anpassen zu können.  
Oft hat die Darstellungskomponente auch Zugriff auf alle Variabeln der Spielkomponente.  

Der Programmierer der Spielkomponente muss
dokummntieren, wann und mit welchen Argumenten `update` aufgerufen wird! 

- `new-game()`:
  - `update('new_game')`
- `check_result(player)`:
  - `update('winner', win_line=<Positionen pos in der win-line>, winner=<'X' oder 'O'>)`
  - `update('draw')`
- `play(pos)`:
  - `update('play', player=<'X' oder 'O'>, pos=<Position>)`

````python
EMPTY = '.'
players = 'OX'
board = [EMPTY] * 9
win_lines = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
             (0, 3, 6), (1, 4, 7), (2, 5, 8),
             (0, 4, 8), (2, 4, 6),
             ]
result = None


def set_result(value):
    global result
    result = value


def pos2idx(pos):
    return 3*pos[1] + pos[0]


def idx2pos(i):
    return divmod(i, 3)[::-1]


def get_player():
    i = board.count(EMPTY) % 2
    return players[i]


def get_win_line(player):
    for i, j, k in win_lines:
        if board[i] == board[j] == board[k] == player:
            return i, j, k


def update(event, **kwargs):
    print(event, kwargs)


def new_game():
    board[:] = [EMPTY] * 9
    set_result(None)
    update('new_game')


def check_result(player):
    if (win_line := get_win_line(player)):
        set_result(player)
        update('winner', win_line=[idx2pos(i) for i in win_line], winner=player)
    elif board.count(EMPTY) == 0:
        set_result('draw')
        update('draw')


def is_inside(pos):
    col, row = pos
    return 0 <= col < 3 and 0 <= row < 3


def play(pos):
    '''pos: (col, row) tuple'''
    if result or not is_inside(pos):
        return
    player = get_player()
    i = pos2idx(pos)
    board[i] = player
    update('play', player=player, pos=pos)
    check_result(player)


def show():
    s = ''.join(board[:9])
    print('\n'.join((s[:3], s[3:6], s[6:])))
````

````python
new_game()
show()
````

````python
new_game()
for i in range(3):
    play(idx2pos(i))
    play(idx2pos(i+3))
show()
````
