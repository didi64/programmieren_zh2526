### Konfigurierbarer Grid-Controller
Bei vielen Spielen werden Steine auf einem Brett/Gitter verschoben.  
Die Controllerkomponente sieht in diesen Fällen immer ähnlich aus.  

Ziel ist es, eine konfigurierbare Controllerkomponente zu schreiben.
Ein erster Ansatz ist `grid_controller.py`. Dieses Modul hat eine Funktion  
`init(canvas, grid_spec, callbacks)`, wobei
- `canvas`: Widget, das auf die Mausklicks hört,
- `grid_spec`: spezifiziert das Gitter,
- `callbacks`: Dict der Form
    ```python
    {'mouse_down': game.play),
     'n': game.new_game),
     }
    ```
    Schlüssel ist ein Mausevent oder ein Tastenname.
    Der Wert ist die Funktion, die in diesem Fall aufgerufen wird.  
    Ist der Schlüssel ein Mausevent, dann wird der Funktion
    die Argument `pos` übergeben. `pos` ist ein Tuple mit der Gitterspalte und Zeile, in die geklickt wurde (kann ausserhalb es Gitters liegen). 

Nachstehend wird die Spielkomponente als `game` und die Controllerkomponente als `C` importiert.  
Die Darstellungskomponente ist die Funktion `update`.

````python
import widget_helpers as W
import grid_helpers as G
import tiktak_L19 as game
import grid_controller as C


canvas = W.get_canvas()


def update(event, **kwargs):
    canvas.save()
    if event == 'new_game':
        canvas.clear()
        G.draw_grid(canvas, grid_spec, color='blue')
    if event == 'play':
        G.fill_text(canvas, kwargs['player'], kwargs['pos'], grid_spec)
    if event == 'winner':
        canvas.global_alpha = 0.3
        for pos in kwargs['win_line']:
            G.fill_rect(canvas, pos, grid_spec, color='red')
    else:
        print(event, kwargs)
    canvas.restore()


game.update = update
````

````python
grid_spec = (20, 20, 20, 20, 3, 3)
callbacks = {'mouse_down': game.play,
             'n': game.new_game,
             }

C.init(canvas, grid_spec, callbacks)
C.show(grid_color='blue', debug=True)
````
