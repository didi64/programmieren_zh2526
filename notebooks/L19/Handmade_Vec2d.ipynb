### Handgestrickte Vektorklasse

Unterschiede zu richtigen Klassen betreffen die Nutzung:
- Instanziierung: `v =  make_instance(Vec, 1, 2)` statt `v = Vec(2, 3)`
- Methodenaufruf: `get_attr('add')` statt `v.add`

```python
def __init__(self, x=0, y=0):
    self['x'] = x
    self['y'] = y


def show(self):
    print(f'Vec({self['x']}, {self['y']})')


def reverse(self):
    return make_instance(self['__class__'], self['x'], self['y'])


def add(self, other):
    return make_instance(self['__class__'], self['x']+other['x'], self['y']+other['y'])


def scale(self, factor):
    return make_instance(self['__class__'], factor*self['x'], factor*self['y'])


def scalar_product(self, other):
    return self['x']*other['x'] + self['y']*other['y']


def norm(self):
    return (self['x']**2 + self['y']**2) **0.5


def perp(self):
    return make_instance(self['__class__'], -self['y'], self['x'])


Vec = {
    '__name__': 'Vec',
    '__init__': __init__,
    'show': show,
    'reverse': reverse,
    'add': add,
    'scale': scale,
    'scalar_product': scalar_product,
    'norm': norm,
    'perp': perp,
}
```

```python
from handmade_class import make_instance, get_attr


v = make_instance(Vec, 1, 2)
get_attr(v, 'show')()
```

```python
# w = v.reverse()
w = get_attr(v, 'reverse')()
get_attr(w, 'show')()
```

```python
# w = v.add(v)
w = get_attr(v, 'add')(v)
get_attr(w, 'show')()
```

```python
# w = v.scale(10)
w = get_attr(v, 'scale')(10)
get_attr(w, 'show')()
```

```python
# u =  w.perp()
u = get_attr(w, 'perp')()
get_attr(u, 'show')()
```

```python
# w.scalar_product(u)
get_attr(w, 'scalar_product')(u)
```

```python
# Vec.scalar_product(w, u)
Vec['scalar_product'](w, u)
```

```python
# w.norm()
get_attr(w, 'norm')()
```
