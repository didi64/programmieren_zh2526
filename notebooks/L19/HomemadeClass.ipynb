```python
def make_instance(P, *args, **kwargs):
    d = {'__class__': P} 
    __init__ = P.get('__init__')
    if __init__:
        __init__(d, *args, **kwargs)

    return d


def get_attr(d, key):
    assert '__class__' in d, 'dict is not an insance of a class!'

    if key in d:
        return d[key]

    elif key in d['__class__']:
        attr = d['__class__'][key]
        if not callable(attr):
            return attr
        else:
            return lambda *args, **kwargs: attr(d, *args, **kwargs)

    else:
        raise AttributeError(f'Attribute {key} not found!')
```

```python
def __init__(self, name, age):
    self['name'] = name
    self['age'] = age


def birthday(self):
    self['age'] += 1


def show(self):
    print(f'Person(name={self['name']}, age={self['age']})')
```

```python
Person = {
    '__init__': __init__,
    'birthday': birthday,
    'show': show,
}
```

```python
alice = make_instance(Person, 'Aice', 18)
bob = make_instance(Person, 'Bob', 20)
```

```python
get_attr(bob, 'show')()
```

```python
get_attr(alice, 'age')
```

```python
get_attr(bob, 'birthday')()
get_attr(bob, 'show')()
```
