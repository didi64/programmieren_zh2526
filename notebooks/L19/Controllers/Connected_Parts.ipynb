```python
import grid_helpers as G
import big_view


N = 5
cells = [[None]*N for _ in range(N)]
color_key = 'k'
colors = {'r': 'red',
          'b': 'blue',
          'g': 'green',
          'y': 'yellow',
          'o': 'orange',
          'k': 'black',
          }


def set_color_key(c):
    global color_key
    color_key = c


def update(event, **kwargs):
    print(event, kwargs)


def set_cell(pos, value):
    col, row = pos
    cells[row][col] = value


def get_cell(pos):
    col, row = pos
    return cells[row][col]


def add(pos):
    set_cell(pos, color_key)
    update('add', pos=pos, color=colors[color_key])


def remove(pos):
    set_cell(pos, ' ')
    update('remove', pos=pos)


def test():
    keys = 'rgbyok'
    for i in range(N*N):
        c = keys[i % 6]
        set_color_key(c)
        row, col = divmod(i, N)
        add((col, row))

    for i in range(5):
        remove((i, i))


test()
cells
```

```python
def is_inside(pt):
    x, y = pt
    return 0 <= x < N and 0 <= y < N


def get_next(pt0):
    x, y = pt0
    pts = [(x+1, y), (x-1, y), (x, y+1), (x, y-1)]
    return [pt for pt in pts if is_inside(pt) and get_cell(pt) == get_cell(pt0)]


def get_component(start, component):
    pts = {start}
    while pts:
        pt0 = pts.pop()
        component.add(pt0)
        for pt in get_next(pt0):
            if pt not in component:
                pts.add(pt)
```

```python
def get_all_components():
    components = {}
    pts = set((x, y) for x in range(N) for y in range(N))
    while pts:
        start = pts.pop()
        component = set()
        get_component(start, component)
        pts -= component
        key = get_cell(start)
        components.setdefault(key, []).append(component)

    return components
```
